## 📚 Краткая теория

### 🧬 **Наследование**
Механизм создания нового класса на основе существующего. Позволяет переиспользовать код и создавать иерархии классов.

```python
class Родитель:
    # атрибуты и методы родителя

class Ребенок(Родитель):  # Наследование
    # + дополнительные атрибуты и методы
```

### 🔄 **Переопределение методов**
Дочерний класс может заменить реализацию метода родительского класса, сохранив его название.

### ⚡ **super()**
Функция для доступа к методам родительского класса из дочернего. Особенно полезна при переопределении методов.

```python
class Дочерний(Родитель):
    def метод(self):
        super().метод()  # Вызов родительского метода
        # Дополнительная логика
```

### 🦆 **Утиная типизация**
"Если что-то ходит как утка и крякает как утка, то это утка". В Python тип объекта определяется его поведением (методами), а не явным указанием класса.

```python
def функция(объект):
    объект.метод()  # Не важно какой класс, главное — чтобы был метод
```

### 🎭 **Полиморфизм**
Возможность использовать объекты разных классов через одинаковый интерфейс. Разные классы могут иметь методы с одинаковыми именами, но разной реализацией.

- `Dog` — переопределяет `make_sound()` на `"Woof!"`

#### **Задача 5: Геометрические фигуры** 📐
Создайте базовый класс `Shape` с методами:
- `area()` → возвращает 0
- `description()` → `"This is a shape"`

Наследники:
- `Rectangle(width, height)` — переопределяет `area()` (площадь = width × height)
- `Circle(radius)` — переопределяет `area()` (площадь = 3.14 × radius²)

Напишите функцию `print_area(shape)`, которая принимает любой объект типа `Shape` (или наследника) и выводит его описание и площадь.

**Демонстрация полиморфизма:**
```python
shapes = [Rectangle(4, 5), Circle(3)]
for shape in shapes:
    print_area(shape)
```
