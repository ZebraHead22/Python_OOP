## üß© –†–∞–∑–±–æ—Ä –∑–∞–¥–∞—á —Å –ø–æ–¥—Ä–æ–±–Ω—ã–º–∏ –ø–æ—è—Å–Ω–µ–Ω–∏—è–º–∏

### –ó–∞–¥–∞—á–∞ 1-—Ä–∞–∑–±–æ—Ä: "–£–º–Ω—ã–π —Å–≤–µ—Ç–æ—Ñ–æ—Ä"

**–ó–∞–¥–∞—á–∞:** –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å `TrafficLight` (—Å–≤–µ—Ç–æ—Ñ–æ—Ä), –∫–æ—Ç–æ—Ä—ã–π:
1. –ò–º–µ–µ—Ç —Ç—Ä–∏ —Ü–≤–µ—Ç–∞: –∫—Ä–∞—Å–Ω—ã–π, –∂—ë–ª—Ç—ã–π, –∑–µ–ª—ë–Ω—ã–π
2. –ú–æ–∂–µ—Ç –ø–µ—Ä–µ–∫–ª—é—á–∞—Ç—å—Å—è –º–µ–∂–¥—É —Ü–≤–µ—Ç–∞–º–∏
3. –•—Ä–∞–Ω–∏—Ç —Ç–µ–∫—É—â–µ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ
4. –°—á–∏—Ç–∞–µ—Ç, —Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –ø–µ—Ä–µ–∫–ª—é—á–∏–ª—Å—è

**–ü–æ—à–∞–≥–æ–≤–æ–µ —Ä–µ—à–µ–Ω–∏–µ —Å –æ–±—ä—è—Å–Ω–µ–Ω–∏—è–º–∏:**

````python
class TrafficLight:
    """
    –®–∞–≥ 2: –û–ø—Ä–µ–¥–µ–ª—è–µ–º –∞—Ç—Ä–∏–±—É—Ç—ã –∫–ª–∞—Å—Å–∞
    –ê—Ç—Ä–∏–±—É—Ç—ã –∫–ª–∞—Å—Å–∞ - –æ–±—â–∏–µ –¥–ª—è –≤—Å–µ—Ö –æ–±—ä–µ–∫—Ç–æ–≤.
    COLORS - —ç—Ç–æ —Å–ø–∏—Å–æ–∫ –≤–æ–∑–º–æ–∂–Ω—ã—Ö —Ü–≤–µ—Ç–æ–≤. –û–Ω –æ–¥–∏–Ω–∞–∫–æ–≤ –¥–ª—è –≤—Å–µ—Ö —Å–≤–µ—Ç–æ—Ñ–æ—Ä–æ–≤.
    """
    COLORS = ["–∫—Ä–∞—Å–Ω—ã–π", "–∂—ë–ª—Ç—ã–π", "–∑–µ–ª—ë–Ω—ã–π"]
    
    def __init__(self):
        """
        –®–∞–≥ 3: –°–æ–∑–¥–∞—ë–º –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä __init__
        –ö–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä –≤—ã–∑—ã–≤–∞–µ—Ç—Å—è –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ –æ–±—ä–µ–∫—Ç–∞.
        self - —Å—Å—ã–ª–∫–∞ –Ω–∞ —Å–æ–∑–¥–∞–≤–∞–µ–º—ã–π –æ–±—ä–µ–∫—Ç.
        """
        self.current_color = "–∫—Ä–∞—Å–Ω—ã–π"  # –ê—Ç—Ä–∏–±—É—Ç —ç–∫–∑–µ–º–ø–ª—è—Ä–∞
        self.switch_count = 0           # –ê—Ç—Ä–∏–±—É—Ç —ç–∫–∑–µ–º–ø–ª—è—Ä–∞
        print(f"–°–æ–∑–¥–∞–Ω —Å–≤–µ—Ç–æ—Ñ–æ—Ä. –ù–∞—á–∞–ª—å–Ω—ã–π —Ü–≤–µ—Ç: {self.current_color}")
    
    def switch_color(self):
        """
        –®–∞–≥ 4: –°–æ–∑–¥–∞—ë–º –º–µ—Ç–æ–¥ –¥–ª—è –ø–µ—Ä–µ–∫–ª—é—á–µ–Ω–∏—è
        –ú–µ—Ç–æ–¥ - —ç—Ç–æ —Ñ—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç –æ–±—ä–µ–∫—Ç—É.
        –ü–µ—Ä–≤—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä –í–°–ï–ì–î–ê self.
        """
        current_index = TrafficLight.COLORS.index(self.current_color)
        next_index = (current_index + 1) % len(TrafficLight.COLORS)
        next_color = TrafficLight.COLORS[next_index]
        
        old_color = self.current_color
        self.current_color = next_color
        self.switch_count += 1
        
        print(f"–°–≤–µ—Ç–æ—Ñ–æ—Ä –ø–µ—Ä–µ–∫–ª—é—á–∏–ª—Å—è: {old_color} ‚Üí {next_color}")
        print(f"–í—Å–µ–≥–æ –ø–µ—Ä–µ–∫–ª—é—á–µ–Ω–∏–π: {self.switch_count}")
    
    def get_status(self):
        """–®–∞–≥ 5: –°–æ–∑–¥–∞—ë–º –º–µ—Ç–æ–¥ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏"""
        return f"–¢–µ–∫—É—â–∏–π —Ü–≤–µ—Ç: {self.current_color}, –ø–µ—Ä–µ–∫–ª—é—á–µ–Ω–∏–π: {self.switch_count}"

# –®–∞–≥ 6: –ò—Å–ø–æ–ª—å–∑—É–µ–º –Ω–∞—à –∫–ª–∞—Å—Å
print("=== –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è —Ä–∞–±–æ—Ç—ã —Å–≤–µ—Ç–æ—Ñ–æ—Ä–∞ ===")
light1 = TrafficLight()
light1.switch_color()
light1.switch_color()
light1.switch_color()
print(f"\n–°—Ç–∞—Ç—É—Å —Å–≤–µ—Ç–æ—Ñ–æ—Ä–∞: {light1.get_status()}")

print("\n=== –°–æ–∑–¥–∞—ë–º –≤—Ç–æ—Ä–æ–π —Å–≤–µ—Ç–æ—Ñ–æ—Ä ===")
light2 = TrafficLight()
light2.switch_color()
print(f"–ü–µ—Ä–≤—ã–π —Å–≤–µ—Ç–æ—Ñ–æ—Ä: {light1.get_status()}")
print(f"–í—Ç–æ—Ä–æ–π —Å–≤–µ—Ç–æ—Ñ–æ—Ä: {light2.get_status()}")
````

### –ó–∞–¥–∞—á–∞ 2-—Ä–∞–∑–±–æ—Ä: "–ë–∞–Ω–∫–æ–≤—Å–∫–∏–π —Å—á—ë—Ç —Å –∏—Å—Ç–æ—Ä–∏–µ–π"

**–ó–∞–¥–∞—á–∞:** –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å `BankAccount`, –∫–æ—Ç–æ—Ä—ã–π:
1. –•—Ä–∞–Ω–∏—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –≤–ª–∞–¥–µ–ª—å—Ü–µ –∏ –±–∞–ª–∞–Ω—Å–µ
2. –í–µ–¥—ë—Ç –∏—Å—Ç–æ—Ä–∏—é –≤—Å–µ—Ö –æ–ø–µ—Ä–∞—Ü–∏–π
3. –ü–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ—Ç –ø–æ–ø–æ–ª–Ω–µ–Ω–∏–µ, —Å–Ω—è—Ç–∏–µ –∏ –ø–µ—Ä–µ–≤–æ–¥
4. –°—á–∏—Ç–∞–µ—Ç –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—á–µ—Ç–æ–≤ –≤ –±–∞–Ω–∫–µ

**–†–µ—à–µ–Ω–∏–µ —Å –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏—è–º–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —à–∞–≥–∞:**

````python
class BankAccount:
    """
    –ê—Ç—Ä–∏–±—É—Ç—ã –∫–ª–∞—Å—Å–∞:
    - total_accounts: –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–∑–¥–∞–Ω–Ω—ã—Ö —Å—á–µ—Ç–æ–≤
    - bank_name: –Ω–∞–∑–≤–∞–Ω–∏–µ –±–∞–Ω–∫–∞ (–æ–±—â–µ–µ –¥–ª—è –≤—Å–µ—Ö —Å—á–µ—Ç–æ–≤)
    - interest_rate: –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–∞—è —Å—Ç–∞–≤–∫–∞
    """
    total_accounts = 0
    bank_name = "Python Bank"
    interest_rate = 0.05  # 5% –≥–æ–¥–æ–≤—ã—Ö
    
    def __init__(self, owner, initial_balance=0):
        """
        –ö–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä –æ–±—ä–µ–∫—Ç–∞.
        –í—ã–∑—ã–≤–∞–µ—Ç—Å—è –ø—Ä–∏ —Å–æ–∑–¥–∞–Ω–∏–∏ –∫–∞–∂–¥–æ–≥–æ –Ω–æ–≤–æ–≥–æ —Å—á—ë—Ç–∞.
        """
        BankAccount.total_accounts += 1
        self.account_number = BankAccount.total_accounts
        self.owner = owner
        self.balance = initial_balance
        self.transaction_history = []
        
        if initial_balance > 0:
            self._add_to_history("initial_deposit", initial_balance)
        
        print(f"–°–æ–∑–¥–∞–Ω —Å—á—ë—Ç ‚Ññ{self.account_number} –¥–ª—è {self.owner}")
        print(f"–ù–∞—á–∞–ª—å–Ω—ã–π –±–∞–ª–∞–Ω—Å: {self.balance}")
    
    def _add_to_history(self, transaction_type, amount):
        """
        –í–Ω—É—Ç—Ä–µ–Ω–Ω–∏–π –º–µ—Ç–æ–¥ (–Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å _) –¥–ª—è –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –≤ –∏—Å—Ç–æ—Ä–∏—é.
        –ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –≤–Ω—É—Ç—Ä–∏ –∫–ª–∞—Å—Å–∞.
        """
        transaction = {
            "type": transaction_type,
            "amount": amount,
            "balance_after": self.balance
        }
        self.transaction_history.append(transaction)
    
    def deposit(self, amount):
        """–ü–æ–ø–æ–ª–Ω–µ–Ω–∏–µ —Å—á—ë—Ç–∞."""
        if amount <= 0:
            return "–û—à–∏–±–∫–∞: —Å—É–º–º–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–π"
        
        self.balance += amount
        self._add_to_history("deposit", amount)
        
        print(f"–°—á—ë—Ç {self.account_number}: –ø–æ–ø–æ–ª–Ω–µ–Ω–∏–µ +{amount}")
        print(f"–ù–æ–≤—ã–π –±–∞–ª–∞–Ω—Å: {self.balance}")
        
        return self.balance
    
    def withdraw(self, amount):
        """–°–Ω—è—Ç–∏–µ —Å–æ —Å—á—ë—Ç–∞."""
        if amount <= 0:
            return "–û—à–∏–±–∫–∞: —Å—É–º–º–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–π"
        
        if amount > self.balance:
            return "–û—à–∏–±–∫–∞: –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Å—Ä–µ–¥—Å—Ç–≤"
        
        self.balance -= amount
        self._add_to_history("withdrawal", -amount)
        
        print(f"–°—á—ë—Ç {self.account_number}: —Å–Ω—è—Ç–∏–µ -{amount}")
        print(f"–ù–æ–≤—ã–π –±–∞–ª–∞–Ω—Å: {self.balance}")
        
        return self.balance
    
    def transfer(self, target_account, amount):
        """–ü–µ—Ä–µ–≤–æ–¥ –Ω–∞ –¥—Ä—É–≥–æ–π —Å—á—ë—Ç."""
        print(f"\n=== –ü–µ—Ä–µ–≤–æ–¥ —Å–æ —Å—á—ë—Ç–∞ {self.account_number} –Ω–∞ —Å—á—ë—Ç {target_account.account_number} ===")
        
        result = self.withdraw(amount)
        
        if isinstance(result, (int, float)):
            target_account.deposit(amount)
            self._add_to_history("transfer_out", -amount)
            target_account._add_to_history("transfer_in", amount)
            return f"–ü–µ—Ä–µ–≤–æ–¥ —É—Å–ø–µ—à–µ–Ω. –í–∞—à –±–∞–ª–∞–Ω—Å: {self.balance}"
        else:
            return f"–û—à–∏–±–∫–∞ –ø–µ—Ä–µ–≤–æ–¥–∞: {result}"
    
    def calculate_interest(self):
        """–†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç –ø—Ä–æ—Ü–µ–Ω—Ç—ã –ø–æ —Å—á—ë—Ç—É."""
        interest = self.balance * BankAccount.interest_rate
        print(f"–ù–∞—á–∏—Å–ª–µ–Ω—ã –ø—Ä–æ—Ü–µ–Ω—Ç—ã: {interest}")
        return interest
    
    def get_statement(self, last_n=None):
        """–ü–æ–ª—É—á–∏—Ç—å –≤—ã–ø–∏—Å–∫—É –ø–æ —Å—á—ë—Ç—É."""
        statement = f"\n=== –í—ã–ø–∏—Å–∫–∞ –ø–æ —Å—á—ë—Ç—É ‚Ññ{self.account_number} ===\n"
        statement += f"–í–ª–∞–¥–µ–ª–µ—Ü: {self.owner}\n"
        statement += f"–ë–∞–ª–∞–Ω—Å: {self.balance}\n"
        statement += f"–ë–∞–Ω–∫: {BankAccount.bank_name}\n"
        
        statement += "\n–ò—Å—Ç–æ—Ä–∏—è –æ–ø–µ—Ä–∞—Ü–∏–π:\n"
        
        if last_n:
            transactions = self.transaction_history[-last_n:]
        else:
            transactions = self.transaction_history
        
        for i, t in enumerate(transactions, 1):
            if t["amount"] >= 0:
                operation = f"+{t['amount']}"
            else:
                operation = f"{t['amount']}"
            
            statement += f"{i}. {t['type']}: {operation}, –±–∞–ª–∞–Ω—Å: {t['balance_after']}\n"
        
        return statement
    
    @classmethod
    def get_bank_info(cls):
        """
        –ú–µ—Ç–æ–¥ –∫–ª–∞—Å—Å–∞ (—Ä–∞–±–æ—Ç–∞–µ—Ç –Ω–∞ —É—Ä–æ–≤–Ω–µ –∫–ª–∞—Å—Å–∞, –∞ –Ω–µ –æ–±—ä–µ–∫—Ç–∞).
        cls - —ç—Ç–æ —Å—Å—ã–ª–∫–∞ –Ω–∞ –∫–ª–∞—Å—Å (–∞–Ω–∞–ª–æ–≥ self –¥–ª—è –º–µ—Ç–æ–¥–æ–≤ –∫–ª–∞—Å—Å–∞).
        """
        info = f"\n=== –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –±–∞–Ω–∫–µ ===\n"
        info += f"–ù–∞–∑–≤–∞–Ω–∏–µ: {cls.bank_name}\n"
        info += f"–í—Å–µ–≥–æ —Å—á–µ—Ç–æ–≤: {cls.total_accounts}\n"
        info += f"–ü—Ä–æ—Ü–µ–Ω—Ç–Ω–∞—è —Å—Ç–∞–≤–∫–∞: {cls.interest_rate * 100}%\n"
        return info

# –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è —Ä–∞–±–æ—Ç—ã
print("=" * 50)
print("–°–û–ó–î–ê–ù–ò–ï –°–ß–ï–¢–û–í")
print("=" * 50)

account1 = BankAccount("–ò–≤–∞–Ω –ò–≤–∞–Ω–æ–≤", 1000)
print()
account2 = BankAccount("–ê–Ω–Ω–∞ –ü–µ—Ç—Ä–æ–≤–∞")
print()

print("=" * 50)
print("–û–ü–ï–†–ê–¶–ò–ò –°–û –°–ß–ï–¢–ê–ú–ò")
print("=" * 50)

account1.deposit(500)
print()
account1.withdraw(200)
print()
account1.transfer(account2, 300)
print()

print("=" * 50)
print("–ò–ù–§–û–†–ú–ê–¶–ò–Ø –û –°–ß–ï–¢–ê–•")
print("=" * 50)

print(account1.get_statement(last_n=3))
print(account2.get_statement())

print("=" * 50)
print("–†–ê–ë–û–¢–ê –° –ê–¢–†–ò–ë–£–¢–ê–ú–ò –ö–õ–ê–°–°–ê")
print("=" * 50)

print(BankAccount.get_bank_info())
BankAccount.interest_rate = 0.07
print(f"–ù–æ–≤–∞—è –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–∞—è —Å—Ç–∞–≤–∫–∞: {BankAccount.interest_rate * 100}%")
````

### –ó–∞–¥–∞—á–∞ 2: "–ë–∏–±–ª–∏–æ—Ç–µ–∫–∞ –∫–Ω–∏–≥"
**–£—Ä–æ–≤–µ–Ω—å:** –ù–∞—á–∞–ª—å–Ω—ã–π

–°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å `Book` –∏ –∫–ª–∞—Å—Å `Library`:
1. –ö–ª–∞—Å—Å `Book` —Ö—Ä–∞–Ω–∏—Ç: –Ω–∞–∑–≤–∞–Ω–∏–µ, –∞–≤—Ç–æ—Ä–∞, –≥–æ–¥, —Å—Ç–∞—Ç—É—Å (–¥–æ—Å—Ç—É–ø–Ω–∞/–≤—ã–¥–∞–Ω–∞)
2. –ö–ª–∞—Å—Å `Library` —É–ø—Ä–∞–≤–ª—è–µ—Ç –∫–Ω–∏–≥–∞–º–∏: –¥–æ–±–∞–≤–ª—è–µ—Ç, —É–¥–∞–ª—è–µ—Ç, –∏—â–µ—Ç, –≤—ã–¥–∞—ë—Ç
3. –î–æ–±–∞–≤—å—Ç–µ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –ø–æ–∏—Å–∫–∞ –ø–æ –∞–≤—Ç–æ—Ä—É –∏ –ø–æ –≥–æ–¥—É
4. –í–µ–¥–∏—Ç–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É: —Å–∫–æ–ª—å–∫–æ –∫–Ω–∏–≥, —Å–∫–æ–ª—å–∫–æ –≤—ã–¥–∞–Ω–æ

**–ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è:**






````python
library = Library("–ì–ª–∞–≤–Ω–∞—è –±–∏–±–ª–∏–æ—Ç–µ–∫–∞")

book1 = Book("–ü—Ä–µ—Å—Ç—É–ø–ª–µ–Ω–∏–µ –∏ –Ω–∞–∫–∞–∑–∞–Ω–∏–µ", "–î–æ—Å—Ç–æ–µ–≤—Å–∫–∏–π", 1866)
book2 = Book("–í–æ–π–Ω–∞ –∏ –º–∏—Ä", "–¢–æ–ª—Å—Ç–æ–π", 1869)

library.add_book(book1)
library.add_book(book2)

library.borrow_book("–í–æ–π–Ω–∞ –∏ –º–∏—Ä")
print(library.get_available_books())  # –¢–æ–ª—å–∫–æ "–ü—Ä–µ—Å—Ç—É–ø–ª–µ–Ω–∏–µ –∏ –Ω–∞–∫–∞–∑–∞–Ω–∏–µ"

library.return_book("–í–æ–π–Ω–∞ –∏ –º–∏—Ä")
print(library.find_by_author("–¢–æ–ª—Å—Ç–æ–π"))  # –ù–∞—Ö–æ–¥–∏—Ç –∫–Ω–∏–≥—É
````

### –ó–∞–¥–∞—á–∞ 3: "–§–∏–≥—É—Ä—ã –Ω–∞ –ø–ª–æ—Å–∫–æ—Å—Ç–∏"
**–£—Ä–æ–≤–µ–Ω—å:** –ù–∞—á–∞–ª—å–Ω—ã–π

–°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å—ã –¥–ª—è –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏—Ö —Ñ–∏–≥—É—Ä:
1. `Point` ‚Äî —Ç–æ—á–∫–∞ —Å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏ (x, y)
2. `Circle` ‚Äî –æ–∫—Ä—É–∂–Ω–æ—Å—Ç—å (—Ü–µ–Ω—Ç—Ä –∏ —Ä–∞–¥–∏—É—Å)
3. `Rectangle` ‚Äî –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ (–¥–≤–µ —Ç–æ—á–∫–∏)

````python
point1 = Point(0, 0)
point2 = Point(3, 4)

circle = Circle(point1, 5)
rectangle = Rectangle(point1, point2)

print(f"–†–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏: {point1.distance_to(point2)}")
print(f"–ü–ª–æ—â–∞–¥—å –∫—Ä—É–≥–∞: {circle.area():.2f}")
print(f"–ü–µ—Ä–∏–º–µ—Ç—Ä –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–∞: {rectangle.perimeter()}")

# –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ:
print(f"–ü–µ—Ä–µ—Å–µ–∫–∞—é—Ç—Å—è –ª–∏ –∫—Ä—É–≥ –∏ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫? {circle.intersects(rectangle)}")
````

### –ó–∞–¥–∞—á–∞ 5: "–°–∏—Å—Ç–µ–º–∞ –±—Ä–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏—è"
**–£—Ä–æ–≤–µ–Ω—å:** –°—Ä–µ–¥–Ω–∏–π

–°–æ–∑–¥–∞–π—Ç–µ —Å–∏—Å—Ç–µ–º—É –±—Ä–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏—è –±–∏–ª–µ—Ç–æ–≤:
1. `Event` ‚Äî –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–µ (–Ω–∞–∑–≤–∞–Ω–∏–µ, –¥–∞—Ç–∞, –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–µ—Å—Ç)
2. `Ticket` ‚Äî –±–∏–ª–µ—Ç (–Ω–æ–º–µ—Ä, –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏–µ, —Å—Ç–∞—Ç—É—Å: —Å–≤–æ–±–æ–¥–µ–Ω/–∑–∞–±—Ä–æ–Ω–∏—Ä–æ–≤–∞–Ω/–ø—Ä–æ–¥–∞–Ω)
3. `BookingSystem` ‚Äî —Å–∏—Å—Ç–µ–º–∞ –±—Ä–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏—è

**–û—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏:**
- –ë–∏–ª–µ—Ç—ã –Ω–µ —Å–æ–∑–¥–∞—é—Ç—Å—è –∑–∞—Ä–∞–Ω–µ–µ, –∞ –≥–µ–Ω–µ—Ä–∏—Ä—É—é—Ç—Å—è –ø—Ä–∏ –±—Ä–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏–∏
- –ü–æ–¥–¥–µ—Ä–∂–∫–∞ —Ä–∞–∑–Ω—ã—Ö —Ç–∏–ø–æ–≤ –±–∏–ª–µ—Ç–æ–≤ (–æ–±—ã—á–Ω—ã–π, VIP, —Å—Ç—É–¥–µ–Ω—á–µ—Å–∫–∏–π)
- –û—á–µ—Ä–µ–¥—å –Ω–∞ –æ–∂–∏–¥–∞–Ω–∏–µ, –µ—Å–ª–∏ –≤—Å–µ –±–∏–ª–µ—Ç—ã –ø—Ä–æ–¥–∞–Ω—ã
- –ò—Å—Ç–æ—Ä–∏—è –±—Ä–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è

````python
concert = Event("–ö–æ–Ω—Ü–µ—Ä—Ç", "2024-12-31", 100)
system = BookingSystem()

# –ë—Ä–æ–Ω–∏—Ä—É–µ–º –±–∏–ª–µ—Ç—ã
ticket1 = system.book_ticket(concert, "–æ–±—ã—á–Ω—ã–π", "–ò–≤–∞–Ω")
ticket2 = system.book_ticket(concert, "VIP", "–ê–Ω–Ω–∞")

# –û—Ç–º–µ–Ω—è–µ–º –±—Ä–æ–Ω—å
system.cancel_booking(ticket1)

# –ü–æ–∫—É–ø–∞–µ–º –±–∏–ª–µ—Ç
system.purchase_ticket(ticket2)

# –°–º–æ—Ç—Ä–∏–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É
print(f"–°–≤–æ–±–æ–¥–Ω–æ –º–µ—Å—Ç: {concert.available_seats}")
print(f"–í—ã—Ä—É—á–∫–∞: {system.total_revenue}")
````

  - –ü–æ–∫–∞–∑–∞—Ç—å –∏–Ω–≤–µ–Ω—Ç–∞—Ä—å

**–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ:**
1. –†–µ–∞–ª–∏–∑—É–π—Ç–µ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ –ø–æ –≤–µ—Å—É –∏–Ω–≤–µ–Ω—Ç–∞—Ä—è
