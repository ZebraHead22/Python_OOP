Реализуйте два плеера:
- `AudioPlayer`: поддерживает `mp3`, `wav`. При `play()` выводит «Воспроизведение аудио: <имя файла>».
- `VideoPlayer`: поддерживает `mp4`, `avi`. При `play()` выводит «Воспроизведение видео: <имя файла>».

```python
from abc import ABC, abstractmethod

class MediaPlayer(ABC):
    @property
    @abstractmethod
    def supported_extensions(self) -> list[str]:
        pass

    @abstractmethod
    def play(self) -> None:
        pass

    @abstractmethod
    def pause(self) -> None:
        pass

    @abstractmethod
    def stop(self) -> None:
        pass

    def play_file(self, filename: str):
        ext = filename.split('.')[-1]
        if ext in self.supported_extensions:
            self.play()
        else:
            print(f"Формат {ext} не поддерживается")

# Реализуйте AudioPlayer и VideoPlayer
# TODO

# Проверка
# ap = AudioPlayer("song.mp3")
# ap.play_file("song.mp3")
# ap.play_file("video.mp4")  # ошибка
```

```python
from typing import Protocol

class WithArea(Protocol):
    def area(self) -> float: ...

class WithPerimeter(Protocol):
    def perimeter(self) -> float: ...

# Протокол, объединяющий оба (можно использовать как тип)
class ShapeProtocol(WithArea, WithPerimeter, Protocol):
    pass

def describe_shape(shape: ShapeProtocol) -> None:
    print(f"Площадь: {shape.area()}, Периметр: {shape.perimeter()}")

class Rectangle:
    def __init__(self, width: float, height: float):
        self.width = width
        self.height = height

    def area(self) -> float:
        return self.width * self.height

    def perimeter(self) -> float:
        return 2 * (self.width + self.height)

class Circle:
    def __init__(self, radius: float):
        self.radius = radius

    def area(self) -> float:
        return 3.14159 * self.radius ** 2

    def perimeter(self) -> float:
        return 2 * 3.14159 * self.radius

# Проверка
# rect = Rectangle(3, 4)
# circle = Circle(5)
# describe_shape(rect)
# describe_shape(circle)
```

```python
# Шаблон
from abc import ABC, abstractmethod

class DatabaseConnection(ABC):
    @abstractmethod
    def connect(self) -> None:
        pass

    @abstractmethod
    def execute(self, query: str) -> list[tuple]:
        pass

    @abstractmethod
    def disconnect(self) -> None:
        pass

# Реализуйте классы
# TODO

def run_queries(db: DatabaseConnection, queries: list[str]):
    db.connect()
    try:
        for q in queries:
            result = db.execute(q)
            print(f"Результат '{q}': {result}")
    finally:
        db.disconnect()

# Проверка
# sqlite = SQLiteConnection()
# run_queries(sqlite, ["SELECT * FROM users", "SELECT * FROM orders"])
```

### Задача 7. Протокол JSONConvertible
Определите протокол `JSONConvertible` с методами:
- `to_json(self) -> str`
- `from_json(cls, json_str: str) -> 'JSONConvertible'` (классовый метод)

Напишите функцию `save_to_file(obj: JSONConvertible, filename: str)`, которая сохраняет результат `to_json()` в файл, и функцию `load_from_file(cls, filename: str)`, которая читает JSON-строку и возвращает объект, созданный через `from_json`.

Создайте класс `Book` с полями `title`, `author`, `year`. Реализуйте для него протокол (сериализация в JSON и десериализация). Проверьте сохранение и загрузку.

```python
import json
from typing import Protocol, Type, TypeVar

T = TypeVar('T', bound='JSONConvertible')

class JSONConvertible(Protocol):
    def to_json(self) -> str:
        ...

    @classmethod
    def from_json(cls: Type[T], json_str: str) -> T:
        ...

def save_to_file(obj: JSONConvertible, filename: str) -> None:
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(obj.to_json())

def load_from_file(cls: Type[T], filename: str) -> T:
    with open(filename, 'r', encoding='utf-8') as f:
        json_str = f.read()
    return cls.from_json(json_str)

class Book:
    def __init__(self, title: str, author: str, year: int):
        self.title = title
        self.author = author
        self.year = year

    def to_json(self) -> str:
        data = {"title": self.title, "author": self.author, "year": self.year}
        return json.dumps(data, ensure_ascii=False)

    @classmethod
    def from_json(cls, json_str: str) -> 'Book':
        data = json.loads(json_str)
        return cls(data["title"], data["author"], data["year"])

# Проверка
# b = Book("Война и мир", "Толстой", 1869)
# save_to_file(b, "book.json")
# b2 = load_from_file(Book, "book.json")
# print(b2.title, b2.author, b2.year)
```

```python
import time
from abc import ABC, abstractmethod
from typing import Protocol

class SortingAlgorithm(ABC):
    @abstractmethod
    def sort(self, arr: list[int]) -> list[int]:
        pass

class Measurable(Protocol):
    def time_complexity(self) -> str:
        pass

# Реализуйте BubbleSort и QuickSort, унаследовав SortingAlgorithm
# и добавив метод time_complexity
# TODO

def compare_sorts(sorts: list[SortingAlgorithm], data: list[int]) -> None:
    for s in sorts:
        arr_copy = data.copy()
        start = time.perf_counter()
        result = s.sort(arr_copy)
        end = time.perf_counter()
        complexity = s.time_complexity() if isinstance(s, Measurable) else "unknown"
        print(f"{s.__class__.__name__}: сложность {complexity}, время {end-start:.6f} сек")

# Проверка
# test_data = [5, 2, 9, 1, 5, 6] * 100  # достаточно большой массив
# compare_sorts([BubbleSort(), QuickSort()], test_data)
```

## 📌 Полезные подсказки

- Для абстрактных свойств используйте связку `@property` и `@abstractmethod` (в любом порядке, но обычно `@property` сверху).
- Протоколы с `@runtime_checkable` позволяют использовать `isinstance`, но проверяют только наличие методов (не сигнатуры).
- При пересечении протоколов можно объявить новый протокол, унаследовав от нескольких: `class Combined(Proto1, Proto2, Protocol): pass`.
- Для замеров времени используйте `time.perf_counter()`, а не `time.time()` – выше точность.
