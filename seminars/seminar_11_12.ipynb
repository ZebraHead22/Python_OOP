## Задача 1. Банковский счёт (инкапсуляция, property)

## Задача 3. Класс-декоратор для логирования вызовов (__call__)

Создайте класс `Logger`, который принимает функцию в `__init__` и сохраняет её. Объект этого класса должен быть вызываемым (через `__call__`). При вызове он выводит сообщение "Вызов функции <имя> с аргументами <args>", затем вызывает оригинальную функцию и возвращает результат. Используйте этот декоратор для функции `add(a, b)`.

```python
@Logger
def add(a, b):
    return a + b

print(add(3, 5))   # Вызов функции add с аргументами (3, 5) -> 8
```

## Задача 4. Контекстный менеджер для временного изменения глобальной переменной (__enter__, __exit__)

Реализуйте класс `TempGlobal`, который в конструкторе принимает имя глобальной переменной (строку) и новое значение. При входе в блок `with` сохраняет старое значение и устанавливает новое, при выходе восстанавливает старое. Используйте `globals()`.

```python
x = 10
with TempGlobal("x", 100):
    print(x)   # 100
print(x)       # 10
```

## Задача 6. Магический список с доступом по индексу и срезу (__getitem__, __setitem__, __len__)

Создайте класс `MagicList`, который внутри хранит обычный список. Реализуйте:
- `__getitem__(self, index)` – возвращает элемент (поддерживает срезы)
- `__setitem__(self, index, value)` – изменяет элемент
- `__len__(self)` – возвращает длину
- `append(self, item)` – добавляет элемент в конец

```python
ml = MagicList()
ml.append(1); ml.append(2); ml.append(3)
print(ml[0])        # 1
ml[1] = 99
print(ml[1:3])      # [99, 3]
print(len(ml))      # 3
```

## Задача 10. Вызываемый класс для кэширования результатов (__call__ + __new__)

Реализуйте класс `CachedFunction`, который при создании принимает функцию (как декоратор). Объект должен быть вызываемым. При вызове с одинаковыми аргументами результат должен браться из кэша (словаря), а не перевычисляться. Используйте `__new__` для того, чтобы сделать декоратор работающим и для классов (не обязательно, но для практики). При первом вызове с новыми аргументами выводите "Вычисляем", при повторном – "Из кэша".

```python
@CachedFunction
def slow_square(x):
    import time
    time.sleep(1)
    return x * x

print(slow_square(4))   # вычисляем -> 16 (задержка)
print(slow_square(4))   # из кэша -> 16 (мгновенно)
