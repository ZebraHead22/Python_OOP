### –ó–∞–¥–∞—á–∞: "–°–µ–∫—Ä–µ—Ç–Ω—ã–π —è—â–∏–∫"
–°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å `SecretBox`, –∫–æ—Ç–æ—Ä—ã–π —Ö—Ä–∞–Ω–∏—Ç —Å–µ–∫—Ä–µ—Ç–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ.
- –°–æ–æ–±—â–µ–Ω–∏–µ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –∑–∞—â–∏—â–µ–Ω–æ –æ—Ç –ø—Ä—è–º–æ–≥–æ –¥–æ—Å—Ç—É–ø–∞
- –ú–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å —Å–æ–æ–±—â–µ–Ω–∏–µ —Ç–æ–ª—å–∫–æ —Å –ø—Ä–∞–≤–∏–ª—å–Ω—ã–º –ø–∞—Ä–æ–ª–µ–º
- –ú–æ–∂–Ω–æ –∏–∑–º–µ–Ω–∏—Ç—å —Å–æ–æ–±—â–µ–Ω–∏–µ —Ç–æ–ª—å–∫–æ —Å –ø—Ä–∞–≤–∏–ª—å–Ω—ã–º –ø–∞—Ä–æ–ª–µ–º

**–ü—Ä–æ—Å—Ç–∞—è –≤–µ—Ä—Å–∏—è ‚Äî —Å–¥–µ–ª–∞–µ–º –≤–º–µ—Å—Ç–µ:**

```python
# –†–∞–∑–±–∏—Ä–∞–µ–º –≤–º–µ—Å—Ç–µ: –°–µ–∫—Ä–µ—Ç–Ω—ã–π —è—â–∏–∫
print("üîê –ü—Ä–æ—Å—Ç–∞—è –∑–∞–¥–∞—á–∞: –°–µ–∫—Ä–µ—Ç–Ω—ã–π —è—â–∏–∫\n")

class SecretBox:
    def __init__(self, password):
        # –ü—É–±–ª–∏—á–Ω—ã–π –∞—Ç—Ä–∏–±—É—Ç - –ø–∞—Ä–æ–ª—å
        self.password = password
        
        # –ó–∞—â–∏—â—ë–Ω–Ω—ã–π –∞—Ç—Ä–∏–±—É—Ç - —Å–æ–æ–±—â–µ–Ω–∏–µ (–æ–¥–Ω–æ –ø–æ–¥—á—ë—Ä–∫–∏–≤–∞–Ω–∏–µ)
        self._message = "–°–µ–∫—Ä–µ—Ç–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ"
        
        # –ü—Ä–∏–≤–∞—Ç–Ω—ã–π –∞—Ç—Ä–∏–±—É—Ç - —Å—á—ë—Ç—á–∏–∫ –ø–æ–ø—ã—Ç–æ–∫ (–¥–≤–∞ –ø–æ–¥—á—ë—Ä–∫–∏–≤–∞–Ω–∏—è)
        self.__attempts = 0
    
    def get_message(self, user_password):
        """–ü–æ–ª—É—á–∏—Ç—å —Å–æ–æ–±—â–µ–Ω–∏–µ, –µ—Å–ª–∏ –ø–∞—Ä–æ–ª—å –≤–µ—Ä–Ω—ã–π"""
        if user_password == self.password:
            print("–ü–∞—Ä–æ–ª—å –≤–µ—Ä–Ω—ã–π! –°–æ–æ–±—â–µ–Ω–∏–µ:", self._message)
            return self._message
        else:
            self.__attempts += 1
            print(f"–ù–µ–≤–µ—Ä–Ω—ã–π –ø–∞—Ä–æ–ª—å! –ü–æ–ø—ã—Ç–æ–∫: {self.__attempts}")
            return None
    
    def change_password(self, old_password, new_password):
        """–ò–∑–º–µ–Ω–∏—Ç—å –ø–∞—Ä–æ–ª—å"""
        if old_password == self.password:
            self.password = new_password
            print("–ü–∞—Ä–æ–ª—å –∏–∑–º–µ–Ω—ë–Ω!")
            return True
        print("–°—Ç–∞—Ä—ã–π –ø–∞—Ä–æ–ª—å –Ω–µ–≤–µ—Ä–µ–Ω!")
        return False

# –î–∞–≤–∞–π—Ç–µ —Å–æ–∑–¥–∞–¥–∏–º —è—â–∏–∫ –∏ –ø–æ–ø—Ä–æ–±—É–µ–º —Å –Ω–∏–º –ø–æ—Ä–∞–±–æ—Ç–∞—Ç—å
print("1. –°–æ–∑–¥–∞—ë–º —Å–µ–∫—Ä–µ—Ç–Ω—ã–π —è—â–∏–∫:")
my_box = SecretBox("1234")
print(f"   –ü–∞—Ä–æ–ª—å —è—â–∏–∫–∞: {my_box.password}")
print()

print("2. –ü—ã—Ç–∞–µ–º—Å—è –ø–æ–ª—É—á–∏—Ç—å —Å–æ–æ–±—â–µ–Ω–∏–µ:")
# –í–µ—Ä–Ω—ã–π –ø–∞—Ä–æ–ª—å
my_box.get_message("1234")

# –ù–µ–≤–µ—Ä–Ω—ã–π –ø–∞—Ä–æ–ª—å
my_box.get_message("0000")
print()

print("3. –ú–µ–Ω—è–µ–º –ø–∞—Ä–æ–ª—å:")
my_box.change_password("1234", "–Ω–æ–≤—ã–π–ø–∞—Ä–æ–ª—å")
print(f"   –ù–æ–≤—ã–π –ø–∞—Ä–æ–ª—å: {my_box.password}")
print()

print("4. –ß—Ç–æ —Ç–∞–∫–æ–µ _protected –∏ __private?")
print("   _message - –∑–∞—â–∏—â—ë–Ω–Ω—ã–π –∞—Ç—Ä–∏–±—É—Ç (–º–æ–∂–Ω–æ –ø—Ä–æ—á–∏—Ç–∞—Ç—å, –Ω–æ –ª—É—á—à–µ –Ω–µ —Ç—Ä–æ–≥–∞—Ç—å)")
print("   __attempts - –ø—Ä–∏–≤–∞—Ç–Ω—ã–π –∞—Ç—Ä–∏–±—É—Ç (Python —Å–∫—Ä—ã–ª –µ–≥–æ –∏–º—è)")
print()

print("5. –ü–æ–ø—Ä–æ–±—É–µ–º –ø–æ–ª—É—á–∏—Ç—å –¥–æ—Å—Ç—É–ø –∫ –∞—Ç—Ä–∏–±—É—Ç–∞–º –Ω–∞–ø—Ä—è–º—É—é:")
print(f"   my_box._message = '{my_box._message}'")
print("   –ú–æ–∂–Ω–æ –ø—Ä–æ—á–∏—Ç–∞—Ç—å, –Ω–æ –Ω–µ –Ω—É–∂–Ω–æ!")

try:
    print(f"   my_box.__attempts = {my_box.__attempts}")
except AttributeError:
    print("   –û—à–∏–±–∫–∞! __attempts –Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω –Ω–∞–ø—Ä—è–º—É—é")
    
print("\n–ü–æ–∑–¥—Ä–∞–≤–ª—è—é! –í—ã —Å–æ–∑–¥–∞–ª–∏ —Å–≤–æ–π –ø–µ—Ä–≤—ã–π –∫–ª–∞—Å—Å —Å –∏–Ω–∫–∞–ø—Å—É–ª—è—Ü–∏–µ–π! üéâ")
```

–°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å `AlarmClock`, –∫–æ—Ç–æ—Ä—ã–π:

### –í–æ–ø—Ä–æ—Å 2: –í —á—ë–º —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É @property –∏ –æ–±—ã—á–Ω—ã–º –º–µ—Ç–æ–¥–æ–º?
```python
# –° @property
class Example1:
    @property
    def value(self):
        return self._value

obj = Example1()
print(obj.value)  # –ë–µ–∑ —Å–∫–æ–±–æ–∫!

# –ë–µ–∑ @property  
class Example2:
    def get_value(self):
        return self._value

obj = Example2()
print(obj.get_value())  # –°–æ —Å–∫–æ–±–∫–∞–º–∏!
```
