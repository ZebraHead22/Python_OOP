# 📚 Теоретическое введение: Инкапсуляция в Python

## 🔐 Что такое инкапсуляция?

**Инкапсуляция** — это принцип объектно-ориентированного программирования (ООП), при котором:

* данные объекта защищаются от прямого изменения извне,
* доступ к данным происходит через методы,
* внутренняя реализация скрыта от пользователя класса.

Проще говоря:

> Мы прячем «внутренности» объекта и даём доступ только через контролируемый интерфейс.

---

## 🏗 Зачем это нужно?

Инкапсуляция помогает:

1. ✅ Защитить данные от случайного изменения
2. ✅ Контролировать корректность значений
3. ✅ Сделать код более предсказуемым
4. ✅ Упростить поддержку и масштабирование

Если любой код может напрямую менять внутренние переменные — легко получить ошибку, которую трудно найти.

---

## 🐍 Инкапсуляция "по-питоньи"

В Python нет строгой приватности как в Java или C++.
Вместо этого используется **соглашение об именовании**.

### 1️⃣ Публичные атрибуты

```python
self.name = "Анна"
```

Можно свободно использовать извне.

---

### 2️⃣ Защищённые атрибуты (`_protected`)

```python
self._balance = 1000
```

Один символ `_` означает:

> "Это внутренняя переменная. Пожалуйста, не трогайте её напрямую."

Это не запрет — это договорённость между программистами.

---

### 3️⃣ Приватные атрибуты (`__private`)

```python
self.__password = "1234"
```

Два подчёркивания вызывают **name mangling** — Python автоматически переименует переменную:

```python
_ClassName__password
```

Это усложняет прямой доступ и защищает от случайного использования.

Важно:
Это **не абсолютная защита**, а механизм предотвращения случайных ошибок.

---

## 🎛 Контролируемый доступ: @property

Иногда нам нужно:

* разрешить читать значение,
* но запретить или контролировать его изменение.

Для этого используется `@property`.

### Пример:

```python
class User:
    def __init__(self, age):
        self._age = age
    
    @property
    def age(self):
        return self._age
    
    @age.setter
    def age(self, value):
        if value < 0:
            raise ValueError("Возраст не может быть отрицательным")
        self._age = value
```

Теперь:

```python
user.age = 25   # OK
user.age = -5   # Ошибка
```

Снаружи это выглядит как обычная переменная,
но внутри — работает логика проверки.

---

## 🏭 @classmethod и @staticmethod

В Python есть три типа методов:

### 1️⃣ Обычный метод

```python
def method(self):
```

Работает с конкретным объектом.

---

### 2️⃣ @classmethod

```python
@classmethod
def method(cls):
```

Работает с классом, а не с объектом.

Используется когда:

* нужно создать альтернативный конструктор,
* нужно работать с данными класса.

---

### 3️⃣ @staticmethod

```python
@staticmethod
def method():
```

Обычная функция внутри класса.

Используется когда:

* метод логически связан с классом,
* но не использует ни объект (`self`), ни класс (`cls`).

---

## 🧠 Главное понимание

Инкапсуляция — это не про «спрятать навсегда».
Это про:

* контроль доступа
* защиту логики
* ответственность класса за свои данные

## 🧩 Разбираем первую задачу вместе

### Задача: "Секретный ящик"
Создайте класс `SecretBox`, который хранит секретное сообщение.
- Сообщение должно быть защищено от прямого доступа
- Можно получить сообщение только с правильным паролем
- Можно изменить сообщение только с правильным паролем

**Простая версия — сделаем вместе:**

```python
# Разбираем вместе: Секретный ящик
print("🔐 Простая задача: Секретный ящик\n")

class SecretBox:
    def __init__(self, password):
        # Публичный атрибут - пароль
        self.password = password
        
        # Защищённый атрибут - сообщение (одно подчёркивание)
        self._message = "Секретное сообщение"
        
        # Приватный атрибут - счётчик попыток (два подчёркивания)
        self.__attempts = 0
    
    def get_message(self, user_password):
        """Получить сообщение, если пароль верный"""
        if user_password == self.password:
            print("Пароль верный! Сообщение:", self._message)
            return self._message
        else:
            self.__attempts += 1
            print(f"Неверный пароль! Попыток: {self.__attempts}")
            return None
    
    def change_password(self, old_password, new_password):
        """Изменить пароль"""
        if old_password == self.password:
            self.password = new_password
            print("Пароль изменён!")
            return True
        print("Старый пароль неверен!")
        return False

# Давайте создадим ящик и попробуем с ним поработать
print("1. Создаём секретный ящик:")
my_box = SecretBox("1234")
print(f"   Пароль ящика: {my_box.password}")
print()

print("2. Пытаемся получить сообщение:")
# Верный пароль
my_box.get_message("1234")

# Неверный пароль
my_box.get_message("0000")
print()

print("3. Меняем пароль:")
my_box.change_password("1234", "новыйпароль")
print(f"   Новый пароль: {my_box.password}")
print()

print("4. Что такое _protected и __private?")
print("   _message - защищённый атрибут (можно прочитать, но лучше не трогать)")
print("   __attempts - приватный атрибут (Python скрыл его имя)")
print()

print("5. Попробуем получить доступ к атрибутам напрямую:")
print(f"   my_box._message = '{my_box._message}'")
print("   Можно прочитать, но не нужно!")

try:
    print(f"   my_box.__attempts = {my_box.__attempts}")
except AttributeError:
    print("   Ошибка! __attempts недоступен напрямую")
    
print("\nПоздравляю! Вы создали свой первый класс с инкапсуляцией! 🎉")
```

## ❓ Частые вопросы и ошибки

### Вопрос 1: Зачем нужно self._атрибут если к нему всё равно можно получить доступ?
**Ответ:** `_атрибут` — это соглашение. Мы говорим другим программистам: "Это внутренняя переменная, лучше её не трогать напрямую". Python доверяет нам, что мы будем соблюдать это правило.

### Вопрос 2: В чём разница между @property и обычным методом?
```python
# С @property
class Example1:
    @property
    def value(self):
        return self._value

obj = Example1()
print(obj.value)  # Без скобок!

# Без @property  
class Example2:
    def get_value(self):
        return self._value

obj = Example2()
print(obj.get_value())  # Со скобками!
```

### Вопрос 3: Когда использовать @classmethod, а когда @staticmethod?
- **`@classmethod`** — когда нужен доступ к классу (например, для создания объекта)
- **`@staticmethod`** — когда не нужен доступ ни к чему (просто функция в классе)

### Частые ошибки:
1. **Забыли self в методах:**
```python
class Wrong:
    def method():  # ❌ Нет self!
        pass

class Right:
    def method(self):  # ✅ Есть self!
        pass
```

2. **Перепутали self и cls:**
```python
class Wrong:
    @classmethod
    def method(self):  # ❌ Должно быть cls!
        pass

class Right:
    @classmethod
    def method(cls):  # ✅ Правильно!
        pass
```

3. **Использовали property как метод:**
```python
obj = MyClass()
print(obj.property())  # ❌ Со скобками
print(obj.property)    # ✅ Без скобок
```
