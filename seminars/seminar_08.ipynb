### Задача 1: Калькулятор с историей
**Условие:** Создайте класс `Calculator`, который можно вызывать как функцию для выполнения математических операций. Калькулятор должен хранить историю операций.

**Требования:**
1. При вызове объекта с аргументами `(число1, операция, число2)` должен возвращаться результат
2. Поддерживаемые операции: `+`, `-`, `*`, `/`
3. Каждая операция должна сохраняться в истории
4. Реализуйте метод `show_history()` для вывода всех операций

```python
# Пример использования:
calc = Calculator()
print(calc(5, '+', 3))  # 8
print(calc(10, '*', 2)) # 20
calc.show_history()     # ['5 + 3 = 8', '10 * 2 = 20']

````python
class Calculator:

    _OPERATIONS = ['+', '-', '*', '/']

    def __init__(self):
        self.history = list()

    @staticmethod
    def values_validate(a, b):
        if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
            raise TypeError('A & B must be INT or FLOAT type!')
        
        return a, b
        
    def __call__(self, a, operator, b):
        a, b = self.values_validate(a, b)
        
        if operator not in self._OPERATIONS: 
            raise ValueError(f'Operator is not valid. Accessible operators store in {self._OPERATIONS}')
                             
        if operator == '+':
            self.history.append(f'{a} + {b} = {a+b}')
            return a + b

        if operator == '-':
            self.history.append(f'{a} - {b} = {a-b}')
            return a - b

        if operator == '*':
            self.history.append(f'{a} * {b} = {a*b}')
            return a * b

        if operator == '/':
        
            if b == 0:
                raise ArithmeticError('Divide by zero!')
            
            self.history.append(f'{a} / {b} = {a / b:.2f}')
            return f'{a / b:.2f}'
        
    def __len__(self):
        return len(self.history)    

    def show_history(self):
        print('History:')
        print('--'+'\n--'.join(self.history))
        # print(f'Count of operation: {len(self.history)}')
````

````python
try:
    calc = Calculator()
    print(calc(5, '+', 3))  # 8
    print(calc(10, '-', 7)) # 20
    print(calc(10, '*', 7)) # 20
    print(calc(10, '/', 7)) # 20
    
except Exception as e:
    print(f'ERROR: {e}')

finally:
    calc.show_history()     # ['5 + 3 = 8', '10 * 2 = 20']
    print(len(calc))
````

````python


# Шаблон для задачи 4
class Config:

    _allowed_keys = ['user', 'theme', 'language']

    def __init__(self):
        self.config = {}
        print('Create empty config file...')
        
    @staticmethod
    def key_validate(key):
        if key not in Config._allowed_keys:
                raise ValueError(f'Invalid key. Allowed keys is {Config._allowed_keys}')
        return key

    def load(self, settings):
        if not isinstance(settings, dict):
            raise TypeError('Setting nust be load at dictionary type!')
    
        for key in settings.keys():
            key = self.key_validate(key)
            
        if 'user' not in settings.keys():
            self.config = settings
            self.config['user'] = None    

        else:
            self.config = settings

        print(self.config)


    def __call__(self, key):
        return self.config[key]
    
    def __getitem__(self, key):
        return self.config[key]
    
    def __setitem__(self, key, value):
        key = self.key_validate(key)
        self.config[key] = value

# Проверка работы конфигурации
try:
    config = Config()
    config.load({"user": "test_user", "theme": 'dark', 'language': 'EN'})
    # config.load({"theme": 'dark', 'language': 'EN'})
    print(config("user"))
    print(config["user"])
except Exception as e:
    print(f'ERROR: {e}')
````
