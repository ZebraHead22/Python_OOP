````python
# –§—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω—ã–π –ø–æ–¥—Ö–æ–¥ (–±–µ–∑ –∫–ª–∞—Å—Å–æ–≤)
print("üìö –£–ø—Ä–∞–≤–ª–µ–Ω–∏–µ —Å—Ç—É–¥–µ–Ω—Ç–∞–º–∏ –±–µ–∑ –∫–ª–∞—Å—Å–æ–≤:")

# –°–æ–∑–¥–∞—ë–º —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –∫–∞–∫ —Å–ª–æ–≤–∞—Ä–∏
student1 = {
    "name": "–ê–Ω–Ω–∞",
    "age": 20,
    "grades": [4, 5, 3, 5]
}

student2 = {
    "name": "–ò–≤–∞–Ω", 
    "age": 21,
    "grades": [3, 4, 4, 5]
}

# –§—É–Ω–∫—Ü–∏–∏ –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å–æ —Å—Ç—É–¥–µ–Ω—Ç–∞–º–∏
def calculate_average(student):
    """–†–∞—Å—Å—á–∏—Ç–∞—Ç—å —Å—Ä–µ–¥–Ω–∏–π –±–∞–ª–ª —Å—Ç—É–¥–µ–Ω—Ç–∞."""
    grades = student["grades"]
    return sum(grades) / len(grades) if grades else 0

def add_grade(student, grade):
    """–î–æ–±–∞–≤–∏—Ç—å –æ—Ü–µ–Ω–∫—É —Å—Ç—É–¥–µ–Ω—Ç—É."""
    student["grades"].append(grade)

def student_info(student):
    """–ü–æ–ª—É—á–∏—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Å—Ç—É–¥–µ–Ω—Ç–µ."""
    avg = calculate_average(student)
    return f"{student['name']}, {student['age']} –ª–µ—Ç, —Å—Ä–µ–¥–Ω–∏–π –±–∞–ª–ª: {avg:.2f}"

# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ
print(f"–°—Ç—É–¥–µ–Ω—Ç 1: {student_info(student1)}")
add_grade(student1, 5)
print(f"–ü–æ—Å–ª–µ –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –æ—Ü–µ–Ω–∫–∏: {student_info(student1)}")

print(f"\n–°—Ç—É–¥–µ–Ω—Ç 2: {student_info(student2)}")

# –ü—Ä–æ–±–ª–µ–º—ã —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–≥–æ –ø–æ–¥—Ö–æ–¥–∞:
print("\n‚ö†Ô∏è –ü—Ä–æ–±–ª–µ–º—ã —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–≥–æ –ø–æ–¥—Ö–æ–¥–∞:")
print("1. –î–∞–Ω–Ω—ã–µ –∏ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞–∑–¥–µ–ª–µ–Ω—ã - –º–æ–∂–Ω–æ –∑–∞–±—ã—Ç—å, –∫–∞–∫–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞–±–æ—Ç–∞—é—Ç —Å –∫–∞–∫–∏–º–∏ –¥–∞–Ω–Ω—ã–º–∏")
print("2. –ù–µ—Ç –≥–∞—Ä–∞–Ω—Ç–∏–∏ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –¥–∞–Ω–Ω—ã—Ö")
print("3. –°–ª–æ–∂–Ω–æ –¥–æ–±–∞–≤–ª—è—Ç—å –Ω–æ–≤–æ–µ –ø–æ–≤–µ–¥–µ–Ω–∏–µ")
print("4. –ö–æ–¥ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –∑–∞–ø—É—Ç–∞–Ω–Ω—ã–º –ø—Ä–∏ —Ä–æ—Å—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—ã")

# –ü—Ä–∏–º–µ—Ä –ø—Ä–æ–±–ª–µ–º—ã: –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–∞—è —Å—Ç—Ä—É–∫—Ç—É—Ä–∞
student3 = {"–∏–º—è": "–ú–∞—Ä–∏—è", "–≤–æ–∑—Ä–∞—Å—Ç": 19}  # –û—à–∏–±–∫–∞ –≤ –∫–ª—é—á–∞—Ö!
print(f"\n–ü—Ä–æ–±–ª–µ–º–∞: —Å—Ç—É–¥–µ–Ω—Ç —Å –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–º–∏ –∫–ª—é—á–∞–º–∏ - {student3}")
print("–§—É–Ω–∫—Ü–∏–∏ –Ω–µ –±—É–¥—É—Ç —Ä–∞–±–æ—Ç–∞—Ç—å –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ!")
````

## üìä –ê—Ç—Ä–∏–±—É—Ç—ã —ç–∫–∑–µ–º–ø–ª—è—Ä–∞

### –•–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ –∞—Ç—Ä–∏–±—É—Ç–æ–≤ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞:
1. **–£–Ω–∏–∫–∞–ª—å–Ω–æ—Å—Ç—å** ‚Äî —É –∫–∞–∂–¥–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ —Å–≤–æ–∏ –∑–Ω–∞—á–µ–Ω–∏—è –∞—Ç—Ä–∏–±—É—Ç–æ–≤
2. **–ò–∑–º–µ–Ω—è–µ–º–æ—Å—Ç—å** ‚Äî –º–æ–∂–Ω–æ –∏–∑–º–µ–Ω—è—Ç—å –ø–æ—Å–ª–µ —Å–æ–∑–¥–∞–Ω–∏—è –æ–±—ä–µ–∫—Ç–∞
3. **–î–∏–Ω–∞–º–∏—á–Ω–æ—Å—Ç—å** ‚Äî –º–æ–∂–Ω–æ –¥–æ–±–∞–≤–ª—è—Ç—å –Ω–æ–≤—ã–µ –∞—Ç—Ä–∏–±—É—Ç—ã –≤ –ª—é–±–æ–π –º–æ–º–µ–Ω—Ç

### –ö–∞–∫ —Ä–∞–±–æ—Ç–∞—é—Ç –∞—Ç—Ä–∏–±—É—Ç—ã:
- –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞—é—Ç—Å—è –≤ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä–µ: `self.–∞—Ç—Ä–∏–±—É—Ç = –∑–Ω–∞—á–µ–Ω–∏–µ`
- –ú–æ–≥—É—Ç –±—ã—Ç—å –¥–æ–±–∞–≤–ª–µ–Ω—ã –ø–æ–∑–∂–µ: `–æ–±—ä–µ–∫—Ç.–Ω–æ–≤—ã–π_–∞—Ç—Ä–∏–±—É—Ç = –∑–Ω–∞—á–µ–Ω–∏–µ`
- –ß–∏—Ç–∞—é—Ç—Å—è —á–µ—Ä–µ–∑ —Ç–æ—á–∫—É: `–æ–±—ä–µ–∫—Ç.–∞—Ç—Ä–∏–±—É—Ç`

````python
# –†–∞–±–æ—Ç–∞ —Å –∞—Ç—Ä–∏–±—É—Ç–∞–º–∏ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞
print("üìù –†–∞–±–æ—Ç–∞ —Å –∞—Ç—Ä–∏–±—É—Ç–∞–º–∏ —ç–∫–∑–µ–º–ø–ª—è—Ä–∞:")

class Book:
    """–ö–ª–∞—Å—Å –¥–ª—è –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è –∫–Ω–∏–≥–∏."""
    
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_borrowed = False  # –ê—Ç—Ä–∏–±—É—Ç —Å–æ –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é

# –°–æ–∑–¥–∞—ë–º –∫–Ω–∏–≥–∏
book1 = Book("–ü—Ä–µ—Å—Ç—É–ø–ª–µ–Ω–∏–µ –∏ –Ω–∞–∫–∞–∑–∞–Ω–∏–µ", "–§—ë–¥–æ—Ä –î–æ—Å—Ç–æ–µ–≤—Å–∫–∏–π")
book2 = Book("–ú–∞—Å—Ç–µ—Ä –∏ –ú–∞—Ä–≥–∞—Ä–∏—Ç–∞", "–ú–∏—Ö–∞–∏–ª –ë—É–ª–≥–∞–∫–æ–≤")

print(f"–ö–Ω–∏–≥–∞ 1: {book1.title} - {book1.author}, –≤–∑—è—Ç–∞? {book1.is_borrowed}")
print(f"–ö–Ω–∏–≥–∞ 2: {book2.title} - {book2.author}, –≤–∑—è—Ç–∞? {book2.is_borrowed}")

# –ò–∑–º–µ–Ω—è–µ–º –∞—Ç—Ä–∏–±—É—Ç—ã
print("\nüìù –ò–∑–º–µ–Ω—è–µ–º –∞—Ç—Ä–∏–±—É—Ç—ã:")
book1.is_borrowed = True  # –ò–∑–º–µ–Ω—è–µ–º —Å—É—â–µ—Å—Ç–≤—É—é—â–∏–π –∞—Ç—Ä–∏–±—É—Ç
book1.year = 1866  # –î–æ–±–∞–≤–ª—è–µ–º –Ω–æ–≤—ã–π –∞—Ç—Ä–∏–±—É—Ç –¥–∏–Ω–∞–º–∏—á–µ—Å–∫–∏!

print(f"–ü–æ—Å–ª–µ –∏–∑–º–µ–Ω–µ–Ω–∏–π:")
print(f"  book1.is_borrowed = {book1.is_borrowed}")
print(f"  book1.year = {book1.year}")

# –£ book2 –Ω–µ—Ç –∞—Ç—Ä–∏–±—É—Ç–∞ year!
print(f"  book2.year —Å—É—â–µ—Å—Ç–≤—É–µ—Ç? {hasattr(book2, 'year')}")

# –î–∏–Ω–∞–º–∏—á–µ—Å–∫–æ–µ –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤? –í Python –º–æ–∂–Ω–æ, –Ω–æ –æ—Å—Ç–æ—Ä–æ–∂–Ω–æ!
print("\n‚ö†Ô∏è –î–∏–Ω–∞–º–∏—á–µ—Å–∫–æ–µ –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –∞—Ç—Ä–∏–±—É—Ç–æ–≤:")
book1.pages = 331
print(f"  –î–æ–±–∞–≤–∏–ª–∏ book1.pages = {book1.pages}")

# –ù–æ –ª—É—á—à–µ —Ç–∞–∫ –Ω–µ –¥–µ–ª–∞—Ç—å - —Ç–µ—Ä—è–µ—Ç—Å—è —Å—Ç—Ä—É–∫—Ç—É—Ä–∞
print("\nüí° –°–æ–≤–µ—Ç: –ª—É—á—à–µ –æ–ø—Ä–µ–¥–µ–ª—è—Ç—å –≤—Å–µ –∞—Ç—Ä–∏–±—É—Ç—ã –≤ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä–µ")
print("–î–∏–Ω–∞–º–∏—á–µ—Å–∫–æ–µ –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –∞—Ç—Ä–∏–±—É—Ç–æ–≤ –º–æ–∂–µ—Ç –∑–∞–ø—É—Ç–∞—Ç—å –∫–æ–¥!")

# –ü—Ä–∏–º–µ—Ä –ø—Ä–æ–±–ª–µ–º—ã
print("\n‚ùå –ü—Ä–∏–º–µ—Ä –ø—Ä–æ–±–ª–µ–º—ã –¥–∏–Ω–∞–º–∏—á–µ—Å–∫–∏—Ö –∞—Ç—Ä–∏–±—É—Ç–æ–≤:")

class Car:
    def __init__(self, model):
        self.model = model

car1 = Car("Toyota")
car2 = Car("BMW")

car1.color = "red"  # –î–æ–±–∞–≤–∏–ª–∏ —Ç–æ–ª—å–∫–æ car1
car2.year = 2020    # –î–æ–±–∞–≤–∏–ª–∏ —Ç–æ–ª—å–∫–æ car2

# –¢–µ–ø–µ—Ä—å –æ–±—ä–µ–∫—Ç—ã –æ–¥–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞ –∏–º–µ—é—Ç —Ä–∞–∑–Ω—ã–µ –∞—Ç—Ä–∏–±—É—Ç—ã!
print(f"  car1 –∏–º–µ–µ—Ç –∞—Ç—Ä–∏–±—É—Ç—ã: {[attr for attr in dir(car1) if not attr.startswith('__')]}")
print(f"  car2 –∏–º–µ–µ—Ç –∞—Ç—Ä–∏–±—É—Ç—ã: {[attr for attr in dir(car2) if not attr.startswith('__')]}")
````

## üõ†Ô∏è –ú–µ—Ç–æ–¥—ã —ç–∫–∑–µ–º–ø–ª—è—Ä–∞

**–ú–µ—Ç–æ–¥—ã —ç–∫–∑–µ–º–ø–ª—è—Ä–∞** ‚Äî —ç—Ç–æ —Ñ—É–Ω–∫—Ü–∏–∏, –∫–æ—Ç–æ—Ä—ã–µ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—Ç –æ–±—ä–µ–∫—Ç—É. –û–Ω–∏ –æ–ø—Ä–µ–¥–µ–ª—è—é—Ç –ø–æ–≤–µ–¥–µ–Ω–∏–µ –æ–±—ä–µ–∫—Ç–∞.

### –û—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏ –º–µ—Ç–æ–¥–æ–≤:
1. **–ü–µ—Ä–≤—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä** ‚Äî –≤—Å–µ–≥–¥–∞ `self` (—Å—Å—ã–ª–∫–∞ –Ω–∞ –æ–±—ä–µ–∫—Ç)
2. **–í—ã–∑—ã–≤–∞—é—Ç—Å—è –æ—Ç –æ–±—ä–µ–∫—Ç–∞** ‚Äî `–æ–±—ä–µ–∫—Ç.–º–µ—Ç–æ–¥()`
3. **–†–∞–±–æ—Ç–∞—é—Ç —Å –¥–∞–Ω–Ω—ã–º–∏ –æ–±—ä–µ–∫—Ç–∞** ‚Äî –∏–º–µ—é—Ç –¥–æ—Å—Ç—É–ø –∫ `self.–∞—Ç—Ä–∏–±—É—Ç—ã`
4. **–ú–æ–≥—É—Ç –∏–∑–º–µ–Ω—è—Ç—å —Å–æ—Å—Ç–æ—è–Ω–∏–µ –æ–±—ä–µ–∫—Ç–∞** ‚Äî –º–µ–Ω—è—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è –∞—Ç—Ä–∏–±—É—Ç–æ–≤

### –ó–∞—á–µ–º –Ω—É–∂–Ω—ã –º–µ—Ç–æ–¥—ã?
1. **–ò–Ω–∫–∞–ø—Å—É–ª—è—Ü–∏—è** ‚Äî –¥–∞–Ω–Ω—ã–µ –∏ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è —Ä–∞–±–æ—Ç—ã —Å –Ω–∏–º–∏ –≤ –æ–¥–Ω–æ–º –º–µ—Å—Ç–µ
2. **–ü–æ–≤—Ç–æ—Ä–Ω–æ–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ** ‚Äî –æ–¥–∏–Ω –º–µ—Ç–æ–¥ —Ä–∞–±–æ—Ç–∞–µ—Ç –¥–ª—è –≤—Å–µ—Ö –æ–±—ä–µ–∫—Ç–æ–≤
3. **–ê–±—Å—Ç—Ä–∞–∫—Ü–∏—è** ‚Äî —Å–∫—Ä—ã–≤–∞—é—Ç —Å–ª–æ–∂–Ω–æ—Å—Ç—å –æ–ø–µ—Ä–∞—Ü–∏–π

````python
# –°–æ–∑–¥–∞—ë–º –∫–ª–∞—Å—Å —Å –º–µ—Ç–æ–¥–∞–º–∏
print("üõ†Ô∏è –°–æ–∑–¥–∞—ë–º –∫–ª–∞—Å—Å —Å –º–µ—Ç–æ–¥–∞–º–∏:")

class BankAccount:
    """–ö–ª–∞—Å—Å –¥–ª—è –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è –±–∞–Ω–∫–æ–≤—Å–∫–æ–≥–æ —Å—á—ë—Ç–∞."""
    
    def __init__(self, owner, balance=0):
        """–ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ—Ç —Å—á—ë—Ç."""
        self.owner = owner
        self.balance = balance
        self.transactions = []  # –ò—Å—Ç–æ—Ä–∏—è –æ–ø–µ—Ä–∞—Ü–∏–π
    
    def deposit(self, amount):
        """–ü–æ–ø–æ–ª–Ω–∏—Ç—å —Å—á—ë—Ç."""
        if amount > 0:
            self.balance += amount
            self.transactions.append(f"–ü–æ–ø–æ–ª–Ω–µ–Ω–∏–µ: +{amount}")
            print(f"–°—á—ë—Ç –ø–æ–ø–æ–ª–Ω–µ–Ω –Ω–∞ {amount}. –ù–æ–≤—ã–π –±–∞–ª–∞–Ω—Å: {self.balance}")
        else:
            print("–û—à–∏–±–∫–∞: —Å—É–º–º–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–π")
    
    def withdraw(self, amount):
        """–°–Ω—è—Ç—å —Å–æ —Å—á—ë—Ç–∞."""
        if amount > 0:
            if amount <= self.balance:
                self.balance -= amount
                self.transactions.append(f"–°–Ω—è—Ç–∏–µ: -{amount}")
                print(f"–°–Ω—è—Ç–æ {amount}. –ù–æ–≤—ã–π –±–∞–ª–∞–Ω—Å: {self.balance}")
            else:
                print("–û—à–∏–±–∫–∞: –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Å—Ä–µ–¥—Å—Ç–≤")
        else:
            print("–û—à–∏–±–∫–∞: —Å—É–º–º–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–π")
    
    def get_balance(self):
        """–ü–æ–ª—É—á–∏—Ç—å —Ç–µ–∫—É—â–∏–π –±–∞–ª–∞–Ω—Å."""
        return self.balance
    
    def print_statement(self):
        """–í—ã–≤–µ—Å—Ç–∏ –≤—ã–ø–∏—Å–∫—É –ø–æ —Å—á—ë—Ç—É."""
        print(f"\n=== –í—ã–ø–∏—Å–∫–∞ –ø–æ —Å—á—ë—Ç—É ===")
        print(f"–í–ª–∞–¥–µ–ª–µ—Ü: {self.owner}")
        print(f"–¢–µ–∫—É—â–∏–π –±–∞–ª–∞–Ω—Å: {self.balance}")
        print("–ò—Å—Ç–æ—Ä–∏—è –æ–ø–µ—Ä–∞—Ü–∏–π:")
        for i, transaction in enumerate(self.transactions, 1):
            print(f"  {i}. {transaction}")

# –°–æ–∑–¥–∞—ë–º —Å—á—ë—Ç –∏ —Ä–∞–±–æ—Ç–∞–µ–º —Å –Ω–∏–º
print("–°–æ–∑–¥–∞—ë–º –±–∞–Ω–∫–æ–≤—Å–∫–∏–π —Å—á—ë—Ç:")
account = BankAccount("–ò–≤–∞–Ω –ü–µ—Ç—Ä–æ–≤", 1000)

print(f"\n–ù–∞—á–∞–ª—å–Ω—ã–π –±–∞–ª–∞–Ω—Å: {account.get_balance()}")

# –ò—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç–æ–¥—ã
account.deposit(500)
account.withdraw(200)
account.withdraw(1500)  # –ü–æ–ø—ã—Ç–∫–∞ —Å–Ω—è—Ç—å –±–æ–ª—å—à–µ, —á–µ–º –µ—Å—Ç—å

account.print_statement()

# –ö–∞–∫ Python –≤—ã–∑—ã–≤–∞–µ—Ç –º–µ—Ç–æ–¥—ã?
print("\nüîç –ö–∞–∫ Python –≤—ã–∑—ã–≤–∞–µ—Ç –º–µ—Ç–æ–¥—ã?")
print("–ö–æ–≥–¥–∞ –º—ã –ø–∏—à–µ–º account.deposit(500), Python –¥–µ–ª–∞–µ—Ç:")
print("  1. –ò—â–µ—Ç –º–µ—Ç–æ–¥ deposit –≤ –∫–ª–∞—Å—Å–µ BankAccount")
print("  2. –í—ã–∑—ã–≤–∞–µ—Ç BankAccount.deposit(account, 500)")
print("  3. account –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –ø–µ—Ä–µ–¥–∞—ë—Ç—Å—è –∫–∞–∫ –ø–µ—Ä–≤—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä self")

# –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è: –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–π —Å–ø–æ—Å–æ–± –≤—ã–∑–æ–≤–∞ –º–µ—Ç–æ–¥–∞
print("\n–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã–π —Å–ø–æ—Å–æ–± –≤—ã–∑–æ–≤–∞ –º–µ—Ç–æ–¥–∞:")
print(f"  account.deposit(300) - –æ–±—ã—á–Ω—ã–π —Å–ø–æ—Å–æ–±")
print(f"  BankAccount.deposit(account, 300) - –ø—Ä—è–º–æ–π –≤—ã–∑–æ–≤ —á–µ—Ä–µ–∑ –∫–ª–∞—Å—Å")

# –ü—Ä–æ–≤–µ—Ä–∏–º
print(f"\n–ë–∞–ª–∞–Ω—Å –¥–æ: {account.get_balance()}")
BankAccount.deposit(account, 300)  # –Ø–≤–Ω–∞—è –ø–µ—Ä–µ–¥–∞—á–∞ self
print(f"–ë–∞–ª–∞–Ω—Å –ø–æ—Å–ª–µ: {account.get_balance()}")

print("\n‚úÖ –í—ã–≤–æ–¥: account.method(...) —ç—Ç–æ –ø—Ä–æ—Å—Ç–æ —É–¥–æ–±–Ω—ã–π —Å–∏–Ω—Ç–∞–∫—Å–∏—Å!")
print("–ù–∞ —Å–∞–º–æ–º –¥–µ–ª–µ –≤—ã–∑—ã–≤–∞–µ—Ç—Å—è Class.method(object, ...)")
````

## üë• –ê—Ç—Ä–∏–±—É—Ç—ã –∫–ª–∞—Å—Å–∞

### –•–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ –∞—Ç—Ä–∏–±—É—Ç–æ–≤ –∫–ª–∞—Å—Å–∞:
1. **–û–±—â–∏–µ –¥–ª—è –≤—Å–µ—Ö –æ–±—ä–µ–∫—Ç–æ–≤** ‚Äî –æ–¥–∏–Ω–∞–∫–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —É –≤—Å–µ—Ö —ç–∫–∑–µ–º–ø–ª—è—Ä–æ–≤
2. **–ò–∑–º–µ–Ω—è—é—Ç—Å—è —á–µ—Ä–µ–∑ –∫–ª–∞—Å—Å** ‚Äî `–ö–ª–∞—Å—Å.–∞—Ç—Ä–∏–±—É—Ç = –∑–Ω–∞—á–µ–Ω–∏–µ`
3. **–ú–æ–≥—É—Ç –±—ã—Ç—å –ø–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª–µ–Ω—ã** ‚Äî –æ–±—ä–µ–∫—Ç –º–æ–∂–µ—Ç –∏–º–µ—Ç—å —Å–≤–æ—ë –∑–Ω–∞—á–µ–Ω–∏–µ
4. **–ò—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è –¥–ª—è –∫–æ–Ω—Å—Ç–∞–Ω—Ç –∏ –æ–±—â–∏—Ö –¥–∞–Ω–Ω—ã—Ö**

### –ö–æ–≥–¥–∞ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∞—Ç—Ä–∏–±—É—Ç—ã –∫–ª–∞—Å—Å–∞?
1. –ö–æ–Ω—Å—Ç–∞–Ω—Ç—ã (–Ω–∞–ø—Ä–∏–º–µ—Ä, –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–µ –∫–æ–Ω—Å—Ç–∞–Ω—Ç—ã)
2. –°—á—ë—Ç—á–∏–∫–∏ –æ–±—ä–µ–∫—Ç–æ–≤
3. –û–±—â–∏–µ –Ω–∞—Å—Ç—Ä–æ–π–∫–∏ –¥–ª—è –≤—Å–µ—Ö –æ–±—ä–µ–∫—Ç–æ–≤
4. –ó–Ω–∞—á–µ–Ω–∏—è –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é

### –°–∏–Ω—Ç–∞–∫—Å–∏—Å:
```python
class MyClass:
    –∞—Ç—Ä–∏–±—É—Ç_–∫–ª–∞—Å—Å–∞ = –∑–Ω–∞—á–µ–Ω–∏–µ  # –û–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –≤–Ω–µ –º–µ—Ç–æ–¥–æ–≤
    
    def __init__(self):
        self.–∞—Ç—Ä–∏–±—É—Ç_—ç–∫–∑–µ–º–ø–ª—è—Ä–∞ = –∑–Ω–∞—á–µ–Ω–∏–µ

````python

# –ê—Ç—Ä–∏–±—É—Ç—ã –∫–ª–∞—Å—Å–∞
print("üë• –ê—Ç—Ä–∏–±—É—Ç—ã –∫–ª–∞—Å—Å–∞:")

class Dog:
    """–ö–ª–∞—Å—Å –¥–ª—è –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è —Å–æ–±–∞–∫–∏."""
    
    # –ê—Ç—Ä–∏–±—É—Ç—ã –∫–ª–∞—Å—Å–∞ (–æ–±—â–∏–µ –¥–ª—è –≤—Å–µ—Ö —Å–æ–±–∞–∫)
    species = "Canis familiaris"  # –í–∏–¥
    total_dogs = 0  # –°—á—ë—Ç—á–∏–∫ –≤—Å–µ—Ö —Å–æ–∑–¥–∞–Ω–Ω—ã—Ö —Å–æ–±–∞–∫
    
    def __init__(self, name, breed, age):
        """–ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ—Ç —Å–æ–±–∞–∫—É."""
        # –ê—Ç—Ä–∏–±—É—Ç—ã —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ (—É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –¥–ª—è –∫–∞–∂–¥–æ–π —Å–æ–±–∞–∫–∏)
        self.name = name
        self.breed = breed
        self.age = age
        
        # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á—ë—Ç—á–∏–∫ —Å–æ–±–∞–∫
        Dog.total_dogs += 1  # –û–±—Ä–∞—â–∞–µ–º—Å—è —á–µ—Ä–µ–∑ –∏–º—è –∫–ª–∞—Å—Å–∞!
    
    def bark(self):
        """–°–æ–±–∞–∫–∞ –ª–∞–µ—Ç."""
        return f"{self.name} –≥–æ–≤–æ—Ä–∏—Ç: –ì–∞–≤!"
    
    @classmethod
    def get_total_dogs(cls):
        """–ü–æ–ª—É—á–∏—Ç—å –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–∑–¥–∞–Ω–Ω—ã—Ö —Å–æ–±–∞–∫."""
        return cls.total_dogs

# –°–æ–∑–¥–∞—ë–º —Å–æ–±–∞–∫
print("–°–æ–∑–¥–∞—ë–º —Å–æ–±–∞–∫:")
dog1 = Dog("–ë–æ–±–∏–∫", "–î–≤–æ—Ä–Ω—è–∂–∫–∞", 3)
dog2 = Dog("–®–∞—Ä–∏–∫", "–û–≤—á–∞—Ä–∫–∞", 5)
dog3 = Dog("–ú—É—Ä–∫–∞", "–¢–∞–∫—Å–∞", 2)  # –î–∞, –∫–æ—Ç—ã —Ç–æ–∂–µ –º–æ–≥—É—Ç –±—ã—Ç—å —Å–æ–±–∞–∫–∞–º–∏ –≤ –Ω–∞—à–µ–π –ø—Ä–æ–≥—Ä–∞–º–º–µ :)

print(f"\n–ê—Ç—Ä–∏–±—É—Ç—ã –∫–ª–∞—Å—Å–∞ (–æ–±—â–∏–µ):")
print(f"  Dog.species = {Dog.species}")
print(f"  –í—Å–µ–≥–æ —Å–æ–∑–¥–∞–Ω–æ —Å–æ–±–∞–∫: {Dog.get_total_dogs()}")

print(f"\n–ê—Ç—Ä–∏–±—É—Ç—ã —ç–∫–∑–µ–º–ø–ª—è—Ä–∞ (—É–Ω–∏–∫–∞–ª—å–Ω—ã–µ):")
print(f"  dog1.name = {dog1.name}, dog1.breed = {dog1.breed}")
print(f"  dog2.name = {dog2.name}, dog2.breed = {dog2.breed}")

# –î–æ—Å—Ç—É–ø –∫ –∞—Ç—Ä–∏–±—É—Ç–∞–º –∫–ª–∞—Å—Å–∞ —á–µ—Ä–µ–∑ —ç–∫–∑–µ–º–ø–ª—è—Ä
print(f"\n–î–æ—Å—Ç—É–ø –∫ –∞—Ç—Ä–∏–±—É—Ç–∞–º –∫–ª–∞—Å—Å–∞ —á–µ—Ä–µ–∑ —ç–∫–∑–µ–º–ø–ª—è—Ä—ã:")
print(f"  dog1.species = {dog1.species}")  # –ú–æ–∂–Ω–æ —á–µ—Ä–µ–∑ —ç–∫–∑–µ–º–ø–ª—è—Ä
print(f"  dog2.species = {dog2.species}")  # –í—Å–µ–≥–¥–∞ –æ–¥–Ω–æ –∏ —Ç–æ –∂–µ –∑–Ω–∞—á–µ–Ω–∏–µ

# –ò–∑–º–µ–Ω–µ–Ω–∏–µ –∞—Ç—Ä–∏–±—É—Ç–∞ –∫–ª–∞—Å—Å–∞
print(f"\n–ò–∑–º–µ–Ω—è–µ–º –∞—Ç—Ä–∏–±—É—Ç –∫–ª–∞—Å—Å–∞:")
Dog.species = "Canis lupus familiaris"  # –ú–µ–Ω—è–µ–º —á–µ—Ä–µ–∑ –∫–ª–∞—Å—Å
print(f"  –ü–æ—Å–ª–µ Dog.species = 'Canis lupus familiaris':")
print(f"  Dog.species = {Dog.species}")
print(f"  dog1.species = {dog1.species}")  # –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –æ–±–Ω–æ–≤–∏–ª–æ—Å—å!
print(f"  dog2.species = {dog2.species}")

# –ß—Ç–æ –µ—Å–ª–∏ –∏–∑–º–µ–Ω–∏—Ç—å —á–µ—Ä–µ–∑ —ç–∫–∑–µ–º–ø–ª—è—Ä?
print(f"\n–ò–∑–º–µ–Ω—è–µ–º species —á–µ—Ä–µ–∑ —ç–∫–∑–µ–º–ø–ª—è—Ä:")
dog1.species = "–û—Å–æ–±–∞—è –ø–æ—Ä–æ–¥–∞"  # –°–æ–∑–¥–∞—ë—Ç—Å—è –∞—Ç—Ä–∏–±—É—Ç —ç–∫–∑–µ–º–ø–ª—è—Ä–∞!
print(f"  –ü–æ—Å–ª–µ dog1.species = '–û—Å–æ–±–∞—è –ø–æ—Ä–æ–¥–∞':")
print(f"  Dog.species = {Dog.species}")  # –ê—Ç—Ä–∏–±—É—Ç –∫–ª–∞—Å—Å–∞ –Ω–µ –∏–∑–º–µ–Ω–∏–ª—Å—è
print(f"  dog1.species = {dog1.species}")  # –£ dog1 —Ç–µ–ø–µ—Ä—å —Å–≤–æ–π species
print(f"  dog2.species = {dog2.species}")  # dog2 –ø–æ-–ø—Ä–µ–∂–Ω–µ–º—É –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –∞—Ç—Ä–∏–±—É—Ç –∫–ª–∞—Å—Å–∞

# –ö–∞–∫ Python –∏—â–µ—Ç –∞—Ç—Ä–∏–±—É—Ç—ã?
print("\nüîç –ö–∞–∫ Python –∏—â–µ—Ç –∞—Ç—Ä–∏–±—É—Ç—ã?")
print("–ö–æ–≥–¥–∞ –º—ã –ø–∏—à–µ–º dog1.species, Python:")
print("  1. –ò—â–µ—Ç –∞—Ç—Ä–∏–±—É—Ç species –≤ —Å–∞–º–æ–º –æ–±—ä–µ–∫—Ç–µ dog1")
print("  2. –ï—Å–ª–∏ –Ω–µ –Ω–∞—Ö–æ–¥–∏—Ç, –∏—â–µ—Ç –≤ –∫–ª–∞—Å—Å–µ Dog")
print("  3. –ï—Å–ª–∏ –Ω–µ –Ω–∞—Ö–æ–¥–∏—Ç —Ç–∞–º, –∏—â–µ—Ç –≤ —Ä–æ–¥–∏—Ç–µ–ª—å—Å–∫–∏—Ö –∫–ª–∞—Å—Å–∞—Ö")

print("\n–ü–æ—Å–ª–µ dog1.species = '–û—Å–æ–±–∞—è –ø–æ—Ä–æ–¥–∞':")
print(f"  'species' in dog1.__dict__ = {'species' in dog1.__dict__}")
print(f"  'species' in dog2.__dict__ = {'species' in dog2.__dict__}")

# –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ: —Å–∫–∏–¥–∫–∞ –¥–ª—è –≤—Å–µ—Ö —Ç–æ–≤–∞—Ä–æ–≤
print("\nüìä –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ –∞—Ç—Ä–∏–±—É—Ç–æ–≤ –∫–ª–∞—Å—Å–∞:")

class Product:
    """–ö–ª–∞—Å—Å –¥–ª—è –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è —Ç–æ–≤–∞—Ä–∞."""
    
    discount = 0.1  # 10% —Å–∫–∏–¥–∫–∞ –¥–ª—è –≤—Å–µ—Ö —Ç–æ–≤–∞—Ä–æ–≤
    
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    def get_price_with_discount(self):
        """–ü–æ–ª—É—á–∏—Ç—å —Ü–µ–Ω—É —Å–æ —Å–∫–∏–¥–∫–æ–π."""
        return self.price * (1 - Product.discount)
    
    @classmethod
    def set_discount(cls, new_discount):
        """–£—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –Ω–æ–≤—É—é —Å–∫–∏–¥–∫—É –¥–ª—è –≤—Å–µ—Ö —Ç–æ–≤–∞—Ä–æ–≤."""
        cls.discount = new_discount
        print(f"–£—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∞ –Ω–æ–≤–∞—è —Å–∫–∏–¥–∫–∞: {new_discount * 100}%")

# –°–æ–∑–¥–∞—ë–º —Ç–æ–≤–∞—Ä—ã
product1 = Product("–ù–æ—É—Ç–±—É–∫", 50000)
product2 = Product("–¢–µ–ª–µ—Ñ–æ–Ω", 30000)

print(f"–¢–æ–≤–∞—Ä 1: {product1.name}, —Ü–µ–Ω–∞: {product1.price}, —Ü–µ–Ω–∞ —Å–æ —Å–∫–∏–¥–∫–æ–π: {product1.get_price_with_discount():.0f}")
print(f"–¢–æ–≤–∞—Ä 2: {product2.name}, —Ü–µ–Ω–∞: {product2.price}, —Ü–µ–Ω–∞ —Å–æ —Å–∫–∏–¥–∫–æ–π: {product2.get_price_with_discount():.0f}")

# –ú–µ–Ω—è–µ–º —Å–∫–∏–¥–∫—É –¥–ª—è –≤—Å–µ—Ö —Ç–æ–≤–∞—Ä–æ–≤
Product.set_discount(0.2)  # 20% —Å–∫–∏–¥–∫–∞

print(f"\n–ü–æ—Å–ª–µ –∏–∑–º–µ–Ω–µ–Ω–∏—è —Å–∫–∏–¥–∫–∏:")
print(f"  {product1.name}: —Ü–µ–Ω–∞ —Å–æ —Å–∫–∏–¥–∫–æ–π: {product1.get_price_with_discount():.0f}")
print(f"  {product2.name}: —Ü–µ–Ω–∞ —Å–æ —Å–∫–∏–¥–∫–æ–π: {product2.get_price_with_discount():.0f}")
````

## üîç –ß—Ç–æ —Ç–∞–∫–æ–µ `__dict__`?

**`__dict__`** ‚Äî —ç—Ç–æ —Å–ø–µ—Ü–∏–∞–ª—å–Ω—ã–π –∞—Ç—Ä–∏–±—É—Ç, –∫–æ—Ç–æ—Ä—ã–π —Å–æ–¥–µ—Ä–∂–∏—Ç —Å–ª–æ–≤–∞—Ä—å —Å –∞—Ç—Ä–∏–±—É—Ç–∞–º–∏ –æ–±—ä–µ–∫—Ç–∞ –∏–ª–∏ –∫–ª–∞—Å—Å–∞.

### –ß—Ç–æ —Ö—Ä–∞–Ω–∏—Ç—Å—è –≤ `__dict__`?
- –î–ª—è **–æ–±—ä–µ–∫—Ç–æ–≤**: –∞—Ç—Ä–∏–±—É—Ç—ã —ç–∫–∑–µ–º–ø–ª—è—Ä–∞
- –î–ª—è **–∫–ª–∞—Å—Å–æ–≤**: –∞—Ç—Ä–∏–±—É—Ç—ã –∫–ª–∞—Å—Å–∞, –º–µ—Ç–æ–¥—ã, –∏ –¥—Ä—É–≥—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é

### –ó–∞—á–µ–º –Ω—É–∂–Ω–æ –∑–Ω–∞—Ç—å –ø—Ä–æ `__dict__`?
1. **–û—Ç–ª–∞–¥–∫–∞** ‚Äî –º–æ–∂–Ω–æ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å, –∫–∞–∫–∏–µ –∞—Ç—Ä–∏–±—É—Ç—ã –µ—Å—Ç—å —É –æ–±—ä–µ–∫—Ç–∞
2. **–î–∏–Ω–∞–º–∏—á–µ—Å–∫–æ–µ —Å–æ–∑–¥–∞–Ω–∏–µ –∞—Ç—Ä–∏–±—É—Ç–æ–≤** ‚Äî –º–æ–∂–Ω–æ –¥–æ–±–∞–≤–ª—è—Ç—å –∞—Ç—Ä–∏–±—É—Ç—ã –ø—Ä—è–º–æ –≤ —Å–ª–æ–≤–∞—Ä—å
3. **–ü–æ–Ω–∏–º–∞–Ω–∏–µ —É—Å—Ç—Ä–æ–π—Å—Ç–≤–∞ Python** ‚Äî –≤–∏–¥–∏–º, –∫–∞–∫ Python —Ö—Ä–∞–Ω–∏—Ç –∞—Ç—Ä–∏–±—É—Ç—ã

### –í–∞–∂–Ω–æ:
- –ù–µ –≤—Å–µ –æ–±—ä–µ–∫—Ç—ã –∏–º–µ—é—Ç `__dict__` (–Ω–∞–ø—Ä–∏–º–µ—Ä, –≤—Å—Ç—Ä–æ–µ–Ω–Ω—ã–µ —Ç–∏–ø—ã)
- –ò—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å `__dict__` –Ω–∞–ø—Ä—è–º—É—é –Ω—É–∂–Ω–æ –æ—Å—Ç–æ—Ä–æ–∂–Ω–æ
- –î–ª—è –æ–±—ã—á–Ω–æ–π —Ä–∞–±–æ—Ç—ã –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –æ–±—ã—á–Ω—ã–π —Å–∏–Ω—Ç–∞–∫—Å–∏—Å —Å —Ç–æ—á–∫–æ–π

````python
# –ò–∑—É—á–∞–µ–º __dict__
print("üîç –ò–∑—É—á–∞–µ–º __dict__:")

class SimpleClass:
    """–ü—Ä–æ—Å—Ç–æ–π –∫–ª–∞—Å—Å –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏ __dict__."""
    
    class_attr = "–Ø –∞—Ç—Ä–∏–±—É—Ç –∫–ª–∞—Å—Å–∞"
    
    def __init__(self, value):
        self.instance_attr = value
    
    def method(self):
        return "–Ø –º–µ—Ç–æ–¥"

# –°–æ–∑–¥–∞—ë–º –æ–±—ä–µ–∫—Ç
obj = SimpleClass("–Ø –∞—Ç—Ä–∏–±—É—Ç —ç–∫–∑–µ–º–ø–ª—è—Ä–∞")

print("1. __dict__ –æ–±—ä–µ–∫—Ç–∞:")
print(f"   obj.__dict__ = {obj.__dict__}")
print(f"   –í –Ω—ë–º —Ç–æ–ª—å–∫–æ –∞—Ç—Ä–∏–±—É—Ç—ã —ç–∫–∑–µ–º–ø–ª—è—Ä–∞: {list(obj.__dict__.keys())}")

print("\n2. __dict__ –∫–ª–∞—Å—Å–∞:")
print(f"   SimpleClass.__dict__ keys (–ø–µ—Ä–≤—ã–µ 10): {list(SimpleClass.__dict__.keys())[:10]}")
print(f"   SimpleClass.__dict__['class_attr'] = {SimpleClass.__dict__['class_attr']}")
print(f"   SimpleClass.__dict__['method'] = {SimpleClass.__dict__['method']}")

# –ö–∞–∫ Python –∏—â–µ—Ç –∞—Ç—Ä–∏–±—É—Ç—ã?
print("\nüîç –ö–∞–∫ Python –∏—â–µ—Ç –∞—Ç—Ä–∏–±—É—Ç—ã (–Ω–∞ –ø—Ä–∏–º–µ—Ä–µ obj.class_attr):")
print("  1. –°–º–æ—Ç—Ä–∏—Ç –≤ obj.__dict__ - –Ω–µ –Ω–∞—Ö–æ–¥–∏—Ç 'class_attr'")
print("  2. –°–º–æ—Ç—Ä–∏—Ç –≤ SimpleClass.__dict__ - –Ω–∞—Ö–æ–¥–∏—Ç!")
print(f"  –†–µ–∑—É–ª—å—Ç–∞—Ç: obj.class_attr = {obj.class_attr}")

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –∞—Ç—Ä–∏–±—É—Ç–∞ —á–µ—Ä–µ–∑ __dict__
print("\n3. –î–æ–±–∞–≤–ª—è–µ–º –∞—Ç—Ä–∏–±—É—Ç —á–µ—Ä–µ–∑ __dict__:")
obj.__dict__['new_attr'] = '–î–æ–±–∞–≤–ª–µ–Ω —á–µ—Ä–µ–∑ __dict__'
print(f"   obj.new_attr = {obj.new_attr}")
print(f"   obj.__dict__ = {obj.__dict__}")

# –ò–∑–º–µ–Ω–µ–Ω–∏–µ –∞—Ç—Ä–∏–±—É—Ç–∞ —á–µ—Ä–µ–∑ __dict__
print("\n4. –ò–∑–º–µ–Ω—è–µ–º –∞—Ç—Ä–∏–±—É—Ç —á–µ—Ä–µ–∑ __dict__:")
obj.__dict__['instance_attr'] = '–ò–∑–º–µ–Ω–µ–Ω–æ —á–µ—Ä–µ–∑ __dict__'
print(f"   obj.instance_attr = {obj.instance_attr}")

# –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏–π –ø—Ä–∏–º–µ—Ä: –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≤ —Å–ª–æ–≤–∞—Ä—å
print("\nüìä –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏–π –ø—Ä–∏–º–µ—Ä: –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –æ–±—ä–µ–∫—Ç–∞ –≤ —Å–ª–æ–≤–∞—Ä—å:")

class Person:
    def __init__(self, name, age, city):
        self.name = name
        self.age = age
        self.city = city
    
    def to_dict(self):
        """–ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å –æ–±—ä–µ–∫—Ç –≤ —Å–ª–æ–≤–∞—Ä—å."""
        return self.__dict__.copy()  # –ö–æ–ø–∏—Ä—É–µ–º, —á—Ç–æ–±—ã –Ω–µ –º–µ–Ω—è—Ç—å –æ—Ä–∏–≥–∏–Ω–∞–ª
    
    @classmethod
    def from_dict(cls, data):
        """–°–æ–∑–¥–∞—Ç—å –æ–±—ä–µ–∫—Ç –∏–∑ —Å–ª–æ–≤–∞—Ä—è."""
        return cls(**data)  # –†–∞—Å–ø–∞–∫–æ–≤—ã–≤–∞–µ–º —Å–ª–æ–≤–∞—Ä—å –∫–∞–∫ –∞—Ä–≥—É–º–µ–Ω—Ç—ã

person = Person("–ê–Ω–Ω–∞", 25, "–ú–æ—Å–∫–≤–∞")
print(f"–û–±—ä–µ–∫—Ç: {person.name}, {person.age} –ª–µ—Ç, {person.city}")

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ —Å–ª–æ–≤–∞—Ä—å
person_dict = person.to_dict()
print(f"–°–ª–æ–≤–∞—Ä—å: {person_dict}")

# –°–æ–∑–¥–∞—ë–º –∏–∑ —Å–ª–æ–≤–∞—Ä—è
new_person = Person.from_dict({"name": "–ò–≤–∞–Ω", "age": 30, "city": "–°–ü–±"})
print(f"–ù–æ–≤—ã–π –æ–±—ä–µ–∫—Ç –∏–∑ —Å–ª–æ–≤–∞—Ä—è: {new_person.name}")

# –ü—Ä–µ–¥–æ—Å—Ç–µ—Ä–µ–∂–µ–Ω–∏–µ: –Ω–µ –≤—Å–µ –æ–±—ä–µ–∫—Ç—ã –∏–º–µ—é—Ç __dict__
print("\n‚ö†Ô∏è –ü—Ä–µ–¥–æ—Å—Ç–µ—Ä–µ–∂–µ–Ω–∏–µ: –Ω–µ –≤—Å–µ –æ–±—ä–µ–∫—Ç—ã –∏–º–µ—é—Ç __dict__:")

# –í—Å—Ç—Ä–æ–µ–Ω–Ω—ã–µ —Ç–∏–ø—ã
x = 10
s = "hello"
lst = [1, 2, 3]

print(f"  –£ int –µ—Å—Ç—å __dict__? {hasattr(x, '__dict__')}")
print(f"  –£ str –µ—Å—Ç—å __dict__? {hasattr(s, '__dict__')}")
print(f"  –£ list –µ—Å—Ç—å __dict__? {hasattr(lst, '__dict__')}")

# –°–ª–æ—Ç—ã (–º—ã –∏—Ö –µ—â—ë –Ω–µ –ø—Ä–æ—Ö–æ–¥–∏–ª–∏, –Ω–æ –¥–ª—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏)
print("\nüí° –î–ª—è –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏ –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å __slots__:")
print("  class Optimized:")
print("      __slots__ = ['x', 'y']  # –§–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –Ω–∞–±–æ—Ä –∞—Ç—Ä–∏–±—É—Ç–æ–≤")
print("      def __init__(self, x, y):")
print("          self.x = x")
print("          self.y = y")
print("  –¢–∞–∫–∏–µ –æ–±—ä–µ–∫—Ç—ã –Ω–µ –∏–º–µ—é—Ç __dict__ –∏ –∏—Å–ø–æ–ª—å–∑—É—é—Ç –º–µ–Ω—å—à–µ –ø–∞–º—è—Ç–∏")
````
