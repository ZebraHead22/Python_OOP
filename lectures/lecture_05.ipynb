## 🏗️ Что такое наследование?

**Наследование** — это механизм, позволяющий создавать новый класс на основе существующего.

### Аналогия из реального мира:
- **Родительский класс:** Транспортное средство
- **Дочерние классы:** Автомобиль, Велосипед, Самолёт

Все они являются транспортными средствами, но имеют свои особенности.

### Зачем нужно наследование?
1. **Повторное использование кода** — не нужно писать одно и то же много раз
2. **Создание иерархий** — логическая организация классов
3. **Расширение функциональности** — добавление нового к уже существующему
4. **Полиморфизм** — возможность работать с разными типами одинаково

## 📊 План лекции

1. **Базовое наследование** — создание простой иерархии
2. **Переопределение методов** — изменение поведения родителя
3. **Функция `super()`** — доступ к методам родительского класса
4. **Полиморфизм** — один интерфейс, много реализаций
5. **Утиная типизация** — "Если это выглядит как утка и крякает как утка..."

> 💡 **Важно:** Наследование — это отношение "является" (is-a). Например: "Автомобиль является транспортным средством".

## 👶 Базовое наследование

### Синтаксис наследования в Python:
```python
class РодительскийКласс:
    # методы и атрибуты родителя

class ДочернийКласс(РодительскийКласс):
    # методы и атрибуты дочернего класса
    # + всё, что есть у родителя

````python


## Ячейка 3: Пример простого наследования (Code)


# Простое наследование
print("👶 Простое наследование:\n")

# Родительский класс
class Animal:
    """Базовый класс для всех животных."""
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.is_alive = True
    
    def eat(self):
        """Животное ест."""
        print(f"{self.name} ест...")
    
    def sleep(self):
        """Животное спит."""
        print(f"{self.name} спит...")
    
    def make_sound(self):
        """Животное издаёт звук."""
        print(f"{self.name} издаёт звук...")
    
    def get_info(self):
        """Получить информацию о животном."""
        return f"{self.name}, {self.age} лет, {'жив' if self.is_alive else 'не жив'}"

# Дочерний класс
class Dog(Animal):  # Dog наследует от Animal
    """Класс собаки."""
    
    def __init__(self, name, age, breed):
        # Вызываем конструктор родительского класса
        super().__init__(name, age)  # Инициализируем то, что есть у Animal
        self.breed = breed           # Добавляем новое свойство
        self.tricks = []             # Ещё одно новое свойство
    
    # Переопределяем метод родителя
    def make_sound(self):
        """Собака лает."""
        print(f"{self.name} говорит: Гав! Гав!")
    
    # Добавляем новый метод
    def learn_trick(self, trick):
        """Собака учит новый трюк."""
        self.tricks.append(trick)
        print(f"{self.name} выучил трюк: {trick}")
    
    # Расширяем метод родителя
    def get_info(self):
        """Получить информацию о собаке."""
        base_info = super().get_info()  # Вызываем метод родителя
        return f"{base_info}, порода: {self.breed}, трюков: {len(self.tricks)}"

# Другой дочерний класс
class Cat(Animal):
    """Класс кота."""
    
    def __init__(self, name, age, color):
        super().__init__(name, age)
        self.color = color
        self.lives = 9  # У кота 9 жизней!
    
    def make_sound(self):
        """Кот мяукает."""
        print(f"{self.name} говорит: Мяу!")
    
    def lose_life(self):
        """Кот теряет жизнь."""
        if self.lives > 0:
            self.lives -= 1
            print(f"{self.name} теряет жизнь. Осталось: {self.lives}")
            if self.lives == 0:
                self.is_alive = False
        else:
            print(f"{self.name} уже умер :(")

# Создаём объекты
print("Создаём животных:")
animal = Animal("Неизвестное животное", 3)
dog = Dog("Бобик", 5, "Дворняжка")
cat = Cat("Мурзик", 2, "рыжий")

print(f"\n1. Базовое животное: {animal.get_info()}")
print(f"2. Собака: {dog.get_info()}")
print(f"3. Кот: {cat.get_info()}, цвет: {cat.color}, жизней: {cat.lives}")

# Используем методы
print("\n🐾 Демонстрация поведения:")
print("\nЖивотное:")
animal.eat()
animal.make_sound()

print("\nСобака:")
dog.eat()            # Унаследованный метод
dog.make_sound()     # Переопределённый метод
dog.learn_trick("Сидеть")  # Свой метод
dog.learn_trick("Лежать")
print(f"Трюки собаки: {dog.tricks}")

print("\nКот:")
cat.eat()
cat.make_sound()
cat.lose_life()      # Свой метод
print(f"Жизней осталось: {cat.lives}")

# Проверяем наследование
print("\n🔍 Проверка наследования:")
print(f"dog является Animal? {isinstance(dog, Animal)}")
print(f"cat является Animal? {isinstance(cat, Animal)}")
print(f"animal является Dog? {isinstance(animal, Dog)}")
print(f"animal является Cat? {isinstance(animal, Cat)}")

print(f"\nDog является подклассом Animal? {issubclass(Dog, Animal)}")
print(f"Cat является подклассом Animal? {issubclass(Cat, Animal)}")
print(f"Animal является подклассом Dog? {issubclass(Animal, Dog)}")
````

## 🔄 Переопределение методов

**Переопределение метода** — это создание в дочернем классе метода с тем же именем, что и в родительском классе.

### Зачем переопределять методы?
1. **Изменение поведения** — сделать метод более специфичным для дочернего класса
2. **Добавление функциональности** — расширить то, что делает метод
3. **Исправление поведения** — исправить то, что не подходит для дочернего класса

### Как происходит переопределение?
1. Python ищет метод сначала в классе объекта
2. Если не находит, ищет в родительском классе
3. Затем в родителе родителя и т.д.

### Особенности переопределения в Python:
- Можно полностью заменить метод
- Можно расширить метод с помощью `super()`
- Можно переопределять даже специальные методы (`__str__`, `__init__` и т.д.)

````python
# Переопределение методов
print("🔄 Переопределение методов:\n")

class Vehicle:
    """Базовый класс транспортного средства."""
    
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
        self.speed = 0
        self.is_moving = False
    
    def start(self):
        """Запустить транспортное средство."""
        if not self.is_moving:
            self.is_moving = True
            print(f"{self.brand} {self.model} завёлся")
        else:
            print(f"{self.brand} {self.model} уже заведён")
    
    def stop(self):
        """Остановить транспортное средство."""
        if self.is_moving:
            self.is_moving = False
            self.speed = 0
            print(f"{self.brand} {self.model} остановился")
        else:
            print(f"{self.brand} {self.model} уже остановлен")
    
    def accelerate(self, amount):
        """Ускориться."""
        if self.is_moving:
            self.speed += amount
            print(f"{self.brand} {self.model} ускорился до {self.speed} км/ч")
        else:
            print(f"Сначала нужно завести {self.brand} {self.model}")
    
    def get_info(self):
        """Получить информацию о транспортном средстве."""
        return f"{self.brand} {self.model} ({self.year}), скорость: {self.speed} км/ч"

class Car(Vehicle):
    """Класс автомобиля."""
    
    def __init__(self, brand, model, year, fuel_type):
        super().__init__(brand, model, year)
        self.fuel_type = fuel_type
        self.fuel_level = 100  # в процентах
        self.doors = 4
    
    # Переопределяем метод start
    def start(self):
        """Запустить автомобиль (проверяем топливо)."""
        if self.fuel_level > 0:
            super().start()  # Вызываем метод родителя
            print(f"  Уровень топлива: {self.fuel_level}%")
        else:
            print(f"{self.brand} {self.model}: нет топлива!")
    
    # Переопределяем метод accelerate
    def accelerate(self, amount):
        """Ускориться (расходует топливо)."""
        if self.is_moving:
            # Расходуем топливо при ускорении
            fuel_used = amount * 0.5
            if self.fuel_level >= fuel_used:
                self.fuel_level -= fuel_used
                super().accelerate(amount)  # Вызываем метод родителя
                print(f"  Израсходовано топлива: {fuel_used:.1f}%")
                print(f"  Остаток топлива: {self.fuel_level:.1f}%")
            else:
                print(f"{self.brand} {self.model}: недостаточно топлива!")
        else:
            super().accelerate(amount)  # Вызываем метод родителя
    
    # Добавляем новый метод
    def refuel(self, amount):
        """Заправить автомобиль."""
        self.fuel_level = min(100, self.fuel_level + amount)
        print(f"{self.brand} {self.model} заправлен. Топливо: {self.fuel_level}%")
    
    # Расширяем метод get_info
    def get_info(self):
        """Получить информацию об автомобиле."""
        base_info = super().get_info()
        return f"{base_info}, топливо: {self.fuel_type}, дверей: {self.doors}"

class Bicycle(Vehicle):
    """Класс велосипеда."""
    
    def __init__(self, brand, model, year, type):
        super().__init__(brand, model, year)
        self.type = type  # Горный, шоссейный и т.д.
        self.gears = 18
        self.current_gear = 1
    
    # Полностью переопределяем метод start
    def start(self):
        """Начать движение на велосипеде."""
        self.is_moving = True
        self.speed = 5  # Начинаем с небольшой скорости
        print(f"{self.brand} {self.model}: начинаем крутить педали!")
        print(f"  Начальная скорость: {self.speed} км/ч")
    
    # Переопределяем метод accelerate
    def accelerate(self, amount):
        """Ускориться на велосипеде (переключаем передачи)."""
        if self.is_moving:
            # На велосипеде ускоряемся переключением передач
            if amount > 0:
                if self.current_gear < self.gears:
                    self.current_gear += 1
                    self.speed += 3  # Каждая передача добавляет скорость
                    print(f"{self.brand} {self.model}: переключился на {self.current_gear} передачу")
                    print(f"  Скорость: {self.speed} км/ч")
                else:
                    print(f"{self.brand} {self.model}: уже на максимальной передаче!")
            else:
                print(f"{self.brand} {self.model}: для замедления используйте brake()")
        else:
            print(f"{self.brand} {self.model}: сначала нужно начать движение!")
    
    # Добавляем новый метод
    def brake(self):
        """Затормозить на велосипеде."""
        if self.is_moving and self.speed > 0:
            self.speed = max(0, self.speed - 5)
            print(f"{self.brand} {self.model} тормозит. Скорость: {self.speed} км/ч")
    
    # Расширяем метод get_info
    def get_info(self):
        """Получить информацию о велосипеде."""
        base_info = super().get_info()
        return f"{base_info}, тип: {self.type}, передача: {self.current_gear}/{self.gears}"

# Демонстрация
print("Создаём транспортные средства:")
car = Car("Toyota", "Camry", 2020, "бензин")
bike = Bicycle("Trek", "Marlin", 2021, "горный")

print(f"\n1. Автомобиль: {car.get_info()}")
print(f"2. Велосипед: {bike.get_info()}")

print("\n🚗 Работа с автомобилем:")
car.start()
car.accelerate(20)
car.accelerate(30)
car.accelerate(50)  # Израсходуем топливо
car.refuel(50)
car.accelerate(20)
car.stop()

print("\n🚲 Работа с велосипедом:")
bike.start()  # Переопределённый метод
bike.accelerate(10)  # Переопределённый метод
bike.accelerate(10)
bike.brake()  # Свой метод
bike.accelerate(-5)  # Попытка отрицательного ускорения
bike.stop()

print("\n🔍 Сравнение методов:")
print("У автомобиля и велосипеда есть методы start(), accelerate(), stop(), get_info()")
print("Но работают они по-разному - это и есть переопределение методов!")
````

## 🦸 Функция `super()`

**`super()`** — это функция, которая возвращает объект-прокси родительского класса.

### Зачем нужен `super()`?
1. **Вызов методов родителя** — когда нужно дополнить, а не заменить
2. **Доступ к атрибутам родителя** — когда нужно использовать родительские данные
3. **Решение проблем множественного наследования** — корректный порядок вызова методов

### Как работает `super()`?


````python

class Parent:
    def method(self):
        print("Родительский метод")

class Child(Parent):
    def method(self):
        super().method()  # Вызываем метод родителя
        print("Дочерний метод")
````

````python
# Прямое обращение (не рекомендуется)
class Child(Parent):
    def method(self):
        Parent.method(self)  # Жёстко привязано к Parent
        print("Дочерний метод")

# Использование super() (рекомендуется)
class Child(Parent):
    def method(self):
        super().method()  # Автоматически находит родителя
        print("Дочерний метод")
````

````python



## Ячейка 7: Примеры использования `super()` (Code)


# Использование super()
print("🦸 Функция super() в действии:\n")

# Пример 1: super() в конструкторе
print("1. super() в конструкторе:\n")

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"Person.__init__ вызван для {name}")

class Employee(Person):
    def __init__(self, name, age, employee_id):
        # Вызываем конструктор родителя
        super().__init__(name, age)
        self.employee_id = employee_id
        print(f"Employee.__init__ вызван для {name}, ID: {employee_id}")

class Manager(Employee):
    def __init__(self, name, age, employee_id, department):
        # Вызываем конструктор родителя
        super().__init__(name, age, employee_id)
        self.department = department
        print(f"Manager.__init__ вызван для {name}, отдел: {department}")

print("Создаём объекты:")
person = Person("Иван", 30)
print()
employee = Employee("Анна", 25, "E001")
print()
manager = Manager("Пётр", 40, "M001", "IT")
print()

# Пример 2: super() в методах
print("2. super() в обычных методах:\n")

class Shape:
    def __init__(self, color):
        self.color = color
    
    def area(self):
        """Вычислить площадь."""
        return 0
    
    def info(self):
        """Информация о фигуре."""
        return f"Фигура цвета {self.color}, площадь: {self.area()}"

class Rectangle(Shape):
    def __init__(self, color, width, height):
        super().__init__(color)
        self.width = width
        self.height = height
    
    def area(self):
        """Площадь прямоугольника."""
        return self.width * self.height
    
    def info(self):
        """Информация о прямоугольнике."""
        base_info = super().info()  # Вызываем info() родителя
        return f"{base_info}, ширина: {self.width}, высота: {self.height}"

class Square(Rectangle):
    def __init__(self, color, side):
        # Квадрат - это прямоугольник с равными сторонами
        super().__init__(color, side, side)
    
    def info(self):
        """Информация о квадрате."""
        base_info = super().info()  # Вызываем info() родителя (Rectangle)
        # Но Rectangle.info() уже вызывает Shape.info()
        return f"Квадрат. {base_info}"

print("Создаём фигуры:")
rect = Rectangle("красный", 5, 10)
square = Square("синий", 7)

print(f"Прямоугольник: {rect.info()}")
print(f"Квадрат: {square.info()}")
print()

# Пример 3: super() с множественным наследованием
print("3. super() с множественным наследованием:\n")

class A:
    def __init__(self):
        print("A.__init__")
        self.a = "a"

class B(A):
    def __init__(self):
        print("B.__init__")
        super().__init__()
        self.b = "b"

class C(A):
    def __init__(self):
        print("C.__init__")
        super().__init__()
        self.c = "c"

class D(B, C):
    def __init__(self):
        print("D.__init__")
        super().__init__()
        self.d = "d"

print("Создаём объект D (множественное наследование):")
obj_d = D()
print()

print("Порядок вызова конструкторов (MRO - Method Resolution Order):")
print(f"  D -> B -> C -> A")
print(f"  MRO класса D: {[cls.__name__ for cls in D.__mro__]}")
print()

print("Атрибуты объекта D:")
print(f"  {obj_d.__dict__}")
````

## 🎭 Полиморфизм

**Полиморфизм** (от греческого "много форм") — это возможность объектов разных классов отвечать на одни и те же сообщения (вызовы методов).

### Простой пример:
У нас есть классы `Собака`, `Кот`, `Птица`. У всех есть метод `make_sound()`, но реализация разная:
- Собака: "Гав!"
- Кот: "Мяу!"
- Птица: "Чирик!"

### Зачем нужен полиморфизм?
1. **Универсальный интерфейс** — можем работать с разными объектами одинаково
2. **Гибкость кода** — можно добавлять новые классы, не меняя существующий код
3. **Читаемость** — код становится более понятным и предсказуемым

### Виды полиморфизма в Python:
1. **Полиморфизм через наследование** — переопределение методов в дочерних классах
2. **Полиморфизм через интерфейсы** (утиная типизация) — если объект имеет нужный метод, он подходит
3. **Полиморфизм через перегрузку операторов** — переопределение `+`, `-`, `*` и т.д.

### Пример полиморфизма:
```python
def make_animal_sound(animal):
    animal.make_sound()  # Неважно, кто это - собака, кот или птица

make_animal_sound(собака)  # "Гав!"

````python



## Ячейка 9: Пример полиморфизма (Code)


# Полиморфизм в действии
print("🎭 Полиморфизм в действии:\n")

# Базовый класс
class Shape:
    def area(self):
        """Вычислить площадь фигуры."""
        raise NotImplementedError("Метод area() должен быть реализован в дочернем классе")
    
    def perimeter(self):
        """Вычислить периметр фигуры."""
        raise NotImplementedError("Метод perimeter() должен быть реализован в дочернем классе")
    
    def info(self):
        """Информация о фигуре."""
        return f"Площадь: {self.area():.2f}, периметр: {self.perimeter():.2f}"

# Классы-наследники
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14159 * self.radius ** 2
    
    def perimeter(self):
        return 2 * 3.14159 * self.radius
    
    def info(self):
        return f"Круг радиусом {self.radius}. " + super().info()

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)
    
    def info(self):
        return f"Прямоугольник {self.width}x{self.height}. " + super().info()

class Triangle(Shape):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    
    def area(self):
        # Формула Герона
        p = self.perimeter() / 2
        return (p * (p - self.a) * (p - self.b) * (p - self.c)) ** 0.5
    
    def perimeter(self):
        return self.a + self.b + self.c
    
    def info(self):
        return f"Треугольник со сторонами {self.a}, {self.b}, {self.c}. " + super().info()

# Демонстрация полиморфизма
print("Создаём фигуры:")
shapes = [
    Circle(5),
    Rectangle(4, 6),
    Triangle(3, 4, 5),
    Circle(2.5),
    Rectangle(10, 3)
]

print("\n📊 Информация о всех фигурах:")
for shape in shapes:
    print(f"  {shape.info()}")

# Полиморфная функция
print("\n🧮 Полиморфные операции:")

def print_total_area(shapes_list):
    """Вычислить общую площадь всех фигур."""
    total = 0
    for shape in shapes_list:
        total += shape.area()  # Вызываем area() для каждой фигуры
    print(f"Общая площадь всех фигур: {total:.2f}")

def print_total_perimeter(shapes_list):
    """Вычислить общий периметр всех фигур."""
    total = 0
    for shape in shapes_list:
        total += shape.perimeter()  # Вызываем perimeter() для каждой фигуры
    print(f"Общий периметр всех фигур: {total:.2f}")

def find_largest_shape(shapes_list):
    """Найти фигуру с наибольшей площадью."""
    if not shapes_list:
        return None
    
    largest = shapes_list[0]
    for shape in shapes_list[1:]:
        if shape.area() > largest.area():
            largest = shape
    return largest

# Используем полиморфные функции
print_total_area(shapes)
print_total_perimeter(shapes)

largest = find_largest_shape(shapes)
if largest:
    print(f"Самая большая фигура: {largest.info()}")

# Полиморфизм в коллекциях
print("\n🗂️ Полиморфизм с коллекциями:")

# Список разных фигур
mixed_shapes = [
    {"type": Circle(3), "name": "Круглый стол"},
    {"type": Rectangle(2, 4), "name": "Прямоугольная доска"},
    {"type": Triangle(5, 5, 5), "name": "Треугольная крыша"}
]

print("Площади разных объектов:")
for item in mixed_shapes:
    shape = item["type"]
    name = item["name"]
    print(f"  {name}: площадь = {shape.area():.2f}")

# Полиморфизм с условиями
print("\n🔍 Фильтрация фигур:")

def filter_large_shapes(shapes_list, min_area):
    """Отфильтровать фигуры с площадью больше заданной."""
    result = []
    for shape in shapes_list:
        if shape.area() > min_area:
            result.append(shape)
    return result

large_shapes = filter_large_shapes(shapes, 20)
print(f"Фигуры с площадью > 20: {len(large_shapes)} шт.")
for shape in large_shapes:
    print(f"  {shape.info()}")

print("\n🎯 Суть полиморфизма:")
print("Мы можем работать с Circle, Rectangle и Triangle одинаково,")
print("потому что у всех есть методы area() и perimeter().")
print("Код, который использует эти методы, не знает и не должен знать,")
print("с каким именно типом фигуры он работает!")
````

````python


## Ячейка 11: Пример утиной типизации (Code)


# Утиная типизация в действии
print("🦆 Утиная типизация в действии:\n")

# Пример 1: Простая утиная типизация
print("1. Простая утиная типизация:\n")

class Dog:
    def speak(self):
        return "Гав!"
    
    def walk(self):
        return "Собака идёт на четырёх лапах"

class Cat:
    def speak(self):
        return "Мяу!"
    
    def walk(self):
        return "Кот крадётся бесшумно"

class Robot:
    def speak(self):
        return "Бип-буп!"
    
    def walk(self):
        return "Робот движется на колёсах"

class Car:
    # У машины нет метода speak()!
    def drive(self):
        return "Машина едет"

# Полиморфная функция (утиная типизация)
def make_it_talk(thing):
    """Заставить что-то говорить."""
    try:
        print(thing.speak())
    except AttributeError:
        print(f"Это {type(thing).__name__} не умеет говорить!")

def make_it_move(thing):
    """Заставить что-то двигаться."""
    # Проверяем наличие метода walk или drive
    if hasattr(thing, 'walk'):
        print(thing.walk())
    elif hasattr(thing, 'drive'):
        print(thing.drive())
    else:
        print(f"Это {type(thing).__name__} не умеет двигаться!")

print("Заставляем разных существ и объекты говорить:")
make_it_talk(Dog())
make_it_talk(Cat())
make_it_talk(Robot())
make_it_talk(Car())  # Ошибка! У Car нет метода speak()

print("\nЗаставляем двигаться:")
make_it_move(Dog())
make_it_move(Cat())
make_it_move(Robot())
make_it_move(Car())

# Пример 2: Более сложный пример
print("\n2. Утиная типизация с файлоподобными объектами:\n")

class StringWriter:
    """Класс, который ведёт себя как файл (имеет write())."""
    
    def __init__(self):
        self.content = []
    
    def write(self, text):
        """Записать текст."""
        self.content.append(text)
    
    def get_content(self):
        """Получить записанный текст."""
        return "".join(self.content)

class FileWriter:
    """Класс для записи в файл."""
    
    def __init__(self, filename):
        self.filename = filename
    
    def write(self, text):
        """Записать текст в файл."""
        with open(self.filename, 'a', encoding='utf-8') as f:
            f.write(text)
    
    def read(self):
        """Прочитать файл."""
        with open(self.filename, 'r', encoding='utf-8') as f:
            return f.read()

class Logger:
    """Простой логгер."""
    
    def __init__(self, output):
        # output должен иметь метод write()
        self.output = output
    
    def log(self, message):
        """Записать сообщение в лог."""
        timestamp = "2024-01-25 10:00:00"  # Здесь была бы реальная дата
        self.output.write(f"[{timestamp}] {message}\n")

# Используем утиную типизацию
print("Используем логгер с разными выводами:")

# Логгируем в строку
string_writer = StringWriter()
logger1 = Logger(string_writer)
logger1.log("Первое сообщение")
logger1.log("Второе сообщение")
print("Лог в строке:")
print(string_writer.get_content())

# Логгируем в файл (имитируем)
print("\nЛог в файле (имитация):")
file_writer = FileWriter("log.txt")  # В реальности создался бы файл
logger2 = Logger(file_writer)
logger2.log("Сообщение в файл")
print("Сообщение записано в файл 'log.txt'")

# Пример 3: Утиная типизация с операторами
print("\n3. Утиная типизация и операторы:\n")

class Vector:
    """Простой класс вектора."""
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        """Сложение векторов."""
        # Нам неважно, что такое other
        # Важно, чтобы у него были атрибуты x и y
        return Vector(self.x + other.x, self.y + other.y)
    
    def __str__(self):
        return f"Vector({self.x}, {self.y})"

class Point:
    """Класс точки."""
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f"Point({self.x}, {self.y})"

# Вектор + Вектор
v1 = Vector(1, 2)
v2 = Vector(3, 4)
result1 = v1 + v2  # Vector.__add__ вызывается с другим Vector
print(f"Vector + Vector: {v1} + {v2} = {result1}")

# Вектор + Point (утиная типизация!)
p = Point(5, 6)
result2 = v1 + p  # Vector.__add__ вызывается с Point
print(f"Vector + Point: {v1} + {p} = {result2}")

print("\n🔍 Важно для утиной типизации:")
print("Python не проверяет, что p - это Vector.")
print("Python просто пытается выполнить v1.__add__(p).")
print("Так как у p есть атрибуты x и y, всё работает!")
````

````python
# Сравнение разных подходов
print("📊 Сравнение разных подходов:\n")

# Подход 1: Жёсткая типизация (как в C++/Java)
print("1. Жёсткая типизация (не по-питоньи):")

def calculate_area_1(shape):
    """Вычисление площади с проверкой типа."""
    if isinstance(shape, Circle):
        return 3.14 * shape.radius ** 2
    elif isinstance(shape, Rectangle):
        return shape.width * shape.height
    elif isinstance(shape, Triangle):
        # Формула Герона
        p = (shape.a + shape.b + shape.c) / 2
        return (p * (p - shape.a) * (p - shape.b) * (p - shape.c)) ** 0.5
    else:
        raise TypeError(f"Неизвестный тип фигуры: {type(shape)}")

# Подход 2: Утиная типизация (по-питоньи)
print("\n2. Утиная типизация (по-питоньи):")

def calculate_area_2(shape):
    """Вычисление площади с утиной типизацией."""
    # Просто вызываем метод area()
    # Если у shape есть этот метод - всё работает
    return shape.area()

# Создадим простые классы для демонстрации
class SimpleCircle:
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius ** 2

class SimpleRectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height

class NewShape:
    """Новая фигура, которую мы добавили позже."""
    def __init__(self, side):
        self.side = side
    
    def area(self):
        return self.side ** 2  # Площадь квадрата

# Тестируем оба подхода
print("\n🧪 Тестируем оба подхода:")

shapes_for_test = [
    SimpleCircle(5),
    SimpleRectangle(4, 6),
    NewShape(3)  # Новая фигура, о которой первая функция не знает
]

print("\nС жёсткой типизацией (подход 1):")
for shape in shapes_for_test[:2]:  # Только первые две, третья вызовет ошибку
    try:
        area = calculate_area_1(shape)
        print(f"  Площадь: {area:.2f}")
    except TypeError as e:
        print(f"  Ошибка: {e}")

# NewShape вызовет ошибку в первом подходе
try:
    calculate_area_1(shapes_for_test[2])
except TypeError as e:
    print(f"  Для NewShape: {e}")

print("\nС утиной типизацией (подход 2):")
for shape in shapes_for_test:
    area = calculate_area_2(shape)  # Работает со всеми!
    print(f"  Площадь: {area:.2f}")

print("\n🎯 Выводы:")
print("1. Жёсткая типизация:")
print("   - Нужно знать все типы заранее")
print("   - При добавлении нового типа нужно менять функцию")
print("   - Много проверок isinstance()")

print("\n2. Утиная типизация:")
print("   - Нужно только чтобы был метод area()")
print("   - Новые типы добавляются без изменения функции")
print("   - Меньше кода, больше гибкости")

print("\n💡 Правило Python:")
print("«Лучше просить прощения, чем разрешения» (EAFP)")
print("  - EAFP: Easier to Ask for Forgiveness than Permission")
print("  - Пытаемся сделать, и если не получается - обрабатываем ошибку")
print("\n«Смотри прежде, чем прыгать» (LBYL)")
print("  - LBYL: Look Before You Leap")
print("  - Сначала проверяем, потом делаем")

print("\nПример EAFP (Python-way):")
print("""
try:
    value = obj.method()
except AttributeError:
    # Обработка случая, когда метода нет
    value = default_value
""")

print("\nПример LBYL (не совсем Python-way):")
print("""
if hasattr(obj, 'method'):
    value = obj.method()
else:
    value = default_value
""")
````

## 📚 Итоги лекции

### Ключевые концепции:
1. **Наследование** — создание новых классов на основе существующих
2. **Переопределение методов** — изменение поведения унаследованных методов
3. **`super()`** — доступ к методам и атрибутам родительского класса
4. **Полиморфизм** — возможность объектов разных классов отвечать на одни команды
5. **Утиная типизация** — "если у объекта есть нужный метод, он подходит"

### 🎯 Правила хорошего тона:

#### 1. Про наследование:


````python

# Используйте наследование только для отношений "является"
class Car(Vehicle):          # ✅ Автомобиль является транспортным средством
class Student(Person):       # ✅ Студент является человеком
# class Car(Engine):         # ❌ Автомобиль не является двигателем!
````

#### 2. Про переопределение методов:

````python
class Parent:
    def method(self):
        # базовая реализация

class Child(Parent):
    def method(self):
        # Сначала вызываем метод родителя
        super().method()
        # Потом добавляем свою логику
        print("Дополнительная логика")
````

#### 3. Про утиную типизацию:

````python
# Хорошо: проверяем наличие метода
if hasattr(obj, 'method'):
    obj.method()

# Ещё лучше: пробуем вызвать и обрабатываем ошибку
try:
    obj.method()
except AttributeError:
    print("У объекта нет этого метода")
````
