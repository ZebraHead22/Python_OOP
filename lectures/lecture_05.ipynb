### –°–∏–Ω—Ç–∞–∫—Å–∏—Å –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –≤ Python:
```python
class –†–æ–¥–∏—Ç–µ–ª—å—Å–∫–∏–π–ö–ª–∞—Å—Å:

    # –º–µ—Ç–æ–¥—ã –∏ –∞—Ç—Ä–∏–±—É—Ç—ã –¥–æ—á–µ—Ä–Ω–µ–≥–æ –∫–ª–∞—Å—Å–∞
    # + –≤—Å—ë, —á—Ç–æ –µ—Å—Ç—å —É —Ä–æ–¥–∏—Ç–µ–ª—è

````python


## –Ø—á–µ–π–∫–∞ 3: –ü—Ä–∏–º–µ—Ä –ø—Ä–æ—Å—Ç–æ–≥–æ –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è (Code)


# –ü—Ä–æ—Å—Ç–æ–µ –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ
print("üë∂ –ü—Ä–æ—Å—Ç–æ–µ –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ:\n")

# –†–æ–¥–∏—Ç–µ–ª—å—Å–∫–∏–π –∫–ª–∞—Å—Å
class Animal:
    """–ë–∞–∑–æ–≤—ã–π –∫–ª–∞—Å—Å –¥–ª—è –≤—Å–µ—Ö –∂–∏–≤–æ—Ç–Ω—ã—Ö."""
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.is_alive = True
    
    def eat(self):
        """–ñ–∏–≤–æ—Ç–Ω–æ–µ –µ—Å—Ç."""
        print(f"{self.name} –µ—Å—Ç...")
    
    def sleep(self):
        """–ñ–∏–≤–æ—Ç–Ω–æ–µ —Å–ø–∏—Ç."""
        print(f"{self.name} —Å–ø–∏—Ç...")
    
    def make_sound(self):
        """–ñ–∏–≤–æ—Ç–Ω–æ–µ –∏–∑–¥–∞—ë—Ç –∑–≤—É–∫."""
        print(f"{self.name} –∏–∑–¥–∞—ë—Ç –∑–≤—É–∫...")
    
    def get_info(self):
        """–ü–æ–ª—É—á–∏—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –∂–∏–≤–æ—Ç–Ω–æ–º."""
        return f"{self.name}, {self.age} –ª–µ—Ç, {'–∂–∏–≤' if self.is_alive else '–Ω–µ –∂–∏–≤'}"

# –î–æ—á–µ—Ä–Ω–∏–π –∫–ª–∞—Å—Å
class Dog(Animal):  # Dog –Ω–∞—Å–ª–µ–¥—É–µ—Ç –æ—Ç Animal
    """–ö–ª–∞—Å—Å —Å–æ–±–∞–∫–∏."""
    
    def __init__(self, name, age, breed):
        # –í—ã–∑—ã–≤–∞–µ–º –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä —Ä–æ–¥–∏—Ç–µ–ª—å—Å–∫–æ–≥–æ –∫–ª–∞—Å—Å–∞
        super().__init__(name, age)  # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Ç–æ, —á—Ç–æ –µ—Å—Ç—å —É Animal
        self.breed = breed           # –î–æ–±–∞–≤–ª—è–µ–º –Ω–æ–≤–æ–µ —Å–≤–æ–π—Å—Ç–≤–æ
        self.tricks = []             # –ï—â—ë –æ–¥–Ω–æ –Ω–æ–≤–æ–µ —Å–≤–æ–π—Å—Ç–≤–æ
    
    # –ü–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª—è–µ–º –º–µ—Ç–æ–¥ —Ä–æ–¥–∏—Ç–µ–ª—è
    def make_sound(self):
        """–°–æ–±–∞–∫–∞ –ª–∞–µ—Ç."""
        print(f"{self.name} –≥–æ–≤–æ—Ä–∏—Ç: –ì–∞–≤! –ì–∞–≤!")
    
    # –î–æ–±–∞–≤–ª—è–µ–º –Ω–æ–≤—ã–π –º–µ—Ç–æ–¥
    def learn_trick(self, trick):
        """–°–æ–±–∞–∫–∞ —É—á–∏—Ç –Ω–æ–≤—ã–π —Ç—Ä—é–∫."""
        self.tricks.append(trick)
        print(f"{self.name} –≤—ã—É—á–∏–ª —Ç—Ä—é–∫: {trick}")
    
    # –†–∞—Å—à–∏—Ä—è–µ–º –º–µ—Ç–æ–¥ —Ä–æ–¥–∏—Ç–µ–ª—è
    def get_info(self):
        """–ü–æ–ª—É—á–∏—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Å–æ–±–∞–∫–µ."""
        base_info = super().get_info()  # –í—ã–∑—ã–≤–∞–µ–º –º–µ—Ç–æ–¥ —Ä–æ–¥–∏—Ç–µ–ª—è
        return f"{base_info}, –ø–æ—Ä–æ–¥–∞: {self.breed}, —Ç—Ä—é–∫–æ–≤: {len(self.tricks)}"

# –î—Ä—É–≥–æ–π –¥–æ—á–µ—Ä–Ω–∏–π –∫–ª–∞—Å—Å
class Cat(Animal):
    """–ö–ª–∞—Å—Å –∫–æ—Ç–∞."""
    
    def __init__(self, name, age, color):
        super().__init__(name, age)
        self.color = color
        self.lives = 9  # –£ –∫–æ—Ç–∞ 9 –∂–∏–∑–Ω–µ–π!
    
    def make_sound(self):
        """–ö–æ—Ç –º—è—É–∫–∞–µ—Ç."""
        print(f"{self.name} –≥–æ–≤–æ—Ä–∏—Ç: –ú—è—É!")
    
    def lose_life(self):
        """–ö–æ—Ç —Ç–µ—Ä—è–µ—Ç –∂–∏–∑–Ω—å."""
        if self.lives > 0:
            self.lives -= 1
            print(f"{self.name} —Ç–µ—Ä—è–µ—Ç –∂–∏–∑–Ω—å. –û—Å—Ç–∞–ª–æ—Å—å: {self.lives}")
            if self.lives == 0:
                self.is_alive = False
        else:
            print(f"{self.name} —É–∂–µ —É–º–µ—Ä :(")

# –°–æ–∑–¥–∞—ë–º –æ–±—ä–µ–∫—Ç—ã
print("–°–æ–∑–¥–∞—ë–º –∂–∏–≤–æ—Ç–Ω—ã—Ö:")
animal = Animal("–ù–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–µ –∂–∏–≤–æ—Ç–Ω–æ–µ", 3)
dog = Dog("–ë–æ–±–∏–∫", 5, "–î–≤–æ—Ä–Ω—è–∂–∫–∞")
cat = Cat("–ú—É—Ä–∑–∏–∫", 2, "—Ä—ã–∂–∏–π")

print(f"\n1. –ë–∞–∑–æ–≤–æ–µ –∂–∏–≤–æ—Ç–Ω–æ–µ: {animal.get_info()}")
print(f"2. –°–æ–±–∞–∫–∞: {dog.get_info()}")
print(f"3. –ö–æ—Ç: {cat.get_info()}, —Ü–≤–µ—Ç: {cat.color}, –∂–∏–∑–Ω–µ–π: {cat.lives}")

# –ò—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç–æ–¥—ã
print("\nüêæ –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è –ø–æ–≤–µ–¥–µ–Ω–∏—è:")
print("\n–ñ–∏–≤–æ—Ç–Ω–æ–µ:")
animal.eat()
animal.make_sound()

print("\n–°–æ–±–∞–∫–∞:")
dog.eat()            # –£–Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–Ω—ã–π –º–µ—Ç–æ–¥
dog.make_sound()     # –ü–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã–π –º–µ—Ç–æ–¥
dog.learn_trick("–°–∏–¥–µ—Ç—å")  # –°–≤–æ–π –º–µ—Ç–æ–¥
dog.learn_trick("–õ–µ–∂–∞—Ç—å")
print(f"–¢—Ä—é–∫–∏ —Å–æ–±–∞–∫–∏: {dog.tricks}")

print("\n–ö–æ—Ç:")
cat.eat()
cat.make_sound()
cat.lose_life()      # –°–≤–æ–π –º–µ—Ç–æ–¥
print(f"–ñ–∏–∑–Ω–µ–π –æ—Å—Ç–∞–ª–æ—Å—å: {cat.lives}")

# –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ
print("\nüîç –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è:")
print(f"dog —è–≤–ª—è–µ—Ç—Å—è Animal? {isinstance(dog, Animal)}")
print(f"cat —è–≤–ª—è–µ—Ç—Å—è Animal? {isinstance(cat, Animal)}")
print(f"animal —è–≤–ª—è–µ—Ç—Å—è Dog? {isinstance(animal, Dog)}")
print(f"animal —è–≤–ª—è–µ—Ç—Å—è Cat? {isinstance(animal, Cat)}")

print(f"\nDog —è–≤–ª—è–µ—Ç—Å—è –ø–æ–¥–∫–ª–∞—Å—Å–æ–º Animal? {issubclass(Dog, Animal)}")
print(f"Cat —è–≤–ª—è–µ—Ç—Å—è –ø–æ–¥–∫–ª–∞—Å—Å–æ–º Animal? {issubclass(Cat, Animal)}")
print(f"Animal —è–≤–ª—è–µ—Ç—Å—è –ø–æ–¥–∫–ª–∞—Å—Å–æ–º Dog? {issubclass(Animal, Dog)}")
````

- –ú–æ–∂–Ω–æ –ø–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª—è—Ç—å –¥–∞–∂–µ —Å–ø–µ—Ü–∏–∞–ª—å–Ω—ã–µ –º–µ—Ç–æ–¥—ã (`__str__`, `__init__` –∏ —Ç.–¥.)

````python
# –ü–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤
print("üîÑ –ü–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤:\n")

class Vehicle:
    """–ë–∞–∑–æ–≤—ã–π –∫–ª–∞—Å—Å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω–æ–≥–æ —Å—Ä–µ–¥—Å—Ç–≤–∞."""
    
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
        self.speed = 0
        self.is_moving = False
    
    def start(self):
        """–ó–∞–ø—É—Å—Ç–∏—Ç—å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω–æ–µ —Å—Ä–µ–¥—Å—Ç–≤–æ."""
        if not self.is_moving:
            self.is_moving = True
            print(f"{self.brand} {self.model} –∑–∞–≤—ë–ª—Å—è")
        else:
            print(f"{self.brand} {self.model} —É–∂–µ –∑–∞–≤–µ–¥—ë–Ω")
    
    def stop(self):
        """–û—Å—Ç–∞–Ω–æ–≤–∏—Ç—å —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω–æ–µ —Å—Ä–µ–¥—Å—Ç–≤–æ."""
        if self.is_moving:
            self.is_moving = False
            self.speed = 0
            print(f"{self.brand} {self.model} –æ—Å—Ç–∞–Ω–æ–≤–∏–ª—Å—è")
        else:
            print(f"{self.brand} {self.model} —É–∂–µ –æ—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω")
    
    def accelerate(self, amount):
        """–£—Å–∫–æ—Ä–∏—Ç—å—Å—è."""
        if self.is_moving:
            self.speed += amount
            print(f"{self.brand} {self.model} —É—Å–∫–æ—Ä–∏–ª—Å—è –¥–æ {self.speed} –∫–º/—á")
        else:
            print(f"–°–Ω–∞—á–∞–ª–∞ –Ω—É–∂–Ω–æ –∑–∞–≤–µ—Å—Ç–∏ {self.brand} {self.model}")
    
    def get_info(self):
        """–ü–æ–ª—É—á–∏—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω–æ–º —Å—Ä–µ–¥—Å—Ç–≤–µ."""
        return f"{self.brand} {self.model} ({self.year}), —Å–∫–æ—Ä–æ—Å—Ç—å: {self.speed} –∫–º/—á"

class Car(Vehicle):
    """–ö–ª–∞—Å—Å –∞–≤—Ç–æ–º–æ–±–∏–ª—è."""
    
    def __init__(self, brand, model, year, fuel_type):
        super().__init__(brand, model, year)
        self.fuel_type = fuel_type
        self.fuel_level = 100  # –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö
        self.doors = 4
    
    # –ü–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª—è–µ–º –º–µ—Ç–æ–¥ start
    def start(self):
        """–ó–∞–ø—É—Å—Ç–∏—Ç—å –∞–≤—Ç–æ–º–æ–±–∏–ª—å (–ø—Ä–æ–≤–µ—Ä—è–µ–º —Ç–æ–ø–ª–∏–≤–æ)."""
        if self.fuel_level > 0:
            super().start()  # –í—ã–∑—ã–≤–∞–µ–º –º–µ—Ç–æ–¥ —Ä–æ–¥–∏—Ç–µ–ª—è
            print(f"  –£—Ä–æ–≤–µ–Ω—å —Ç–æ–ø–ª–∏–≤–∞: {self.fuel_level}%")
        else:
            print(f"{self.brand} {self.model}: –Ω–µ—Ç —Ç–æ–ø–ª–∏–≤–∞!")
    
    # –ü–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª—è–µ–º –º–µ—Ç–æ–¥ accelerate
    def accelerate(self, amount):
        """–£—Å–∫–æ—Ä–∏—Ç—å—Å—è (—Ä–∞—Å—Ö–æ–¥—É–µ—Ç —Ç–æ–ø–ª–∏–≤–æ)."""
        if self.is_moving:
            # –†–∞—Å—Ö–æ–¥—É–µ–º —Ç–æ–ø–ª–∏–≤–æ –ø—Ä–∏ —É—Å–∫–æ—Ä–µ–Ω–∏–∏
            fuel_used = amount * 0.5
            if self.fuel_level >= fuel_used:
                self.fuel_level -= fuel_used
                super().accelerate(amount)  # –í—ã–∑—ã–≤–∞–µ–º –º–µ—Ç–æ–¥ —Ä–æ–¥–∏—Ç–µ–ª—è
                print(f"  –ò–∑—Ä–∞—Å—Ö–æ–¥–æ–≤–∞–Ω–æ —Ç–æ–ø–ª–∏–≤–∞: {fuel_used:.1f}%")
                print(f"  –û—Å—Ç–∞—Ç–æ–∫ —Ç–æ–ø–ª–∏–≤–∞: {self.fuel_level:.1f}%")
            else:
                print(f"{self.brand} {self.model}: –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Ç–æ–ø–ª–∏–≤–∞!")
        else:
            super().accelerate(amount)  # –í—ã–∑—ã–≤–∞–µ–º –º–µ—Ç–æ–¥ —Ä–æ–¥–∏—Ç–µ–ª—è
    
    # –î–æ–±–∞–≤–ª—è–µ–º –Ω–æ–≤—ã–π –º–µ—Ç–æ–¥
    def refuel(self, amount):
        """–ó–∞–ø—Ä–∞–≤–∏—Ç—å –∞–≤—Ç–æ–º–æ–±–∏–ª—å."""
        self.fuel_level = min(100, self.fuel_level + amount)
        print(f"{self.brand} {self.model} –∑–∞–ø—Ä–∞–≤–ª–µ–Ω. –¢–æ–ø–ª–∏–≤–æ: {self.fuel_level}%")
    
    # –†–∞—Å—à–∏—Ä—è–µ–º –º–µ—Ç–æ–¥ get_info
    def get_info(self):
        """–ü–æ–ª—É—á–∏—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ–± –∞–≤—Ç–æ–º–æ–±–∏–ª–µ."""
        base_info = super().get_info()
        return f"{base_info}, —Ç–æ–ø–ª–∏–≤–æ: {self.fuel_type}, –¥–≤–µ—Ä–µ–π: {self.doors}"

class Bicycle(Vehicle):
    """–ö–ª–∞—Å—Å –≤–µ–ª–æ—Å–∏–ø–µ–¥–∞."""
    
    def __init__(self, brand, model, year, type):
        super().__init__(brand, model, year)
        self.type = type  # –ì–æ—Ä–Ω—ã–π, —à–æ—Å—Å–µ–π–Ω—ã–π –∏ —Ç.–¥.
        self.gears = 18
        self.current_gear = 1
    
    # –ü–æ–ª–Ω–æ—Å—Ç—å—é –ø–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª—è–µ–º –º–µ—Ç–æ–¥ start
    def start(self):
        """–ù–∞—á–∞—Ç—å –¥–≤–∏–∂–µ–Ω–∏–µ –Ω–∞ –≤–µ–ª–æ—Å–∏–ø–µ–¥–µ."""
        self.is_moving = True
        self.speed = 5  # –ù–∞—á–∏–Ω–∞–µ–º —Å –Ω–µ–±–æ–ª—å—à–æ–π —Å–∫–æ—Ä–æ—Å—Ç–∏
        print(f"{self.brand} {self.model}: –Ω–∞—á–∏–Ω–∞–µ–º –∫—Ä—É—Ç–∏—Ç—å –ø–µ–¥–∞–ª–∏!")
        print(f"  –ù–∞—á–∞–ª—å–Ω–∞—è —Å–∫–æ—Ä–æ—Å—Ç—å: {self.speed} –∫–º/—á")
    
    # –ü–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª—è–µ–º –º–µ—Ç–æ–¥ accelerate
    def accelerate(self, amount):
        """–£—Å–∫–æ—Ä–∏—Ç—å—Å—è –Ω–∞ –≤–µ–ª–æ—Å–∏–ø–µ–¥–µ (–ø–µ—Ä–µ–∫–ª—é—á–∞–µ–º –ø–µ—Ä–µ–¥–∞—á–∏)."""
        if self.is_moving:
            # –ù–∞ –≤–µ–ª–æ—Å–∏–ø–µ–¥–µ —É—Å–∫–æ—Ä—è–µ–º—Å—è –ø–µ—Ä–µ–∫–ª—é—á–µ–Ω–∏–µ–º –ø–µ—Ä–µ–¥–∞—á
            if amount > 0:
                if self.current_gear < self.gears:
                    self.current_gear += 1
                    self.speed += 3  # –ö–∞–∂–¥–∞—è –ø–µ—Ä–µ–¥–∞—á–∞ –¥–æ–±–∞–≤–ª—è–µ—Ç —Å–∫–æ—Ä–æ—Å—Ç—å
                    print(f"{self.brand} {self.model}: –ø–µ—Ä–µ–∫–ª—é—á–∏–ª—Å—è –Ω–∞ {self.current_gear} –ø–µ—Ä–µ–¥–∞—á—É")
                    print(f"  –°–∫–æ—Ä–æ—Å—Ç—å: {self.speed} –∫–º/—á")
                else:
                    print(f"{self.brand} {self.model}: —É–∂–µ –Ω–∞ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π –ø–µ—Ä–µ–¥–∞—á–µ!")
            else:
                print(f"{self.brand} {self.model}: –¥–ª—è –∑–∞–º–µ–¥–ª–µ–Ω–∏—è –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ brake()")
        else:
            print(f"{self.brand} {self.model}: —Å–Ω–∞—á–∞–ª–∞ –Ω—É–∂–Ω–æ –Ω–∞—á–∞—Ç—å –¥–≤–∏–∂–µ–Ω–∏–µ!")
    
    # –î–æ–±–∞–≤–ª—è–µ–º –Ω–æ–≤—ã–π –º–µ—Ç–æ–¥
    def brake(self):
        """–ó–∞—Ç–æ—Ä–º–æ–∑–∏—Ç—å –Ω–∞ –≤–µ–ª–æ—Å–∏–ø–µ–¥–µ."""
        if self.is_moving and self.speed > 0:
            self.speed = max(0, self.speed - 5)
            print(f"{self.brand} {self.model} —Ç–æ—Ä–º–æ–∑–∏—Ç. –°–∫–æ—Ä–æ—Å—Ç—å: {self.speed} –∫–º/—á")
    
    # –†–∞—Å—à–∏—Ä—è–µ–º –º–µ—Ç–æ–¥ get_info
    def get_info(self):
        """–ü–æ–ª—É—á–∏—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –≤–µ–ª–æ—Å–∏–ø–µ–¥–µ."""
        base_info = super().get_info()
        return f"{base_info}, —Ç–∏–ø: {self.type}, –ø–µ—Ä–µ–¥–∞—á–∞: {self.current_gear}/{self.gears}"

# –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è
print("–°–æ–∑–¥–∞—ë–º —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–Ω—ã–µ —Å—Ä–µ–¥—Å—Ç–≤–∞:")
car = Car("Toyota", "Camry", 2020, "–±–µ–Ω–∑–∏–Ω")
bike = Bicycle("Trek", "Marlin", 2021, "–≥–æ—Ä–Ω—ã–π")

print(f"\n1. –ê–≤—Ç–æ–º–æ–±–∏–ª—å: {car.get_info()}")
print(f"2. –í–µ–ª–æ—Å–∏–ø–µ–¥: {bike.get_info()}")

print("\nüöó –†–∞–±–æ—Ç–∞ —Å –∞–≤—Ç–æ–º–æ–±–∏–ª–µ–º:")
car.start()
car.accelerate(20)
car.accelerate(30)
car.accelerate(50)  # –ò–∑—Ä–∞—Å—Ö–æ–¥—É–µ–º —Ç–æ–ø–ª–∏–≤–æ
car.refuel(50)
car.accelerate(20)
car.stop()

print("\nüö≤ –†–∞–±–æ—Ç–∞ —Å –≤–µ–ª–æ—Å–∏–ø–µ–¥–æ–º:")
bike.start()  # –ü–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã–π –º–µ—Ç–æ–¥
bike.accelerate(10)  # –ü–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã–π –º–µ—Ç–æ–¥
bike.accelerate(10)
bike.brake()  # –°–≤–æ–π –º–µ—Ç–æ–¥
bike.accelerate(-5)  # –ü–æ–ø—ã—Ç–∫–∞ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–≥–æ —É—Å–∫–æ—Ä–µ–Ω–∏—è
bike.stop()

print("\nüîç –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤:")
print("–£ –∞–≤—Ç–æ–º–æ–±–∏–ª—è –∏ –≤–µ–ª–æ—Å–∏–ø–µ–¥–∞ –µ—Å—Ç—å –º–µ—Ç–æ–¥—ã start(), accelerate(), stop(), get_info()")
print("–ù–æ —Ä–∞–±–æ—Ç–∞—é—Ç –æ–Ω–∏ –ø–æ-—Ä–∞–∑–Ω–æ–º—É - —ç—Ç–æ –∏ –µ—Å—Ç—å –ø–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤!")
````

## ü¶∏ –§—É–Ω–∫—Ü–∏—è `super()`

**`super()`** ‚Äî —ç—Ç–æ —Ñ—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –æ–±—ä–µ–∫—Ç-–ø—Ä–æ–∫—Å–∏ —Ä–æ–¥–∏—Ç–µ–ª—å—Å–∫–æ–≥–æ –∫–ª–∞—Å—Å–∞.

### –ó–∞—á–µ–º –Ω—É–∂–µ–Ω `super()`?
1. **–í—ã–∑–æ–≤ –º–µ—Ç–æ–¥–æ–≤ —Ä–æ–¥–∏—Ç–µ–ª—è** ‚Äî –∫–æ–≥–¥–∞ –Ω—É–∂–Ω–æ –¥–æ–ø–æ–ª–Ω–∏—Ç—å, –∞ –Ω–µ –∑–∞–º–µ–Ω–∏—Ç—å
2. **–î–æ—Å—Ç—É–ø –∫ –∞—Ç—Ä–∏–±—É—Ç–∞–º —Ä–æ–¥–∏—Ç–µ–ª—è** ‚Äî –∫–æ–≥–¥–∞ –Ω—É–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ä–æ–¥–∏—Ç–µ–ª—å—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ
3. **–†–µ—à–µ–Ω–∏–µ –ø—Ä–æ–±–ª–µ–º –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è** ‚Äî –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –ø–æ—Ä—è–¥–æ–∫ –≤—ã–∑–æ–≤–∞ –º–µ—Ç–æ–¥–æ–≤

### –ö–∞–∫ —Ä–∞–±–æ—Ç–∞–µ—Ç `super()`?


````python

class Parent:
    def method(self):
        print("–†–æ–¥–∏—Ç–µ–ª—å—Å–∫–∏–π –º–µ—Ç–æ–¥")

class Child(Parent):
    def method(self):
        super().method()  # –í—ã–∑—ã–≤–∞–µ–º –º–µ—Ç–æ–¥ —Ä–æ–¥–∏—Ç–µ–ª—è
        print("–î–æ—á–µ—Ä–Ω–∏–π –º–µ—Ç–æ–¥")
````

````python
# –ü—Ä—è–º–æ–µ –æ–±—Ä–∞—â–µ–Ω–∏–µ (–Ω–µ —Ä–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è)
class Child(Parent):
    def method(self):
        Parent.method(self)  # –ñ—ë—Å—Ç–∫–æ –ø—Ä–∏–≤—è–∑–∞–Ω–æ –∫ Parent
        print("–î–æ—á–µ—Ä–Ω–∏–π –º–µ—Ç–æ–¥")

# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ super() (—Ä–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è)
class Child(Parent):
    def method(self):
        super().method()  # –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –Ω–∞—Ö–æ–¥–∏—Ç —Ä–æ–¥–∏—Ç–µ–ª—è
        print("–î–æ—á–µ—Ä–Ω–∏–π –º–µ—Ç–æ–¥")
````

````python



## –Ø—á–µ–π–∫–∞ 7: –ü—Ä–∏–º–µ—Ä—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è `super()` (Code)


# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ super()
print("ü¶∏ –§—É–Ω–∫—Ü–∏—è super() –≤ –¥–µ–π—Å—Ç–≤–∏–∏:\n")

# –ü—Ä–∏–º–µ—Ä 1: super() –≤ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä–µ
print("1. super() –≤ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä–µ:\n")

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f"Person.__init__ –≤—ã–∑–≤–∞–Ω –¥–ª—è {name}")

class Employee(Person):
    def __init__(self, name, age, employee_id):
        # –í—ã–∑—ã–≤–∞–µ–º –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä —Ä–æ–¥–∏—Ç–µ–ª—è
        super().__init__(name, age)
        self.employee_id = employee_id
        print(f"Employee.__init__ –≤—ã–∑–≤–∞–Ω –¥–ª—è {name}, ID: {employee_id}")

class Manager(Employee):
    def __init__(self, name, age, employee_id, department):
        # –í—ã–∑—ã–≤–∞–µ–º –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä —Ä–æ–¥–∏—Ç–µ–ª—è
        super().__init__(name, age, employee_id)
        self.department = department
        print(f"Manager.__init__ –≤—ã–∑–≤–∞–Ω –¥–ª—è {name}, –æ—Ç–¥–µ–ª: {department}")

print("–°–æ–∑–¥–∞—ë–º –æ–±—ä–µ–∫—Ç—ã:")
person = Person("–ò–≤–∞–Ω", 30)
print()
employee = Employee("–ê–Ω–Ω–∞", 25, "E001")
print()
manager = Manager("–ü—ë—Ç—Ä", 40, "M001", "IT")
print()

# –ü—Ä–∏–º–µ—Ä 2: super() –≤ –º–µ—Ç–æ–¥–∞—Ö
print("2. super() –≤ –æ–±—ã—á–Ω—ã—Ö –º–µ—Ç–æ–¥–∞—Ö:\n")

class Shape:
    def __init__(self, color):
        self.color = color
    
    def area(self):
        """–í—ã—á–∏—Å–ª–∏—Ç—å –ø–ª–æ—â–∞–¥—å."""
        return 0
    
    def info(self):
        """–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Ñ–∏–≥—É—Ä–µ."""
        return f"–§–∏–≥—É—Ä–∞ —Ü–≤–µ—Ç–∞ {self.color}, –ø–ª–æ—â–∞–¥—å: {self.area()}"

class Rectangle(Shape):
    def __init__(self, color, width, height):
        super().__init__(color)
        self.width = width
        self.height = height
    
    def area(self):
        """–ü–ª–æ—â–∞–¥—å –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–∞."""
        return self.width * self.height
    
    def info(self):
        """–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–µ."""
        base_info = super().info()  # –í—ã–∑—ã–≤–∞–µ–º info() —Ä–æ–¥–∏—Ç–µ–ª—è
        return f"{base_info}, —à–∏—Ä–∏–Ω–∞: {self.width}, –≤—ã—Å–æ—Ç–∞: {self.height}"

class Square(Rectangle):
    def __init__(self, color, side):
        # –ö–≤–∞–¥—Ä–∞—Ç - —ç—Ç–æ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ —Å —Ä–∞–≤–Ω—ã–º–∏ —Å—Ç–æ—Ä–æ–Ω–∞–º–∏
        super().__init__(color, side, side)
    
    def info(self):
        """–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –∫–≤–∞–¥—Ä–∞—Ç–µ."""
        base_info = super().info()  # –í—ã–∑—ã–≤–∞–µ–º info() —Ä–æ–¥–∏—Ç–µ–ª—è (Rectangle)
        # –ù–æ Rectangle.info() —É–∂–µ –≤—ã–∑—ã–≤–∞–µ—Ç Shape.info()
        return f"–ö–≤–∞–¥—Ä–∞—Ç. {base_info}"

print("–°–æ–∑–¥–∞—ë–º —Ñ–∏–≥—É—Ä—ã:")
rect = Rectangle("–∫—Ä–∞—Å–Ω—ã–π", 5, 10)
square = Square("—Å–∏–Ω–∏–π", 7)

print(f"–ü—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫: {rect.info()}")
print(f"–ö–≤–∞–¥—Ä–∞—Ç: {square.info()}")
print()

# –ü—Ä–∏–º–µ—Ä 3: super() —Å –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω—ã–º –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ–º
print("3. super() —Å –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω—ã–º –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ–º:\n")

class A:
    def __init__(self):
        print("A.__init__")
        self.a = "a"

class B(A):
    def __init__(self):
        print("B.__init__")
        super().__init__()
        self.b = "b"

class C(A):
    def __init__(self):
        print("C.__init__")
        super().__init__()
        self.c = "c"

class D(B, C):
    def __init__(self):
        print("D.__init__")
        super().__init__()
        self.d = "d"

print("–°–æ–∑–¥–∞—ë–º –æ–±—ä–µ–∫—Ç D (–º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ):")
obj_d = D()
print()

print("–ü–æ—Ä—è–¥–æ–∫ –≤—ã–∑–æ–≤–∞ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä–æ–≤ (MRO - Method Resolution Order):")
print(f"  D -> B -> C -> A")
print(f"  MRO –∫–ª–∞—Å—Å–∞ D: {[cls.__name__ for cls in D.__mro__]}")
print()

print("–ê—Ç—Ä–∏–±—É—Ç—ã –æ–±—ä–µ–∫—Ç–∞ D:")
print(f"  {obj_d.__dict__}")
````

````python



## –Ø—á–µ–π–∫–∞ 9: –ü—Ä–∏–º–µ—Ä –ø–æ–ª–∏–º–æ—Ä—Ñ–∏–∑–º–∞ (Code)


# –ü–æ–ª–∏–º–æ—Ä—Ñ–∏–∑–º –≤ –¥–µ–π—Å—Ç–≤–∏–∏
print("üé≠ –ü–æ–ª–∏–º–æ—Ä—Ñ–∏–∑–º –≤ –¥–µ–π—Å—Ç–≤–∏–∏:\n")

# –ë–∞–∑–æ–≤—ã–π –∫–ª–∞—Å—Å
class Shape:
    def area(self):
        """–í—ã—á–∏—Å–ª–∏—Ç—å –ø–ª–æ—â–∞–¥—å —Ñ–∏–≥—É—Ä—ã."""
        raise NotImplementedError("–ú–µ—Ç–æ–¥ area() –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω –≤ –¥–æ—á–µ—Ä–Ω–µ–º –∫–ª–∞—Å—Å–µ")
    
    def perimeter(self):
        """–í—ã—á–∏—Å–ª–∏—Ç—å –ø–µ—Ä–∏–º–µ—Ç—Ä —Ñ–∏–≥—É—Ä—ã."""
        raise NotImplementedError("–ú–µ—Ç–æ–¥ perimeter() –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω –≤ –¥–æ—á–µ—Ä–Ω–µ–º –∫–ª–∞—Å—Å–µ")
    
    def info(self):
        """–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Ñ–∏–≥—É—Ä–µ."""
        return f"–ü–ª–æ—â–∞–¥—å: {self.area():.2f}, –ø–µ—Ä–∏–º–µ—Ç—Ä: {self.perimeter():.2f}"

# –ö–ª–∞—Å—Å—ã-–Ω–∞—Å–ª–µ–¥–Ω–∏–∫–∏
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14159 * self.radius ** 2
    
    def perimeter(self):
        return 2 * 3.14159 * self.radius
    
    def info(self):
        return f"–ö—Ä—É–≥ —Ä–∞–¥–∏—É—Å–æ–º {self.radius}. " + super().info()

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)
    
    def info(self):
        return f"–ü—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ {self.width}x{self.height}. " + super().info()

class Triangle(Shape):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    
    def area(self):
        # –§–æ—Ä–º—É–ª–∞ –ì–µ—Ä–æ–Ω–∞
        p = self.perimeter() / 2
        return (p * (p - self.a) * (p - self.b) * (p - self.c)) ** 0.5
    
    def perimeter(self):
        return self.a + self.b + self.c
    
    def info(self):
        return f"–¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ —Å–æ —Å—Ç–æ—Ä–æ–Ω–∞–º–∏ {self.a}, {self.b}, {self.c}. " + super().info()

# –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è –ø–æ–ª–∏–º–æ—Ä—Ñ–∏–∑–º–∞
print("–°–æ–∑–¥–∞—ë–º —Ñ–∏–≥—É—Ä—ã:")
shapes = [
    Circle(5),
    Rectangle(4, 6),
    Triangle(3, 4, 5),
    Circle(2.5),
    Rectangle(10, 3)
]

print("\nüìä –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –≤—Å–µ—Ö —Ñ–∏–≥—É—Ä–∞—Ö:")
for shape in shapes:
    print(f"  {shape.info()}")

# –ü–æ–ª–∏–º–æ—Ä—Ñ–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è
print("\nüßÆ –ü–æ–ª–∏–º–æ—Ä—Ñ–Ω—ã–µ –æ–ø–µ—Ä–∞—Ü–∏–∏:")

def print_total_area(shapes_list):
    """–í—ã—á–∏—Å–ª–∏—Ç—å –æ–±—â—É—é –ø–ª–æ—â–∞–¥—å –≤—Å–µ—Ö —Ñ–∏–≥—É—Ä."""
    total = 0
    for shape in shapes_list:
        total += shape.area()  # –í—ã–∑—ã–≤–∞–µ–º area() –¥–ª—è –∫–∞–∂–¥–æ–π —Ñ–∏–≥—É—Ä—ã
    print(f"–û–±—â–∞—è –ø–ª–æ—â–∞–¥—å –≤—Å–µ—Ö —Ñ–∏–≥—É—Ä: {total:.2f}")

def print_total_perimeter(shapes_list):
    """–í—ã—á–∏—Å–ª–∏—Ç—å –æ–±—â–∏–π –ø–µ—Ä–∏–º–µ—Ç—Ä –≤—Å–µ—Ö —Ñ–∏–≥—É—Ä."""
    total = 0
    for shape in shapes_list:
        total += shape.perimeter()  # –í—ã–∑—ã–≤–∞–µ–º perimeter() –¥–ª—è –∫–∞–∂–¥–æ–π —Ñ–∏–≥—É—Ä—ã
    print(f"–û–±—â–∏–π –ø–µ—Ä–∏–º–µ—Ç—Ä –≤—Å–µ—Ö —Ñ–∏–≥—É—Ä: {total:.2f}")

def find_largest_shape(shapes_list):
    """–ù–∞–π—Ç–∏ —Ñ–∏–≥—É—Ä—É —Å –Ω–∞–∏–±–æ–ª—å—à–µ–π –ø–ª–æ—â–∞–¥—å—é."""
    if not shapes_list:
        return None
    
    largest = shapes_list[0]
    for shape in shapes_list[1:]:
        if shape.area() > largest.area():
            largest = shape
    return largest

# –ò—Å–ø–æ–ª—å–∑—É–µ–º –ø–æ–ª–∏–º–æ—Ä—Ñ–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏
print_total_area(shapes)
print_total_perimeter(shapes)

largest = find_largest_shape(shapes)
if largest:
    print(f"–°–∞–º–∞—è –±–æ–ª—å—à–∞—è —Ñ–∏–≥—É—Ä–∞: {largest.info()}")

# –ü–æ–ª–∏–º–æ—Ä—Ñ–∏–∑–º –≤ –∫–æ–ª–ª–µ–∫—Ü–∏—è—Ö
print("\nüóÇÔ∏è –ü–æ–ª–∏–º–æ—Ä—Ñ–∏–∑–º —Å –∫–æ–ª–ª–µ–∫—Ü–∏—è–º–∏:")

# –°–ø–∏—Å–æ–∫ —Ä–∞–∑–Ω—ã—Ö —Ñ–∏–≥—É—Ä
mixed_shapes = [
    {"type": Circle(3), "name": "–ö—Ä—É–≥–ª—ã–π —Å—Ç–æ–ª"},
    {"type": Rectangle(2, 4), "name": "–ü—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∞—è –¥–æ—Å–∫–∞"},
    {"type": Triangle(5, 5, 5), "name": "–¢—Ä–µ—É–≥–æ–ª—å–Ω–∞—è –∫—Ä—ã—à–∞"}
]

print("–ü–ª–æ—â–∞–¥–∏ —Ä–∞–∑–Ω—ã—Ö –æ–±—ä–µ–∫—Ç–æ–≤:")
for item in mixed_shapes:
    shape = item["type"]
    name = item["name"]
    print(f"  {name}: –ø–ª–æ—â–∞–¥—å = {shape.area():.2f}")

# –ü–æ–ª–∏–º–æ—Ä—Ñ–∏–∑–º —Å —É—Å–ª–æ–≤–∏—è–º–∏
print("\nüîç –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è —Ñ–∏–≥—É—Ä:")

def filter_large_shapes(shapes_list, min_area):
    """–û—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞—Ç—å —Ñ–∏–≥—É—Ä—ã —Å –ø–ª–æ—â–∞–¥—å—é –±–æ–ª—å—à–µ –∑–∞–¥–∞–Ω–Ω–æ–π."""
    result = []
    for shape in shapes_list:
        if shape.area() > min_area:
            result.append(shape)
    return result

large_shapes = filter_large_shapes(shapes, 20)
print(f"–§–∏–≥—É—Ä—ã —Å –ø–ª–æ—â–∞–¥—å—é > 20: {len(large_shapes)} —à—Ç.")
for shape in large_shapes:
    print(f"  {shape.info()}")

print("\nüéØ –°—É—Ç—å –ø–æ–ª–∏–º–æ—Ä—Ñ–∏–∑–º–∞:")
print("–ú—ã –º–æ–∂–µ–º —Ä–∞–±–æ—Ç–∞—Ç—å —Å Circle, Rectangle –∏ Triangle –æ–¥–∏–Ω–∞–∫–æ–≤–æ,")
print("–ø–æ—Ç–æ–º—É —á—Ç–æ —É –≤—Å–µ—Ö –µ—Å—Ç—å –º–µ—Ç–æ–¥—ã area() –∏ perimeter().")
print("–ö–æ–¥, –∫–æ—Ç–æ—Ä—ã–π –∏—Å–ø–æ–ª—å–∑—É–µ—Ç —ç—Ç–∏ –º–µ—Ç–æ–¥—ã, –Ω–µ –∑–Ω–∞–µ—Ç –∏ –Ω–µ –¥–æ–ª–∂–µ–Ω –∑–Ω–∞—Ç—å,")
print("—Å –∫–∞–∫–∏–º –∏–º–µ–Ω–Ω–æ —Ç–∏–ø–æ–º —Ñ–∏–≥—É—Ä—ã –æ–Ω —Ä–∞–±–æ—Ç–∞–µ—Ç!")
````

````python


## –Ø—á–µ–π–∫–∞ 11: –ü—Ä–∏–º–µ—Ä —É—Ç–∏–Ω–æ–π —Ç–∏–ø–∏–∑–∞—Ü–∏–∏ (Code)


# –£—Ç–∏–Ω–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è –≤ –¥–µ–π—Å—Ç–≤–∏–∏
print("ü¶Ü –£—Ç–∏–Ω–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è –≤ –¥–µ–π—Å—Ç–≤–∏–∏:\n")

# –ü—Ä–∏–º–µ—Ä 1: –ü—Ä–æ—Å—Ç–∞—è —É—Ç–∏–Ω–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è
print("1. –ü—Ä–æ—Å—Ç–∞—è —É—Ç–∏–Ω–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è:\n")

class Dog:
    def speak(self):
        return "–ì–∞–≤!"
    
    def walk(self):
        return "–°–æ–±–∞–∫–∞ –∏–¥—ë—Ç –Ω–∞ —á–µ—Ç—ã—Ä—ë—Ö –ª–∞–ø–∞—Ö"

class Cat:
    def speak(self):
        return "–ú—è—É!"
    
    def walk(self):
        return "–ö–æ—Ç –∫—Ä–∞–¥—ë—Ç—Å—è –±–µ—Å—à—É–º–Ω–æ"

class Robot:
    def speak(self):
        return "–ë–∏–ø-–±—É–ø!"
    
    def walk(self):
        return "–†–æ–±–æ—Ç –¥–≤–∏–∂–µ—Ç—Å—è –Ω–∞ –∫–æ–ª—ë—Å–∞—Ö"

class Car:
    # –£ –º–∞—à–∏–Ω—ã –Ω–µ—Ç –º–µ—Ç–æ–¥–∞ speak()!
    def drive(self):
        return "–ú–∞—à–∏–Ω–∞ –µ–¥–µ—Ç"

# –ü–æ–ª–∏–º–æ—Ä—Ñ–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è (—É—Ç–∏–Ω–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è)
def make_it_talk(thing):
    """–ó–∞—Å—Ç–∞–≤–∏—Ç—å —á—Ç–æ-—Ç–æ –≥–æ–≤–æ—Ä–∏—Ç—å."""
    try:
        print(thing.speak())
    except AttributeError:
        print(f"–≠—Ç–æ {type(thing).__name__} –Ω–µ —É–º–µ–µ—Ç –≥–æ–≤–æ—Ä–∏—Ç—å!")

def make_it_move(thing):
    """–ó–∞—Å—Ç–∞–≤–∏—Ç—å —á—Ç–æ-—Ç–æ –¥–≤–∏–≥–∞—Ç—å—Å—è."""
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞–ª–∏—á–∏–µ –º–µ—Ç–æ–¥–∞ walk –∏–ª–∏ drive
    if hasattr(thing, 'walk'):
        print(thing.walk())
    elif hasattr(thing, 'drive'):
        print(thing.drive())
    else:
        print(f"–≠—Ç–æ {type(thing).__name__} –Ω–µ —É–º–µ–µ—Ç –¥–≤–∏–≥–∞—Ç—å—Å—è!")

print("–ó–∞—Å—Ç–∞–≤–ª—è–µ–º —Ä–∞–∑–Ω—ã—Ö —Å—É—â–µ—Å—Ç–≤ –∏ –æ–±—ä–µ–∫—Ç—ã –≥–æ–≤–æ—Ä–∏—Ç—å:")
make_it_talk(Dog())
make_it_talk(Cat())
make_it_talk(Robot())
make_it_talk(Car())  # –û—à–∏–±–∫–∞! –£ Car –Ω–µ—Ç –º–µ—Ç–æ–¥–∞ speak()

print("\n–ó–∞—Å—Ç–∞–≤–ª—è–µ–º –¥–≤–∏–≥–∞—Ç—å—Å—è:")
make_it_move(Dog())
make_it_move(Cat())
make_it_move(Robot())
make_it_move(Car())

# –ü—Ä–∏–º–µ—Ä 2: –ë–æ–ª–µ–µ —Å–ª–æ–∂–Ω—ã–π –ø—Ä–∏–º–µ—Ä
print("\n2. –£—Ç–∏–Ω–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è —Å —Ñ–∞–π–ª–æ–ø–æ–¥–æ–±–Ω—ã–º–∏ –æ–±—ä–µ–∫—Ç–∞–º–∏:\n")

class StringWriter:
    """–ö–ª–∞—Å—Å, –∫–æ—Ç–æ—Ä—ã–π –≤–µ–¥—ë—Ç —Å–µ–±—è –∫–∞–∫ —Ñ–∞–π–ª (–∏–º–µ–µ—Ç write())."""
    
    def __init__(self):
        self.content = []
    
    def write(self, text):
        """–ó–∞–ø–∏—Å–∞—Ç—å —Ç–µ–∫—Å—Ç."""
        self.content.append(text)
    
    def get_content(self):
        """–ü–æ–ª—É—á–∏—Ç—å –∑–∞–ø–∏—Å–∞–Ω–Ω—ã–π —Ç–µ–∫—Å—Ç."""
        return "".join(self.content)

class FileWriter:
    """–ö–ª–∞—Å—Å –¥–ª—è –∑–∞–ø–∏—Å–∏ –≤ —Ñ–∞–π–ª."""
    
    def __init__(self, filename):
        self.filename = filename
    
    def write(self, text):
        """–ó–∞–ø–∏—Å–∞—Ç—å —Ç–µ–∫—Å—Ç –≤ —Ñ–∞–π–ª."""
        with open(self.filename, 'a', encoding='utf-8') as f:
            f.write(text)
    
    def read(self):
        """–ü—Ä–æ—á–∏—Ç–∞—Ç—å —Ñ–∞–π–ª."""
        with open(self.filename, 'r', encoding='utf-8') as f:
            return f.read()

class Logger:
    """–ü—Ä–æ—Å—Ç–æ–π –ª–æ–≥–≥–µ—Ä."""
    
    def __init__(self, output):
        # output –¥–æ–ª–∂–µ–Ω –∏–º–µ—Ç—å –º–µ—Ç–æ–¥ write()
        self.output = output
    
    def log(self, message):
        """–ó–∞–ø–∏—Å–∞—Ç—å —Å–æ–æ–±—â–µ–Ω–∏–µ –≤ –ª–æ–≥."""
        timestamp = "2024-01-25 10:00:00"  # –ó–¥–µ—Å—å –±—ã–ª–∞ –±—ã —Ä–µ–∞–ª—å–Ω–∞—è –¥–∞—Ç–∞
        self.output.write(f"[{timestamp}] {message}\n")

# –ò—Å–ø–æ–ª—å–∑—É–µ–º —É—Ç–∏–Ω—É—é —Ç–∏–ø–∏–∑–∞—Ü–∏—é
print("–ò—Å–ø–æ–ª—å–∑—É–µ–º –ª–æ–≥–≥–µ—Ä —Å —Ä–∞–∑–Ω—ã–º–∏ –≤—ã–≤–æ–¥–∞–º–∏:")

# –õ–æ–≥–≥–∏—Ä—É–µ–º –≤ —Å—Ç—Ä–æ–∫—É
string_writer = StringWriter()
logger1 = Logger(string_writer)
logger1.log("–ü–µ—Ä–≤–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ")
logger1.log("–í—Ç–æ—Ä–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ")
print("–õ–æ–≥ –≤ —Å—Ç—Ä–æ–∫–µ:")
print(string_writer.get_content())

# –õ–æ–≥–≥–∏—Ä—É–µ–º –≤ —Ñ–∞–π–ª (–∏–º–∏—Ç–∏—Ä—É–µ–º)
print("\n–õ–æ–≥ –≤ —Ñ–∞–π–ª–µ (–∏–º–∏—Ç–∞—Ü–∏—è):")
file_writer = FileWriter("log.txt")  # –í —Ä–µ–∞–ª—å–Ω–æ—Å—Ç–∏ —Å–æ–∑–¥–∞–ª—Å—è –±—ã —Ñ–∞–π–ª
logger2 = Logger(file_writer)
logger2.log("–°–æ–æ–±—â–µ–Ω–∏–µ –≤ —Ñ–∞–π–ª")
print("–°–æ–æ–±—â–µ–Ω–∏–µ –∑–∞–ø–∏—Å–∞–Ω–æ –≤ —Ñ–∞–π–ª 'log.txt'")

# –ü—Ä–∏–º–µ—Ä 3: –£—Ç–∏–Ω–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è —Å –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞–º–∏
print("\n3. –£—Ç–∏–Ω–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è –∏ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã:\n")

class Vector:
    """–ü—Ä–æ—Å—Ç–æ–π –∫–ª–∞—Å—Å –≤–µ–∫—Ç–æ—Ä–∞."""
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        """–°–ª–æ–∂–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–æ–≤."""
        # –ù–∞–º –Ω–µ–≤–∞–∂–Ω–æ, —á—Ç–æ —Ç–∞–∫–æ–µ other
        # –í–∞–∂–Ω–æ, —á—Ç–æ–±—ã —É –Ω–µ–≥–æ –±—ã–ª–∏ –∞—Ç—Ä–∏–±—É—Ç—ã x –∏ y
        return Vector(self.x + other.x, self.y + other.y)
    
    def __str__(self):
        return f"Vector({self.x}, {self.y})"

class Point:
    """–ö–ª–∞—Å—Å —Ç–æ—á–∫–∏."""
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f"Point({self.x}, {self.y})"

# –í–µ–∫—Ç–æ—Ä + –í–µ–∫—Ç–æ—Ä
v1 = Vector(1, 2)
v2 = Vector(3, 4)
result1 = v1 + v2  # Vector.__add__ –≤—ã–∑—ã–≤–∞–µ—Ç—Å—è —Å –¥—Ä—É–≥–∏–º Vector
print(f"Vector + Vector: {v1} + {v2} = {result1}")

# –í–µ–∫—Ç–æ—Ä + Point (—É—Ç–∏–Ω–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è!)
p = Point(5, 6)
result2 = v1 + p  # Vector.__add__ –≤—ã–∑—ã–≤–∞–µ—Ç—Å—è —Å Point
print(f"Vector + Point: {v1} + {p} = {result2}")

print("\nüîç –í–∞–∂–Ω–æ –¥–ª—è —É—Ç–∏–Ω–æ–π —Ç–∏–ø–∏–∑–∞—Ü–∏–∏:")
print("Python –Ω–µ –ø—Ä–æ–≤–µ—Ä—è–µ—Ç, —á—Ç–æ p - —ç—Ç–æ Vector.")
print("Python –ø—Ä–æ—Å—Ç–æ –ø—ã—Ç–∞–µ—Ç—Å—è –≤—ã–ø–æ–ª–Ω–∏—Ç—å v1.__add__(p).")
print("–¢–∞–∫ –∫–∞–∫ —É p –µ—Å—Ç—å –∞—Ç—Ä–∏–±—É—Ç—ã x –∏ y, –≤—Å—ë —Ä–∞–±–æ—Ç–∞–µ—Ç!")
````

````python
# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–∞–∑–Ω—ã—Ö –ø–æ–¥—Ö–æ–¥–æ–≤
print("üìä –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–∞–∑–Ω—ã—Ö –ø–æ–¥—Ö–æ–¥–æ–≤:\n")

# –ü–æ–¥—Ö–æ–¥ 1: –ñ—ë—Å—Ç–∫–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è (–∫–∞–∫ –≤ C++/Java)
print("1. –ñ—ë—Å—Ç–∫–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è (–Ω–µ –ø–æ-–ø–∏—Ç–æ–Ω—å–∏):")

def calculate_area_1(shape):
    """–í—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø–ª–æ—â–∞–¥–∏ —Å –ø—Ä–æ–≤–µ—Ä–∫–æ–π —Ç–∏–ø–∞."""
    if isinstance(shape, Circle):
        return 3.14 * shape.radius ** 2
    elif isinstance(shape, Rectangle):
        return shape.width * shape.height
    elif isinstance(shape, Triangle):
        # –§–æ—Ä–º—É–ª–∞ –ì–µ—Ä–æ–Ω–∞
        p = (shape.a + shape.b + shape.c) / 2
        return (p * (p - shape.a) * (p - shape.b) * (p - shape.c)) ** 0.5
    else:
        raise TypeError(f"–ù–µ–∏–∑–≤–µ—Å—Ç–Ω—ã–π —Ç–∏–ø —Ñ–∏–≥—É—Ä—ã: {type(shape)}")

# –ü–æ–¥—Ö–æ–¥ 2: –£—Ç–∏–Ω–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è (–ø–æ-–ø–∏—Ç–æ–Ω—å–∏)
print("\n2. –£—Ç–∏–Ω–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è (–ø–æ-–ø–∏—Ç–æ–Ω—å–∏):")

def calculate_area_2(shape):
    """–í—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø–ª–æ—â–∞–¥–∏ —Å —É—Ç–∏–Ω–æ–π —Ç–∏–ø–∏–∑–∞—Ü–∏–µ–π."""
    # –ü—Ä–æ—Å—Ç–æ –≤—ã–∑—ã–≤–∞–µ–º –º–µ—Ç–æ–¥ area()
    # –ï—Å–ª–∏ —É shape –µ—Å—Ç—å —ç—Ç–æ—Ç –º–µ—Ç–æ–¥ - –≤—Å—ë —Ä–∞–±–æ—Ç–∞–µ—Ç
    return shape.area()

# –°–æ–∑–¥–∞–¥–∏–º –ø—Ä–æ—Å—Ç—ã–µ –∫–ª–∞—Å—Å—ã –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏
class SimpleCircle:
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius ** 2

class SimpleRectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height

class NewShape:
    """–ù–æ–≤–∞—è —Ñ–∏–≥—É—Ä–∞, –∫–æ—Ç–æ—Ä—É—é –º—ã –¥–æ–±–∞–≤–∏–ª–∏ –ø–æ–∑–∂–µ."""
    def __init__(self, side):
        self.side = side
    
    def area(self):
        return self.side ** 2  # –ü–ª–æ—â–∞–¥—å –∫–≤–∞–¥—Ä–∞—Ç–∞

# –¢–µ—Å—Ç–∏—Ä—É–µ–º –æ–±–∞ –ø–æ–¥—Ö–æ–¥–∞
print("\nüß™ –¢–µ—Å—Ç–∏—Ä—É–µ–º –æ–±–∞ –ø–æ–¥—Ö–æ–¥–∞:")

shapes_for_test = [
    SimpleCircle(5),
    SimpleRectangle(4, 6),
    NewShape(3)  # –ù–æ–≤–∞—è —Ñ–∏–≥—É—Ä–∞, –æ –∫–æ—Ç–æ—Ä–æ–π –ø–µ—Ä–≤–∞—è —Ñ—É–Ω–∫—Ü–∏—è –Ω–µ –∑–Ω–∞–µ—Ç
]

print("\n–° –∂—ë—Å—Ç–∫–æ–π —Ç–∏–ø–∏–∑–∞—Ü–∏–µ–π (–ø–æ–¥—Ö–æ–¥ 1):")
for shape in shapes_for_test[:2]:  # –¢–æ–ª—å–∫–æ –ø–µ—Ä–≤—ã–µ –¥–≤–µ, —Ç—Ä–µ—Ç—å—è –≤—ã–∑–æ–≤–µ—Ç –æ—à–∏–±–∫—É
    try:
        area = calculate_area_1(shape)
        print(f"  –ü–ª–æ—â–∞–¥—å: {area:.2f}")
    except TypeError as e:
        print(f"  –û—à–∏–±–∫–∞: {e}")

# NewShape –≤—ã–∑–æ–≤–µ—Ç –æ—à–∏–±–∫—É –≤ –ø–µ—Ä–≤–æ–º –ø–æ–¥—Ö–æ–¥–µ
try:
    calculate_area_1(shapes_for_test[2])
except TypeError as e:
    print(f"  –î–ª—è NewShape: {e}")

print("\n–° —É—Ç–∏–Ω–æ–π —Ç–∏–ø–∏–∑–∞—Ü–∏–µ–π (–ø–æ–¥—Ö–æ–¥ 2):")
for shape in shapes_for_test:
    area = calculate_area_2(shape)  # –†–∞–±–æ—Ç–∞–µ—Ç —Å–æ –≤—Å–µ–º–∏!
    print(f"  –ü–ª–æ—â–∞–¥—å: {area:.2f}")

print("\nüéØ –í—ã–≤–æ–¥—ã:")
print("1. –ñ—ë—Å—Ç–∫–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è:")
print("   - –ù—É–∂–Ω–æ –∑–Ω–∞—Ç—å –≤—Å–µ —Ç–∏–ø—ã –∑–∞—Ä–∞–Ω–µ–µ")
print("   - –ü—Ä–∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏–∏ –Ω–æ–≤–æ–≥–æ —Ç–∏–ø–∞ –Ω—É–∂–Ω–æ –º–µ–Ω—è—Ç—å —Ñ—É–Ω–∫—Ü–∏—é")
print("   - –ú–Ω–æ–≥–æ –ø—Ä–æ–≤–µ—Ä–æ–∫ isinstance()")

print("\n2. –£—Ç–∏–Ω–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è:")
print("   - –ù—É–∂–Ω–æ —Ç–æ–ª—å–∫–æ —á—Ç–æ–±—ã –±—ã–ª –º–µ—Ç–æ–¥ area()")
print("   - –ù–æ–≤—ã–µ —Ç–∏–ø—ã –¥–æ–±–∞–≤–ª—è—é—Ç—Å—è –±–µ–∑ –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏")
print("   - –ú–µ–Ω—å—à–µ –∫–æ–¥–∞, –±–æ–ª—å—à–µ –≥–∏–±–∫–æ—Å—Ç–∏")

print("\nüí° –ü—Ä–∞–≤–∏–ª–æ Python:")
print("¬´–õ—É—á—à–µ –ø—Ä–æ—Å–∏—Ç—å –ø—Ä–æ—â–µ–Ω–∏—è, —á–µ–º —Ä–∞–∑—Ä–µ—à–µ–Ω–∏—è¬ª (EAFP)")
print("  - EAFP: Easier to Ask for Forgiveness than Permission")
print("  - –ü—ã—Ç–∞–µ–º—Å—è —Å–¥–µ–ª–∞—Ç—å, –∏ –µ—Å–ª–∏ –Ω–µ –ø–æ–ª—É—á–∞–µ—Ç—Å—è - –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –æ—à–∏–±–∫—É")
print("\n¬´–°–º–æ—Ç—Ä–∏ –ø—Ä–µ–∂–¥–µ, —á–µ–º –ø—Ä—ã–≥–∞—Ç—å¬ª (LBYL)")
print("  - LBYL: Look Before You Leap")
print("  - –°–Ω–∞—á–∞–ª–∞ –ø—Ä–æ–≤–µ—Ä—è–µ–º, –ø–æ—Ç–æ–º –¥–µ–ª–∞–µ–º")

print("\n–ü—Ä–∏–º–µ—Ä EAFP (Python-way):")
print("""
try:
    value = obj.method()
except AttributeError:
    # –û–±—Ä–∞–±–æ—Ç–∫–∞ —Å–ª—É—á–∞—è, –∫–æ–≥–¥–∞ –º–µ—Ç–æ–¥–∞ –Ω–µ—Ç
    value = default_value
""")

print("\n–ü—Ä–∏–º–µ—Ä LBYL (–Ω–µ —Å–æ–≤—Å–µ–º Python-way):")
print("""
if hasattr(obj, 'method'):
    value = obj.method()
else:
    value = default_value
""")
````
