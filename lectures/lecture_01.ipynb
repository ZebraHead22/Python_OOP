## üí¨ –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏ –∏ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ

### –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏
- –û–¥–Ω–æ—Å—Ç—Ä–æ—á–Ω—ã–µ: `# –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π`
- –ú–Ω–æ–≥–æ—Å—Ç—Ä–æ—á–Ω—ã–µ: `''' –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π '''` –∏–ª–∏ `""" –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π """`

### –ü–µ—Ä–µ–º–µ–Ω–Ω—ã–µ
–í Python –Ω–µ –Ω—É–∂–Ω–æ –æ–±—ä—è–≤–ª—è—Ç—å —Ç–∏–ø –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π! –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ç–æ—Ä —Å–∞–º –æ–ø—Ä–µ–¥–µ–ª–∏—Ç —Ç–∏–ø.

```python
# –û–±—ä—è–≤–ª–µ–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö
x = 10           # int
y = 3.14         # float
name = "Python"  # str
is_valid = True  # bool

# –í—ã–≤–æ–¥ —Ç–∏–ø–∞ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
print(f"–¢–∏–ø x: {type(x)}")        # <class 'int'>
print(f"–¢–∏–ø y: {type(y)}")        # <class 'float'>
print(f"–¢–∏–ø name: {type(name)}")  # <class 'str'>
print(f"–¢–∏–ø is_valid: {type(is_valid)}")  # <class 'bool'>

# –î–∏–Ω–∞–º–∏—á–µ—Å–∫–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è - –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è –º–æ–∂–µ—Ç –º–µ–Ω—è—Ç—å —Ç–∏–ø!
variable = 42
print(f"variable = {variable}, —Ç–∏–ø: {type(variable)}")

variable = "–¢–µ–ø–µ—Ä—å —è —Å—Ç—Ä–æ–∫–∞"
print(f"variable = {variable}, —Ç–∏–ø: {type(variable)}")

variable = 3.14
print(f"variable = {variable}, —Ç–∏–ø: {type(variable)}")
```

## üìä –û—Å–Ω–æ–≤–Ω—ã–µ —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö –≤ Python

### –ß–∏—Å–ª–æ–≤—ã–µ —Ç–∏–ø—ã
1. **int** - —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ (–Ω–µ–æ–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω–æ–π —Ç–æ—á–Ω–æ—Å—Ç–∏!)
2. **float** - —á–∏—Å–ª–∞ —Å –ø–ª–∞–≤–∞—é—â–µ–π —Ç–æ—á–∫–æ–π
3. **complex** - –∫–æ–º–ø–ª–µ–∫—Å–Ω—ã–µ —á–∏—Å–ª–∞

### –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏
1. **str** - —Å—Ç—Ä–æ–∫–∏ (—é–Ω–∏–∫–æ–¥ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é)
2. **list** - –∏–∑–º–µ–Ω—è–µ–º—ã–π –º–∞—Å—Å–∏–≤
3. **tuple** - –Ω–µ–∏–∑–º–µ–Ω—è–µ–º—ã–π –º–∞—Å—Å–∏–≤
4. **range** - –¥–∏–∞–ø–∞–∑–æ–Ω —á–∏—Å–µ–ª

### –ö–æ–ª–ª–µ–∫—Ü–∏–∏
1. **dict** - —Å–ª–æ–≤–∞—Ä—å (—Ö–µ—à-—Ç–∞–±–ª–∏—Ü–∞)
2. **set** - –º–Ω–æ–∂–µ—Å—Ç–≤–æ (—É–Ω–∏–∫–∞–ª—å–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã)

### –õ–æ–≥–∏—á–µ—Å–∫–∏–π —Ç–∏–ø
1. **bool** - True/False

### –°–ø–µ—Ü–∏–∞–ª—å–Ω—ã–µ
1. **NoneType** - –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è

```python
# –¶–µ–ª—ã–µ —á–∏—Å–ª–∞ (int) - –Ω–µ–æ–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å!
big_number = 10**100  # 10 –≤ —Å—Ç–µ–ø–µ–Ω–∏ 100
print(f"–û–≥—Ä–æ–º–Ω–æ–µ —á–∏—Å–ª–æ: {big_number}")
print(f"–ï—â–µ –±–æ–ª—å—à–µ: {big_number * 1000}")

# –û–ø–µ—Ä–∞—Ü–∏–∏ —Å —á–∏—Å–ª–∞–º–∏
a = 10
b = 3

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")      # –û–±—ã—á–Ω–æ–µ –¥–µ–ª–µ–Ω–∏–µ (float)
print(f"{a} // {b} = {a // b}")    # –¶–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω–æ–µ –¥–µ–ª–µ–Ω–∏–µ
print(f"{a} % {b} = {a % b}")      # –û—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è
print(f"{a} ** {b} = {a ** b}")    # –í–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ —Å—Ç–µ–ø–µ–Ω—å

# –ö–æ–º–ø–ª–µ–∫—Å–Ω—ã–µ —á–∏—Å–ª–∞
z = 3 + 4j
print(f"–ö–æ–º–ø–ª–µ–∫—Å–Ω–æ–µ —á–∏—Å–ª–æ: {z}")
print(f"–î–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–∞—è —á–∞—Å—Ç—å: {z.real}")
print(f"–ú–Ω–∏–º–∞—è —á–∞—Å—Ç—å: {z.imag}")
```

```python
# –°—Ç—Ä–æ–∫–∏ –≤ Python
str1 = "–ü—Ä–æ—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞"
str2 = '–¢–æ–∂–µ —Å—Ç—Ä–æ–∫–∞'
str3 = '''–ú–Ω–æ–≥–æ—Å—Ç—Ä–æ—á–Ω–∞—è
—Å—Ç—Ä–æ–∫–∞'''
str4 = """–ï—â–µ –æ–¥–Ω–∞
–º–Ω–æ–≥–æ—Å—Ç—Ä–æ—á–Ω–∞—è"""

print(str1)
print(str2)
print(str3)
print(str4)

# –û–ø–µ—Ä–∞—Ü–∏–∏ —Å–æ —Å—Ç—Ä–æ–∫–∞–º–∏
s = "Python"
print(f"–î–ª–∏–Ω–∞ —Å—Ç—Ä–æ–∫–∏: {len(s)}")
print(f"–ü–µ—Ä–≤—ã–π —Å–∏–º–≤–æ–ª: {s[0]}")    # P
print(f"–ü–æ—Å–ª–µ–¥–Ω–∏–π —Å–∏–º–≤–æ–ª: {s[-1]}") # n (–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∏–Ω–¥–µ–∫—Å—ã!)
print(f"–°—Ä–µ–∑: {s[1:4]}")           # yth
print(f"–°—Ä–µ–∑ —Å —à–∞–≥–æ–º: {s[::2]}")   # Pto

# –ö–æ–Ω–∫–∞—Ç–µ–Ω–∞—Ü–∏—è –∏ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–µ
print("Hello" + " " + "World")
print("Python" * 3)

# f-—Å—Ç—Ä–æ–∫–∏ (—Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ —Å—Ç—Ä–æ–∫)
name = "–ê–Ω–Ω–∞"
age = 20
height = 1.75

message = f"–ò–º—è: {name}, –≤–æ–∑—Ä–∞—Å—Ç: {age}, —Ä–æ—Å—Ç: {height:.2f} –º"
print(message)

# –ú–µ—Ç–æ–¥—ã —Å—Ç—Ä–æ–∫
text = "  Python - –º–æ—â–Ω—ã–π —è–∑—ã–∫!  "
print(f"–ò—Å—Ö–æ–¥–Ω—ã–π —Ç–µ–∫—Å—Ç: '{text}'")
print(f"–£–±—Ä–∞—Ç—å –ø—Ä–æ–±–µ–ª—ã: '{text.strip()}'")
print(f"–í–µ—Ä—Ö–Ω–∏–π —Ä–µ–≥–∏—Å—Ç—Ä: '{text.upper()}'")
print(f"–ù–∏–∂–Ω–∏–π —Ä–µ–≥–∏—Å—Ç—Ä: '{text.lower()}'")
print(f"–ó–∞–º–µ–Ω–∞: '{text.replace('–º–æ—â–Ω—ã–π', '–∫—Ä—É—Ç–æ–π')}'")
print(f"–ù–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å 'Python': {text.strip().startswith('Python')}")
print(f"–ó–∞–∫–∞–Ω—á–∏–≤–∞–µ—Ç—Å—è –Ω–∞ '—è–∑—ã–∫!': {text.strip().endswith('—è–∑—ã–∫!')}")
```

## üìã –°–ø–∏—Å–∫–∏ (list) - –∏–∑–º–µ–Ω—è–µ–º—ã–µ –º–∞—Å—Å–∏–≤—ã

–°–ø–∏—Å–æ–∫ - —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω–∞—è –∏–∑–º–µ–Ω—è–µ–º–∞—è –∫–æ–ª–ª–µ–∫—Ü–∏—è. –ê–Ω–∞–ª–æ–≥ `std::vector` –≤ C++.

```python
# –°–æ–∑–¥–∞–Ω–∏–µ —Å–ø–∏—Å–∫–æ–≤
empty_list = []
numbers = [1, 2, 3, 4, 5]
mixed = [1, "–¥–≤–∞", 3.0, True]
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

print(f"numbers: {numbers}")
print(f"mixed: {mixed}")
print(f"matrix: {matrix}")

# –û–ø–µ—Ä–∞—Ü–∏–∏ —Å–æ —Å–ø–∏—Å–∫–∞–º–∏
print(f"–î–ª–∏–Ω–∞ —Å–ø–∏—Å–∫–∞: {len(numbers)}")
print(f"–ü–µ—Ä–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç: {numbers[0]}")
print(f"–ü–æ—Å–ª–µ–¥–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç: {numbers[-1]}")
print(f"–°—Ä–µ–∑: {numbers[1:4]}")
print(f"–°—Ä–µ–∑ —Å —à–∞–≥–æ–º: {numbers[::2]}")

# –ò–∑–º–µ–Ω–µ–Ω–∏–µ —Å–ø–∏—Å–∫–æ–≤
numbers.append(6)  # –î–æ–±–∞–≤–∏—Ç—å –≤ –∫–æ–Ω–µ—Ü
print(f"–ü–æ—Å–ª–µ append: {numbers}")

numbers.insert(2, 99)  # –í—Å—Ç–∞–≤–∏—Ç—å –Ω–∞ –ø–æ–∑–∏—Ü–∏—é 2
print(f"–ü–æ—Å–ª–µ insert: {numbers}")

numbers.extend([7, 8, 9])  # –†–∞—Å—à–∏—Ä–∏—Ç—å –¥—Ä—É–≥–∏–º —Å–ø–∏—Å–∫–æ–º
print(f"–ü–æ—Å–ª–µ extend: {numbers}")

# –£–¥–∞–ª–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–æ–≤
removed = numbers.pop()  # –£–¥–∞–ª–∏—Ç—å –ø–æ—Å–ª–µ–¥–Ω–∏–π
print(f"–£–¥–∞–ª–µ–Ω: {removed}, —Å–ø–∏—Å–æ–∫: {numbers}")

removed = numbers.pop(2)  # –£–¥–∞–ª–∏—Ç—å –ø–æ –∏–Ω–¥–µ–∫—Å—É
print(f"–£–¥–∞–ª–µ–Ω —ç–ª–µ–º–µ–Ω—Ç —Å –∏–Ω–¥–µ–∫—Å–æ–º 2: {removed}, —Å–ø–∏—Å–æ–∫: {numbers}")

numbers.remove(4)  # –£–¥–∞–ª–∏—Ç—å –ø–µ—Ä–≤–æ–µ –≤—Ö–æ–∂–¥–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è
print(f"–ü–æ—Å–ª–µ remove(4): {numbers}")

# –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞
unsorted = [5, 2, 8, 1, 9]
unsorted.sort()
print(f"–ü–æ—Å–ª–µ sort: {unsorted}")

unsorted.sort(reverse=True)
print(f"–û–±—Ä–∞—Ç–Ω–∞—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞: {unsorted}")

# –ì–µ–Ω–µ—Ä–∞—Ç–æ—Ä—ã —Å–ø–∏—Å–∫–æ–≤ (list comprehension) - –æ—á–µ–Ω—å –º–æ—â–Ω–∞—è —Ñ–∏—á–∞!
squares = [x**2 for x in range(10)]
print(f"–ö–≤–∞–¥—Ä–∞—Ç—ã: {squares}")

even_squares = [x**2 for x in range(10) if x % 2 == 0]
print(f"–ö–≤–∞–¥—Ä–∞—Ç—ã —á–µ—Ç–Ω—ã—Ö: {even_squares}")

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞
print(f"–ï—Å—Ç—å –ª–∏ 3 –≤ —Å–ø–∏—Å–∫–µ? {3 in numbers}")
print(f"–ï—Å—Ç—å –ª–∏ 100 –≤ —Å–ø–∏—Å–∫–µ? {100 in numbers}")
```

```python
# –ö–æ—Ä—Ç–µ–∂–∏ (tuple) - –Ω–µ–∏–∑–º–µ–Ω—è–µ–º—ã–µ —Å–ø–∏—Å–∫–∏
point = (10, 20)
print(f"–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã: {point}")
print(f"X: {point[0]}, Y: {point[1]}")

# –ö–æ—Ä—Ç–µ–∂ –∏–∑ –æ–¥–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ (–æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ —Å –∑–∞–ø—è—Ç–æ–π!)
single = (42,)
not_a_tuple = (42)  # –≠—Ç–æ –ø—Ä–æ—Å—Ç–æ —á–∏—Å–ª–æ!
print(f"–ö–æ—Ä—Ç–µ–∂ –∏–∑ –æ–¥–Ω–æ–≥–æ: {type(single)}")
print(f"–ë–µ–∑ –∑–∞–ø—è—Ç–æ–π: {type(not_a_tuple)}")

# –†–∞—Å–ø–∞–∫–æ–≤–∫–∞ –∫–æ—Ä—Ç–µ–∂–∞
x, y = point
print(f"–†–∞—Å–ø–∞–∫–æ–≤–∞–Ω–æ: x={x}, y={y}")

# –ú–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–∞—è —Ä–∞—Å–ø–∞–∫–æ–≤–∫–∞
a, b, *rest = 1, 2, 3, 4, 5
print(f"a={a}, b={b}, rest={rest}")

# –û–±–º–µ–Ω –∑–Ω–∞—á–µ–Ω–∏–π –±–µ–∑ –≤—Ä–µ–º–µ–Ω–Ω–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
x, y = 10, 20
print(f"–î–æ –æ–±–º–µ–Ω–∞: x={x}, y={y}")
x, y = y, x
print(f"–ü–æ—Å–ª–µ –æ–±–º–µ–Ω–∞: x={x}, y={y}")
```

```python
# –°–ª–æ–≤–∞—Ä–∏ (dict) - —Ö–µ—à-—Ç–∞–±–ª–∏—Ü—ã / –∞—Å—Å–æ—Ü–∏–∞—Ç–∏–≤–Ω—ã–µ –º–∞—Å—Å–∏–≤—ã
# –ö–ª—é—á–æ–º –º–æ–∂–µ—Ç –±—ã—Ç—å –ª—é–±–æ–π –Ω–µ–∏–∑–º–µ–Ω—è–µ–º—ã–π –æ–±—ä–µ–∫—Ç

student = {
    "name": "–ò–≤–∞–Ω",
    "age": 20,
    "courses": ["–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞", "–§–∏–∑–∏–∫–∞", "–ü—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏–µ"],
    "grades": {"–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞": 5, "–§–∏–∑–∏–∫–∞": 4}
}

print(f"–°—Ç—É–¥–µ–Ω—Ç: {student}")
print(f"–ò–º—è: {student['name']}")
print(f"–ö—É—Ä—Å—ã: {student['courses']}")

# –î–æ—Å—Ç—É–ø —Å get (–±–µ–∑–æ–ø–∞—Å–Ω—ã–π)
print(f"–í–æ–∑—Ä–∞—Å—Ç: {student.get('age')}")
print(f"–ê–¥—Ä–µ—Å: {student.get('address', '–ù–µ —É–∫–∞–∑–∞–Ω')}")

# –ò–∑–º–µ–Ω–µ–Ω–∏–µ —Å–ª–æ–≤–∞—Ä—è
student["email"] = "ivan@example.com"  # –î–æ–±–∞–≤–∏—Ç—å
student["age"] = 21  # –ò–∑–º–µ–Ω–∏—Ç—å

# –£–¥–∞–ª–µ–Ω–∏–µ
removed = student.pop("grades")
print(f"–£–¥–∞–ª–µ–Ω—ã –æ—Ü–µ–Ω–∫–∏: {removed}")

# –ò—Ç–µ—Ä–∞—Ü–∏—è –ø–æ —Å–ª–æ–≤–∞—Ä—é
print("\n–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Å—Ç—É–¥–µ–Ω—Ç–µ:")
for key, value in student.items():
    print(f"  {key}: {value}")

# –ì–µ–Ω–µ—Ä–∞—Ç–æ—Ä—ã —Å–ª–æ–≤–∞—Ä–µ–π
squares_dict = {x: x**2 for x in range(1, 6)}
print(f"\n–ö–≤–∞–¥—Ä–∞—Ç—ã —á–∏—Å–µ–ª: {squares_dict}")
```

```python
# –ú–Ω–æ–∂–µ—Å—Ç–≤–∞ (set) - —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –Ω–µ—É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã
primes = {2, 3, 5, 7, 11, 13}
numbers = {1, 2, 3, 4, 5, 6}

print(f"–ü—Ä–æ—Å—Ç—ã–µ —á–∏—Å–ª–∞: {primes}")
print(f"–ß–∏—Å–ª–∞: {numbers}")

# –û–ø–µ—Ä–∞—Ü–∏–∏ —Å –º–Ω–æ–∂–µ—Å—Ç–≤–∞–º–∏
print(f"–ü–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ: {primes & numbers}")  # –û–±—â–∏–µ —ç–ª–µ–º–µ–Ω—Ç—ã
print(f"–û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ: {primes | numbers}")  # –í—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã
print(f"–†–∞–∑–Ω–æ—Å—Ç—å: {primes - numbers}")     # –í primes, –Ω–æ –Ω–µ –≤ numbers
print(f"–°–∏–º–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è —Ä–∞–∑–Ω–æ—Å—Ç—å: {primes ^ numbers}")  # –≠–ª–µ–º–µ–Ω—Ç—ã, –∫–æ—Ç–æ—Ä—ã–µ –µ—Å—Ç—å —Ç–æ–ª—å–∫–æ –≤ –æ–¥–Ω–æ–º

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏
print(f"7 –ø—Ä–æ—Å—Ç–æ–µ —á–∏—Å–ª–æ? {7 in primes}")
print(f"4 –ø—Ä–æ—Å—Ç–æ–µ —á–∏—Å–ª–æ? {4 in primes}")

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –∏ —É–¥–∞–ª–µ–Ω–∏–µ
primes.add(17)
print(f"–ü–æ—Å–ª–µ –¥–æ–±–∞–≤–ª–µ–Ω–∏—è 17: {primes}")

primes.discard(2)  # –£–¥–∞–ª–∏—Ç—å, –µ—Å–ª–∏ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç
print(f"–ü–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è 2: {primes}")
```

## üéÆ –£–ø—Ä–∞–≤–ª—è—é—â–∏–µ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü–∏–∏

### –û—Ç–ª–∏—á–∏—è –æ—Ç C++:
1. –ù–µ—Ç —Ñ–∏–≥—É—Ä–Ω—ã—Ö —Å–∫–æ–±–æ–∫ `{}` 
2. –ù–µ—Ç –∫—Ä—É–≥–ª—ã—Ö —Å–∫–æ–±–æ–∫ –≤ —É—Å–ª–æ–≤–∏—è—Ö (–Ω–æ –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å)
3. –û–±—è–∑–∞—Ç–µ–ª—å–Ω—ã–µ –æ—Ç—Å—Ç—É–ø—ã (4 –ø—Ä–æ–±–µ–ª–∞ –∏–ª–∏ —Ç–∞–±—É–ª—è—Ü–∏—è)
4. –î–≤–æ–µ—Ç–æ—á–∏–µ `:` –ø–æ—Å–ª–µ —É—Å–ª–æ–≤–∏—è/—Ü–∏–∫–ª–∞

### –£—Å–ª–æ–≤–Ω—ã–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã
- `if`
- `elif` (–≤–º–µ—Å—Ç–æ `else if`)
- `else`

### –¶–∏–∫–ª—ã
- `for` (–∏—Ç–µ—Ä–∞—Ü–∏—è –ø–æ –∫–æ–ª–ª–µ–∫—Ü–∏—è–º)
- `while`
- `break`, `continue`

```python
# –£—Å–ª–æ–≤–Ω—ã–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã
age = 20

# –ü—Ä–æ—Å—Ç–æ–π if
if age >= 18:
    print("–°–æ–≤–µ—Ä—à–µ–Ω–Ω–æ–ª–µ—Ç–Ω–∏–π")
    print("–ú–æ–∂–Ω–æ –≥–æ–ª–æ—Å–æ–≤–∞—Ç—å")

# if-elif-else
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"

print(f"–û—Ü–µ–Ω–∫–∞: {grade}")

# –¢–µ—Ä–Ω–∞—Ä–Ω—ã–π –æ–ø–µ—Ä–∞—Ç–æ—Ä
status = "–≤–∑—Ä–æ—Å–ª—ã–π" if age >= 18 else "—Ä–µ–±–µ–Ω–æ–∫"
print(f"–°—Ç–∞—Ç—É—Å: {status}")

# –¶–µ–ø–æ—á–∫–∏ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π (–∫–∞–∫ –≤ –º–∞—Ç–µ–º–∞—Ç–∏–∫–µ)
x = 5
if 0 < x < 10:
    print(f"{x} –º–µ–∂–¥—É 0 –∏ 10")

# –õ–æ–≥–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä—ã: and, or, not (–≤–º–µ—Å—Ç–æ &&, ||, !)
if age >= 18 and score >= 60:
    print("–ü—Ä–æ—à–µ–ª –ø–æ—Ä–æ–≥")
```

```python
# –¶–∏–∫–ª for - –∏—Ç–µ—Ä–∞—Ü–∏—è –ø–æ –∫–æ–ª–ª–µ–∫—Ü–∏—è–º
print("–¶–∏–∫–ª –ø–æ —Å–ø–∏—Å–∫—É:")
for number in [1, 2, 3, 4, 5]:
    print(f"–ß–∏—Å–ª–æ: {number}")

print("\n–¶–∏–∫–ª –ø–æ —Å—Ç—Ä–æ–∫–µ:")
for char in "Python":
    print(char)

print("\n–¶–∏–∫–ª –ø–æ —Å–ª–æ–≤–∞—Ä—é:")
student = {"name": "–ò–≤–∞–Ω", "age": 20, "course": "–ò–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫–∞"}
for key, value in student.items():
    print(f"{key}: {value}")

# –§—É–Ω–∫—Ü–∏—è range()
print("\nrange(5):")
for i in range(5):
    print(i)

print("\nrange(2, 10, 2):")
for i in range(2, 10, 2):  # –æ—Ç 2 –¥–æ 10 —Å —à–∞–≥–æ–º 2
    print(i)

# –¶–∏–∫–ª while
print("\n–¶–∏–∫–ª while:")
counter = 0
while counter < 5:
    print(f"–°—á–µ—Ç—á–∏–∫: {counter}")
    counter += 1

# break –∏ continue
print("\nbreak –∏ continue:")
for i in range(10):
    if i == 3:
        continue  # –ü—Ä–æ–ø—É—Å—Ç–∏—Ç—å 3
    if i == 7:
        break     # –í—ã–π—Ç–∏ –Ω–∞ 7
    print(i)

# else –≤ —Ü–∏–∫–ª–∞—Ö (–≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è, –µ—Å–ª–∏ –Ω–µ –±—ã–ª–æ break)
print("\n–ü–æ–∏—Å–∫ —á–∏—Å–ª–∞:")
for i in range(5):
    if i == 10:  # –¢–∞–∫–æ–≥–æ —á–∏—Å–ª–∞ –Ω–µ—Ç
        break
else:
    print("–ß–∏—Å–ª–æ –Ω–µ –Ω–∞–π–¥–µ–Ω–æ")
```

## üõ†Ô∏è –§—É–Ω–∫—Ü–∏–∏ –≤ Python

### –û—Ç–ª–∏—á–∏—è –æ—Ç C++:
1. –ù–µ—Ç —Ç–∏–ø–∞ –≤–æ–∑–≤—Ä–∞—â–∞–µ–º–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –æ–±—ä—è–≤–ª–µ–Ω–∏–∏
2. –ù–µ—Ç —Ç–∏–ø–æ–≤ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ (–Ω–æ –µ—Å—Ç—å –∞–Ω–Ω–æ—Ç–∞—Ü–∏–∏ —Ç–∏–ø–æ–≤)
3. –ú–æ–≥—É—Ç –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ –∑–Ω–∞—á–µ–Ω–∏–π (—á–µ—Ä–µ–∑ –∫–æ—Ä—Ç–µ–∂)
4. –ü–æ–¥–¥–µ—Ä–∂–∏–≤–∞—é—Ç –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é
5. –ú–æ–≥—É—Ç –ø—Ä–∏–Ω–∏–º–∞—Ç—å –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤

```python
# –ü—Ä–æ—Å—Ç–∞—è —Ñ—É–Ω–∫—Ü–∏—è
def greet():
    print("–ü—Ä–∏–≤–µ—Ç, –º–∏—Ä!")

greet()

# –§—É–Ω–∫—Ü–∏—è —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏
def greet_person(name):
    print(f"–ü—Ä–∏–≤–µ—Ç, {name}!")

greet_person("–ê–Ω–Ω–∞")
greet_person("–ò–≤–∞–Ω")

# –§—É–Ω–∫—Ü–∏—è —Å –≤–æ–∑–≤—Ä–∞—â–∞–µ–º—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º
def add(a, b):
    return a + b

result = add(5, 3)
print(f"5 + 3 = {result}")

# –§—É–Ω–∫—Ü–∏—è, –≤–æ–∑–≤—Ä–∞—â–∞—é—â–∞—è –Ω–µ—Å–∫–æ–ª—å–∫–æ –∑–Ω–∞—á–µ–Ω–∏–π
def min_max(numbers):
    return min(numbers), max(numbers)

minimum, maximum = min_max([5, 2, 8, 1, 9])
print(f"–ú–∏–Ω–∏–º—É–º: {minimum}, –ú–∞–∫—Å–∏–º—É–º: {maximum}")

# –ó–Ω–∞—á–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é
def power(base, exponent=2):
    return base ** exponent

print(f"5^2 = {power(5)}")      # exponent=2 –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é
print(f"2^3 = {power(2, 3)}")   # exponent=3

# –ò–º–µ–Ω–æ–≤–∞–Ω–Ω—ã–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã (keyword arguments)
def create_person(name, age, city="–ú–æ—Å–∫–≤–∞"):
    return f"{name}, {age} –ª–µ—Ç, –≥.{city}"

# –ü–æ—Ä—è–¥–æ–∫ –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤ –º–æ–∂–Ω–æ –º–µ–Ω—è—Ç—å, –µ—Å–ª–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∏–º–µ–Ω–∞
person1 = create_person("–ê–Ω–Ω–∞", 20)
person2 = create_person(age=25, name="–ò–≤–∞–Ω", city="–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥")
print(person1)
print(person2)

# –ü–µ—Ä–µ–º–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤ (*args, **kwargs)
def print_args(*args, **kwargs):
    print(f"–ü–æ–∑–∏—Ü–∏–æ–Ω–Ω—ã–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã (args): {args}")
    print(f"–ò–º–µ–Ω–æ–≤–∞–Ω–Ω—ã–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã (kwargs): {kwargs}")

print_args(1, 2, 3, name="Python", version=3.11)

# –ê–Ω–Ω–æ—Ç–∞—Ü–∏–∏ —Ç–∏–ø–æ–≤ (–Ω–µ –ø—Ä–æ–≤–µ—Ä—è—é—Ç—Å—è, –Ω–æ –ø–æ–º–æ–≥–∞—é—Ç —Ä–∞–∑—Ä–∞–±–æ—Ç—á–∏–∫–∞–º)
def multiply(a: float, b: float) -> float:
    """–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –¥–≤—É—Ö —á–∏—Å–µ–ª."""
    return a * b

print(f"2.5 * 3.5 = {multiply(2.5, 3.5)}")
```

```python
# –õ—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü–∏–∏ - –∞–Ω–æ–Ω–∏–º–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É
# –°–∏–Ω—Ç–∞–∫—Å–∏—Å: lambda –ø–∞—Ä–∞–º–µ—Ç—Ä—ã: –≤—ã—Ä–∞–∂–µ–Ω–∏–µ

# –ü—Ä–æ—Å—Ç–∞—è –ª—è–º–±–¥–∞
square = lambda x: x ** 2
print(f"–ö–≤–∞–¥—Ä–∞—Ç 5: {square(5)}")

# –õ—è–º–±–¥–∞ —Å –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏
add = lambda a, b: a + b
print(f"3 + 7 = {add(3, 7)}")

# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ —Å —Ñ—É–Ω–∫—Ü–∏—è–º–∏ –≤—ã—Å—à–µ–≥–æ –ø–æ—Ä—è–¥–∫–∞
numbers = [1, 2, 3, 4, 5]

# map - –ø—Ä–∏–º–µ–Ω–∏—Ç—å —Ñ—É–Ω–∫—Ü–∏—é –∫–æ –≤—Å–µ–º —ç–ª–µ–º–µ–Ω—Ç–∞–º
squares = list(map(lambda x: x**2, numbers))
print(f"–ö–≤–∞–¥—Ä–∞—Ç—ã: {squares}")

# filter - –æ—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞—Ç—å —ç–ª–µ–º–µ–Ω—Ç—ã
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(f"–ß–µ—Ç–Ω—ã–µ —á–∏—Å–ª–∞: {even_numbers}")

# sorted —Å –∫–ª—é—á–æ–º —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏
students = [
    {"name": "–ê–Ω–Ω–∞", "grade": 90},
    {"name": "–ò–≤–∞–Ω", "grade": 85},
    {"name": "–ú–∞—Ä–∏—è", "grade": 95}
]

# –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø–æ –æ—Ü–µ–Ω–∫–µ
sorted_by_grade = sorted(students, key=lambda s: s["grade"], reverse=True)
print("–°—Ç—É–¥–µ–Ω—Ç—ã –ø–æ –æ—Ü–µ–Ω–∫–∞–º:")
for student in sorted_by_grade:
    print(f"  {student['name']}: {student['grade']}")
```
